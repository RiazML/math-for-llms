Compute the Jacobian of:

**f**(x, y) = [xy + y², x² - y, eˣ]ᵀ

1. Derive the analytical Jacobian matrix
2. Evaluate it at point (1, 2)
3. Verify numerically

```python
# Your solution here
```

### Solution

```python
print("f(x, y) = [xy + y², x² - y, eˣ]ᵀ")
print("\nPartial derivatives:")
print("∂f₁/∂x = y,     ∂f₁/∂y = x + 2y")
print("∂f₂/∂x = 2x,    ∂f₂/∂y = -1")
print("∂f₃/∂x = eˣ,    ∂f₃/∂y = 0")

print("\nJacobian J (3×2):")
print("J = [  y     x+2y  ]")
print("    [ 2x      -1   ]")
print("    [ eˣ       0   ]")
```

```python
# Evaluate at specific point
x, y = 1.0, 2.0

J = np.array([
    [y, x + 2*y],
    [2*x, -1],
    [np.exp(x), 0]
])

print(f"At (x, y) = ({x}, {y}):")
print(f"J = \n{J.round(4)}")
```

```python
# Numerical verification
def f(v):
    x, y = v
    return np.array([x*y + y**2, x**2 - y, np.exp(x)])

h = 1e-7
v = np.array([x, y])
J_numerical = np.zeros((3, 2))

for j in range(2):
    v_plus = v.copy()
    v_plus[j] += h
    J_numerical[:, j] = (f(v_plus) - f(v)) / h

print(f"Numerical Jacobian:\n{J_numerical.round(4)}")
print(f"\nMatch: {np.allclose(J, J_numerical)}")
```

## Exercise 5: Newton's Method

Use Newton's method to find minimum of:

f(x, y) = x² + 4y² + 2x - 8y

Starting from (0, 0)

```python
# Your solution here
```

### Solution

```python
print("f(x, y) = x² + 4y² + 2x - 8y")
print("∇f = [2x + 2, 8y - 8]ᵀ")
print("H = [ 2   0 ]  (constant)")
print("    [ 0   8 ]")

def grad_f(x):
    return np.array([2*x[0] + 2, 8*x[1] - 8])

H = np.array([[2, 0], [0, 8]])
H_inv = np.linalg.inv(H)
```

```python
# Newton's method
x = np.array([0.0, 0.0])

print(f"Starting: x = {x}")
print("\nNewton iteration: x_new = x - H⁻¹∇f")

for i in range(3):
    g = grad_f(x)
    delta = H_inv @ g
    x_new = x - delta
    print(f"Iter {i}: x = {x}, ∇f = {g}, Δx = {-delta.round(4)}, x_new = {x_new}")
    x = x_new

print(f"\nMinimum at: {x}")
print("(Setting ∇f = 0: x = -1, y = 1)")
```

## Exercise 6: Softmax Jacobian Properties

For 3-class softmax, verify that:
1. The Jacobian is symmetric
2. Each row sums to 0
3. Explain why rows sum to 0

```python
# Your solution here
```

### Solution

```python
def softmax(z):
    exp_z = np.exp(z - np.max(z))
    return exp_z / np.sum(exp_z)

def softmax_jacobian(p):
    return np.diag(p) - np.outer(p, p)

z = np.array([1.0, 2.0, 3.0])
p = softmax(z)
J = softmax_jacobian(p)

print(f"z = {z}")
print(f"p = softmax(z) = {p.round(4)}")
print(f"\nJacobian J:\n{J.round(6)}")
```

```python
print("a) Symmetric check:")
print(f"   J = Jᵀ? {np.allclose(J, J.T)}")
print(f"   ||J - Jᵀ|| = {np.linalg.norm(J - J.T):.2e}")

print("\nb) Row sums:")
row_sums = J.sum(axis=1)
print(f"   Row sums = {row_sums.round(10)}")

print("\nc) Column sums:")
col_sums = J.sum(axis=0)
print(f"   Column sums = {col_sums.round(10)}")
```

```python
print("Explanation for row sums = 0:")
print("")
print("Since Σᵢ pᵢ = 1 (probabilities sum to 1), differentiating:")
print("  d/dzⱼ (Σᵢ pᵢ) = 0")
print("  Σᵢ ∂pᵢ/∂zⱼ = 0")
print("")
print("This means sum of each column = 0.")
print("By symmetry, sum of each row = 0 as well.")
```

## Exercise 7: Hessian of Quadratic Form

Show that for f(**x**) = **x**ᵀ**Ax** where **A** is symmetric:

**H** = 2**A**

Verify numerically.

```python
# Your solution here
```

### Solution

```python
print("f(x) = xᵀAx where A is symmetric")
print("")
print("Component form:")
print("f = Σᵢ Σⱼ xᵢ Aᵢⱼ xⱼ")
print("")
print("∂f/∂xₖ = Σⱼ Aₖⱼ xⱼ + Σᵢ xᵢ Aᵢₖ")
print("       = (Ax)ₖ + (Aᵀx)ₖ")
print("       = 2(Ax)ₖ  (since A = Aᵀ)")
print("")
print("So ∇f = 2Ax")
print("")
print("∂²f/∂xₖ∂xₗ = 2Aₖₗ")
print("")
print("Therefore H = 2A")
```

```python
# Numerical verification
A = np.array([[2, 1], [1, 3]])  # Symmetric

def f(x):
    return x @ A @ x

def numerical_hessian(x, h=1e-5):
    n = len(x)
    H = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_pp = x.copy(); x_pp[i] += h; x_pp[j] += h
            x_pm = x.copy(); x_pm[i] += h; x_pm[j] -= h
            x_mp = x.copy(); x_mp[i] -= h; x_mp[j] += h
            x_mm = x.copy(); x_mm[i] -= h; x_mm[j] -= h
            H[i,j] = (f(x_pp) - f(x_pm) - f(x_mp) + f(x_mm)) / (4*h*h)
    return H

x = np.array([1.0, 2.0])
H_numerical = numerical_hessian(x)
H_analytical = 2 * A

print(f"A = \n{A}")
print(f"\n2A = \n{H_analytical}")
print(f"\nNumerical H = \n{H_numerical.round(4)}")
print(f"\nMatch: {np.allclose(H_numerical, H_analytical)}")
```

## Exercise 9: Neural Network Layer Jacobian

For a layer **y** = σ(**Wx** + **b**), compute ∂**y**/∂**x**

Use σ = sigmoid and verify numerically.

```python
# Your solution here
```

### Solution

```python
print("y = σ(Wx + b) = σ(z) where z = Wx + b")
print("")
print("By chain rule:")
print("∂y/∂x = (∂y/∂z)(∂z/∂x)")
print("      = diag(σ'(z)) · W")
print("")
print("For sigmoid σ: σ'(z) = σ(z)(1 - σ(z))")
print("")
print("So ∂y/∂x = diag(y ⊙ (1-y)) · W")
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Layer parameters
W = np.array([[1, 2, 3], [4, 5, 6]])  # 2×3
b = np.array([0.1, 0.2])
x = np.array([1.0, 2.0, 3.0])

# Forward pass
z = W @ x + b
y = sigmoid(z)

# Analytical Jacobian
sigma_prime = y * (1 - y)
J = np.diag(sigma_prime) @ W

print(f"W (2×3):\n{W}")
print(f"\nx = {x}")
print(f"z = Wx + b = {z.round(4)}")
print(f"y = σ(z) = {y.round(4)}")
print(f"σ'(z) = {sigma_prime.round(4)}")
print(f"\n∂y/∂x (2×3):\n{J.round(4)}")
```

```python
# Numerical verification
h = 1e-7
J_num = np.zeros((2, 3))

for j in range(3):
    x_plus = x.copy()
    x_plus[j] += h
    y_plus = sigmoid(W @ x_plus + b)
    J_num[:, j] = (y_plus - y) / h

print(f"Numerical Jacobian:\n{J_num.round(4)}")
print(f"\nMatch: {np.allclose(J, J_num)}")
```

## Exercise 10: Hessian Eigenvalue Analysis

For f(x, y) = x⁴ - 2x²y² + y⁴ = (x² - y²)²:
1. Find all critical points
2. Analyze Hessian eigenvalues at each
3. Classify the critical points

```python
# Your solution here
```

### Solution

```python
print("f(x, y) = x⁴ - 2x²y² + y⁴ = (x² - y²)²")
print("")
print("∂f/∂x = 4x³ - 4xy² = 4x(x² - y²)")
print("∂f/∂y = -4x²y + 4y³ = 4y(y² - x²)")
print("")
print("a) Critical points where ∇f = 0:")
print("  4x(x² - y²) = 0 → x = 0 or x² = y²")
print("  4y(y² - x²) = 0 → y = 0 or y² = x²")
print("")
print("Critical points: (0, 0) and all points on x = ±y")
```

```python
print("b) Hessian:")
print("∂²f/∂x² = 12x² - 4y²")
print("∂²f/∂y² = -4x² + 12y²")
print("∂²f/∂x∂y = -8xy")

def hessian(x, y):
    return np.array([
        [12*x**2 - 4*y**2, -8*x*y],
        [-8*x*y, -4*x**2 + 12*y**2]
    ])

# Representative points
points = [(0, 0), (1, 1), (1, -1), (2, 2)]
```

```python
print("\nAnalysis at key points:")
print("=" * 60)

for pt in points:
    x, y = pt
    H = hessian(x, y)
    eigs = np.linalg.eigvals(H)
    
    if np.all(eigs > 1e-10):
        classification = "Local Minimum"
    elif np.all(eigs < -1e-10):
        classification = "Local Maximum"
    elif np.all(np.abs(eigs) < 1e-10):
        classification = "Degenerate (H = 0)"
    else:
        classification = "Saddle Point"
    
    print(f"\n({x}, {y}):")
    print(f"  H = \n    {H}")
    print(f"  Eigenvalues: {eigs.round(4)}")
    print(f"  Classification: {classification}")
```

```python
# Visualize
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = (X**2 - Y**2)**2

plt.figure(figsize=(10, 8))
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.plot([-2, 2], [-2, 2], 'r--', label='x = y (global min, f=0)')
plt.plot([-2, 2], [2, -2], 'r--', label='x = -y (global min, f=0)')
plt.plot(0, 0, 'ko', markersize=10, label='Origin (degenerate)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x, y) = (x² - y²)²')
plt.legend()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Summary

### Key Takeaways

1. **Jacobian**: Matrix of first-order partial derivatives for vector-valued functions
   - J_ij = ∂f_i/∂x_j
   - Chain rule: J_{f∘g} = J_f · J_g

2. **Hessian**: Matrix of second-order partial derivatives for scalar functions
   - H_ij = ∂²f/∂x_i∂x_j
   - Always symmetric (Clairaut's theorem)

3. **Critical Point Classification**:
   - All eigenvalues > 0: Local minimum
   - All eigenvalues < 0: Local maximum
   - Mixed signs: Saddle point
   - Zero eigenvalues: Degenerate (need higher-order analysis)

4. **Newton's Method**: x_{k+1} = x_k - H⁻¹∇f
   - Uses Hessian for faster convergence
   - Quadratic convergence for well-behaved problems
