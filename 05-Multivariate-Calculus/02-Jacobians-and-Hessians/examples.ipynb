# Jacobians and Hessians - Examples

Practical demonstrations of Jacobian and Hessian matrices and their applications in machine learning.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Example 1: Basic Jacobian Computation

For a vector-valued function **f**: ℝⁿ → ℝᵐ, the Jacobian is an m×n matrix of partial derivatives.

Consider f(x, y) = [x² + y, xy, x - y²]ᵀ

```python
print("For f(x, y) = [x² + y, xy, x - y²]ᵀ")
print("\nPartial derivatives:")
print("∂f₁/∂x = 2x,  ∂f₁/∂y = 1")
print("∂f₂/∂x = y,   ∂f₂/∂y = x")
print("∂f₃/∂x = 1,   ∂f₃/∂y = -2y")

print("\nJacobian matrix (3×2):")
print("J = [ 2x    1  ]")
print("    [  y    x  ]")
print("    [  1   -2y ]")
```

```python
def f(v):
    """Vector-valued function."""
    x, y = v
    return np.array([x**2 + y, x*y, x - y**2])

def jacobian_analytic(x, y):
    """Analytical Jacobian."""
    return np.array([
        [2*x, 1],
        [y, x],
        [1, -2*y]
    ])

# Compute at specific point
point = (2.0, 3.0)
J = jacobian_analytic(*point)

print(f"At point {point}:")
print(f"J = \n{J}")
```

```python
# Numerical verification
h = 1e-7
J_numerical = np.zeros((3, 2))
v = np.array(point)

for j in range(2):
    v_plus = v.copy()
    v_plus[j] += h
    J_numerical[:, j] = (f(v_plus) - f(v)) / h

print(f"Numerical Jacobian:\n{J_numerical.round(4)}")
print(f"\nMatch: {np.allclose(J, J_numerical)}")
```

## Example 2: Jacobian Chain Rule

For composed functions **f** ∘ **g**, the chain rule becomes matrix multiplication:

$$\mathbf{J}_{\mathbf{f} \circ \mathbf{g}} = \mathbf{J}_{\mathbf{f}} \cdot \mathbf{J}_{\mathbf{g}}$$

```python
print("g: ℝ² → ℝ²: g(x, y) = [x + y, xy]")
print("f: ℝ² → ℝ²: f(u, v) = [u², uv]")
print("Composition: (f ∘ g)(x, y)")

print("\nJacobian of g:")
print("J_g = [ 1   1 ]")
print("      [ y   x ]")

print("\nJacobian of f (at g(x,y)):")
print("J_f = [ 2u    0 ]")
print("      [  v    u ]")

print("\nChain rule: J_{f∘g} = J_f × J_g")
```

```python
x, y = 2.0, 3.0

# g and its Jacobian
u, v = x + y, x * y  # g(x, y)
J_g = np.array([[1, 1], [y, x]])

# f and its Jacobian at g(x,y)
J_f = np.array([[2*u, 0], [v, u]])

# Chain rule: multiply Jacobians
J_composition = J_f @ J_g

print(f"At (x, y) = ({x}, {y}):")
print(f"g(x, y) = ({u}, {v})")
print(f"\nJ_g = \n{J_g}")
print(f"\nJ_f = \n{J_f}")
print(f"\nJ_f × J_g = \n{J_composition}")
```

```python
# Verify by direct numerical computation
def composition(v):
    x, y = v
    u = x + y
    w = x * y
    return np.array([u**2, u*w])

v_in = np.array([x, y])
h = 1e-7
J_direct = np.zeros((2, 2))

for j in range(2):
    v_plus = v_in.copy()
    v_plus[j] += h
    J_direct[:, j] = (composition(v_plus) - composition(v_in)) / h

print(f"Direct numerical Jacobian:\n{J_direct.round(4)}")
print(f"\nChain rule result matches: {np.allclose(J_composition, J_direct)}")
```

## Example 3: Basic Hessian Computation

For a scalar function f: ℝⁿ → ℝ, the Hessian is the matrix of second partial derivatives:

$$H_{ij} = \frac{\partial^2 f}{\partial x_i \partial x_j}$$

```python
print("f(x, y) = x³ - 2xy + y²")
print("\nFirst derivatives:")
print("∂f/∂x = 3x² - 2y")
print("∂f/∂y = -2x + 2y")

print("\nSecond derivatives:")
print("∂²f/∂x² = 6x")
print("∂²f/∂y² = 2")
print("∂²f/∂x∂y = -2")

print("\nHessian:")
print("H = [ 6x   -2 ]")
print("    [ -2    2 ]")
```

```python
def f(v):
    x, y = v
    return x**3 - 2*x*y + y**2

def hessian_analytic(x, y):
    return np.array([
        [6*x, -2],
        [-2, 2]
    ])

point = (1.0, 2.0)
H = hessian_analytic(*point)

print(f"At point {point}:")
print(f"H = \n{H}")

# Check symmetry (Clairaut's theorem)
print(f"\nSymmetric? {np.allclose(H, H.T)}")

# Eigenvalues determine curvature
eigenvalues = np.linalg.eigvals(H)
print(f"Eigenvalues: {eigenvalues.round(4)}")
```

## Example 4: Critical Point Classification

At a critical point (where ∇f = 0), the Hessian determines the type:

| Hessian Property | Point Type |
|-----------------|------------|
| All positive eigenvalues | Local minimum |
| All negative eigenvalues | Local maximum |
| Mixed signs | Saddle point |

```python
print("Example 1: f(x, y) = x² - y²")
print("\n∇f = [2x, -2y]ᵀ = 0 at (0, 0)")
print("\nH = [ 2   0 ]")
print("    [ 0  -2 ]")

H = np.array([[2, 0], [0, -2]])
eigenvalues = np.linalg.eigvals(H)

print(f"\nEigenvalues: {eigenvalues}")
print("One positive, one negative → SADDLE POINT")
```

```python
print("Example 2: f(x, y) = x² + y²")
print("H = [ 2   0 ]")
print("    [ 0   2 ]")

H2 = np.array([[2, 0], [0, 2]])
eigenvalues2 = np.linalg.eigvals(H2)
print(f"Eigenvalues: {eigenvalues2}")
print("All positive → LOCAL MINIMUM")
```

```python
print("Example 3: f(x, y) = -x² - y²")
print("H = [ -2   0 ]")
print("    [  0  -2 ]")

H3 = np.array([[-2, 0], [0, -2]])
eigenvalues3 = np.linalg.eigvals(H3)
print(f"Eigenvalues: {eigenvalues3}")
print("All negative → LOCAL MAXIMUM")
```

## Example 5: Newton's Method for Optimization

Newton's method uses second-order information (Hessian) for faster convergence:

$$\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}^{-1} \nabla f(\mathbf{x}_k)$$

For quadratic functions, Newton converges in one step!

```python
print("Minimize f(x, y) = (x - 1)² + 2(y - 2)²")
print("\n∇f = [2(x-1), 4(y-2)]ᵀ")
print("H = [ 2   0 ]  (constant)")
print("    [ 0   4 ]")

def f(v):
    x, y = v
    return (x - 1)**2 + 2*(y - 2)**2

def grad_f(v):
    x, y = v
    return np.array([2*(x - 1), 4*(y - 2)])

def hessian_f(v):
    return np.array([[2, 0], [0, 4]])
```

```python
# Newton's method
x = np.array([0.0, 0.0])
history = [x.copy()]

print(f"Starting point: {x}")
print(f"{'Iter':<6} {'x':^20} {'f(x)':^12} {'||∇f||':^12}")
print("-" * 52)

for i in range(5):
    g = grad_f(x)
    H = hessian_f(x)
    
    print(f"{i:<6} [{x[0]:.4f}, {x[1]:.4f}] {f(x):^12.6f} {np.linalg.norm(g):^12.6f}")
    
    # Newton step: x_new = x - H^(-1) * g
    delta = np.linalg.solve(H, g)
    x = x - delta
    history.append(x.copy())

print(f"\nFinal: {x} (optimal is [1, 2])")
print("\nNote: For quadratic functions, Newton converges in 1 step!")
```

## Example 6: Newton's Method on Rosenbrock Function

The Rosenbrock function f(x, y) = (1-x)² + 100(y-x²)² is a classic optimization test problem with a curved valley.

```python
def rosenbrock(v):
    x, y = v
    return (1 - x)**2 + 100*(y - x**2)**2

def grad_rosenbrock(v):
    x, y = v
    dx = -2*(1 - x) - 400*x*(y - x**2)
    dy = 200*(y - x**2)
    return np.array([dx, dy])

def hessian_rosenbrock(v):
    x, y = v
    h11 = 2 - 400*(y - x**2) + 800*x**2
    h12 = -400*x
    h22 = 200
    return np.array([[h11, h12], [h12, h22]])
```

```python
# Newton's method with damping
x = np.array([0.0, 0.0])
history = [x.copy()]

print(f"Starting: x = {x}, f(x) = {rosenbrock(x)}")
print(f"\n{'Iter':<6} {'x':^24} {'f(x)':^14}")
print("-" * 46)

for i in range(20):
    g = grad_rosenbrock(x)
    H = hessian_rosenbrock(x)
    
    # Add regularization if Hessian is nearly singular
    H = H + 0.1 * np.eye(2)
    
    delta = np.linalg.solve(H, g)
    
    # Line search (damped Newton)
    alpha = 1.0
    while rosenbrock(x - alpha*delta) > rosenbrock(x) and alpha > 1e-10:
        alpha *= 0.5
    
    x = x - alpha * delta
    history.append(x.copy())
    
    if i < 5 or i % 5 == 0:
        print(f"{i:<6} [{x[0]:.6f}, {x[1]:.6f}] {rosenbrock(x):^14.8f}")

print(f"\nFinal: [{x[0]:.6f}, {x[1]:.6f}]")
print("Minimum at: [1, 1]")
```

```python
# Visualize optimization path
history = np.array(history)

x_range = np.linspace(-0.5, 1.5, 100)
y_range = np.linspace(-0.5, 1.5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = (1 - X)**2 + 100*(Y - X**2)**2

plt.figure(figsize=(10, 8))
plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20), cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.plot(history[:, 0], history[:, 1], 'r.-', markersize=10, linewidth=2, label='Newton path')
plt.plot(1, 1, 'g*', markersize=15, label='Minimum (1, 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Newton\'s Method on Rosenbrock Function')
plt.legend()
plt.tight_layout()
plt.show()
```

## Example 7: Softmax Jacobian

For the softmax function p_i = exp(z_i) / Σexp(z_j), the Jacobian has a special structure:

$$J_{ij} = \frac{\partial p_i}{\partial z_j} = p_i(\delta_{ij} - p_j)$$

In matrix form: **J** = diag(**p**) - **p p**ᵀ

```python
def softmax(z):
    """Numerically stable softmax."""
    exp_z = np.exp(z - np.max(z))
    return exp_z / np.sum(exp_z)

def softmax_jacobian(p):
    """Compute Jacobian from softmax output p."""
    return np.diag(p) - np.outer(p, p)

z = np.array([1.0, 2.0, 3.0])
p = softmax(z)
J = softmax_jacobian(p)

print(f"Input z = {z}")
print(f"Softmax p = {p.round(4)}")
print(f"\nJacobian (3×3):\n{J.round(4)}")
```

```python
# Properties of softmax Jacobian
print("Properties of Softmax Jacobian:\n")

print(f"1. Symmetric: J = Jᵀ? {np.allclose(J, J.T)}")
print(f"   ||J - Jᵀ|| = {np.linalg.norm(J - J.T):.2e}")

print(f"\n2. Row sums: {J.sum(axis=1).round(10)}")
print("   (≈ 0 because Σpᵢ = 1, so Σ ∂pᵢ/∂zⱼ = 0)")

print(f"\n3. Column sums: {J.sum(axis=0).round(10)}")
print("   (Also ≈ 0 by symmetry)")
```

```python
# Numerical verification
h = 1e-7
J_numerical = np.zeros((3, 3))
for j in range(3):
    z_plus = z.copy()
    z_plus[j] += h
    J_numerical[:, j] = (softmax(z_plus) - softmax(z)) / h

print("Numerical Jacobian:")
print(J_numerical.round(4))
print(f"\nMatch with analytical: {np.allclose(J, J_numerical)}")
```

## Example 8: Loss Landscape Analysis via Hessian Eigenvalues

Hessian eigenvalues reveal the curvature of the loss landscape:
- Large eigenvalues → steep directions (fast learning)
- Small eigenvalues → flat directions (slow learning)
- Negative eigenvalues → saddle points

```python
print("f(x, y) = x⁴ + y⁴ - 4xy + 1")
print("(This function has interesting critical points)")

def f(v):
    x, y = v
    return x**4 + y**4 - 4*x*y + 1

def grad_f(v):
    x, y = v
    return np.array([4*x**3 - 4*y, 4*y**3 - 4*x])

def hessian_f(v):
    x, y = v
    return np.array([
        [12*x**2, -4],
        [-4, 12*y**2]
    ])
```

```python
# Analyze critical points
critical_points = [(0, 0), (1, 1), (-1, -1)]

print("Critical Point Analysis:")
print("=" * 60)

for point in critical_points:
    x = np.array(point)
    f_val = f(x)
    g = grad_f(x)
    H = hessian_f(x)
    eigenvalues = np.linalg.eigvals(H)
    
    # Classification
    if np.all(eigenvalues > 0):
        point_type = "Local Minimum"
    elif np.all(eigenvalues < 0):
        point_type = "Local Maximum"
    else:
        point_type = "Saddle Point"
    
    print(f"\nPoint {point}:")
    print(f"  f(x) = {f_val:.4f}")
    print(f"  ||∇f|| = {np.linalg.norm(g):.6f}")
    print(f"  Hessian eigenvalues: {eigenvalues.round(4)}")
    print(f"  Classification: {point_type}")
```

## Example 9: Hessian of Quadratic Form

For a quadratic form f(**x**) = **x**ᵀ**Ax** + **b**ᵀ**x** + c:
- Gradient: ∇f = (**A** + **A**ᵀ)**x** + **b**
- Hessian: **H** = **A** + **A**ᵀ

If **A** is symmetric: **H** = 2**A**

```python
# Example with symmetric A
A = np.array([[2, 1], [1, 3]])
b = np.array([1, 2])
c = 5

print(f"A = \n{A}")
print(f"b = {b}")
print(f"c = {c}")

def f(x):
    return x @ A @ x + b @ x + c

# Analytical Hessian
H_analytic = A + A.T
print(f"\nAnalytical H = A + Aᵀ = \n{H_analytic}")
```

```python
# Numerical Hessian verification
x = np.array([1.0, 2.0])
h = 1e-5
H_numerical = np.zeros((2, 2))

for i in range(2):
    for j in range(2):
        x_pp = x.copy(); x_pp[i] += h; x_pp[j] += h
        x_pm = x.copy(); x_pm[i] += h; x_pm[j] -= h
        x_mp = x.copy(); x_mp[i] -= h; x_mp[j] += h
        x_mm = x.copy(); x_mm[i] -= h; x_mm[j] -= h
        H_numerical[i,j] = (f(x_pp) - f(x_pm) - f(x_mp) + f(x_mm)) / (4*h*h)

print(f"Numerical Hessian:\n{H_numerical.round(4)}")
print(f"\nMatch: {np.allclose(H_analytic, H_numerical)}")
```

## Example 10: Backpropagation as Jacobian Products

Backpropagation is essentially computing vector-Jacobian products (VJPs):

$$\frac{\partial L}{\partial \mathbf{x}} = \mathbf{J}^T \frac{\partial L}{\partial \mathbf{y}}$$

For a layer **y** = **f**(**x**), we propagate gradients backward using the transpose Jacobian.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

# Network architecture
n_input = 3
n_hidden = 4
n_output = 2

# Initialize weights
np.random.seed(42)
x = np.random.randn(n_input)
t = np.random.randn(n_output)  # target
W1 = np.random.randn(n_hidden, n_input) * 0.1
b1 = np.zeros(n_hidden)
W2 = np.random.randn(n_output, n_hidden) * 0.1
b2 = np.zeros(n_output)

print("Two-layer network: x → h → y → L")
print("h = σ(W₁x + b₁)")
print("y = W₂h + b₂")
print("L = ||y - t||²")
```

```python
# Forward pass
z1 = W1 @ x + b1
h = sigmoid(z1)
z2 = W2 @ h + b2
y = z2  # Linear output
L = np.sum((y - t)**2)

print("Forward pass:")
print(f"  x shape: {x.shape}")
print(f"  h shape: {h.shape}")
print(f"  y shape: {y.shape}")
print(f"  Loss L = {L:.4f}")
```

```python
# Backward pass (using Jacobians)
print("Backward pass using Jacobians:")

# ∂L/∂y
dL_dy = 2 * (y - t)  # Shape: (n_output,)
print(f"\n∂L/∂y shape: {dL_dy.shape}")

# Jacobian of y = W₂h w.r.t. h is W₂ (n_output × n_hidden)
# ∂L/∂h = (∂y/∂h)ᵀ × ∂L/∂y = W₂ᵀ × ∂L/∂y
J_y_h = W2  # Shape: (n_output, n_hidden)
dL_dh = J_y_h.T @ dL_dy  # Shape: (n_hidden,)
print(f"J_y_h (∂y/∂h) shape: {J_y_h.shape}")
print(f"∂L/∂h = J_y_hᵀ × ∂L/∂y, shape: {dL_dh.shape}")

# Jacobian of h = σ(z₁) is diag(σ'(z₁)) (n_hidden × n_hidden)
sigma_prime = h * (1 - h)  # Diagonal elements
dL_dz1 = dL_dh * sigma_prime  # Element-wise (implicit diag mult)
print(f"∂L/∂z₁ shape: {dL_dz1.shape}")
```

```python
# Weight gradients
dL_dW2 = np.outer(dL_dy, h)  # Shape: (n_output, n_hidden)
dL_db2 = dL_dy
dL_dW1 = np.outer(dL_dz1, x)  # Shape: (n_hidden, n_input)
dL_db1 = dL_dz1

print("Gradient shapes:")
print(f"  ∂L/∂W₂: {dL_dW2.shape} (same as W₂)")
print(f"  ∂L/∂b₂: {dL_db2.shape} (same as b₂)")
print(f"  ∂L/∂W₁: {dL_dW1.shape} (same as W₁)")
print(f"  ∂L/∂b₁: {dL_db1.shape} (same as b₁)")
```

## Example 11: Numerical Hessian Computation

For complex functions, we can compute the Hessian numerically using finite differences:

$$H_{ij} \approx \frac{f(x + h e_i + h e_j) - f(x + h e_i - h e_j) - f(x - h e_i + h e_j) + f(x - h e_i - h e_j)}{4h^2}$$

```python
def f(v):
    """A complex function."""
    x, y = v
    return np.sin(x*y) + x**2 + y**3

def numerical_hessian(f, x, h=1e-5):
    """Compute Hessian using finite differences."""
    n = len(x)
    H = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            x_pp = x.copy(); x_pp[i] += h; x_pp[j] += h
            x_pm = x.copy(); x_pm[i] += h; x_pm[j] -= h
            x_mp = x.copy(); x_mp[i] -= h; x_mp[j] += h
            x_mm = x.copy(); x_mm[i] -= h; x_mm[j] -= h
            
            H[i, j] = (f(x_pp) - f(x_pm) - f(x_mp) + f(x_mm)) / (4 * h * h)
    
    return H

x = np.array([1.0, 2.0])
H = numerical_hessian(f, x)

print("f(x, y) = sin(xy) + x² + y³")
print(f"\nAt point {x}:")
print(f"Numerical Hessian:\n{H.round(4)}")
```

```python
# Analytical Hessian for comparison
# ∂f/∂x = y·cos(xy) + 2x
# ∂f/∂y = x·cos(xy) + 3y²
# ∂²f/∂x² = -y²·sin(xy) + 2
# ∂²f/∂y² = -x²·sin(xy) + 6y
# ∂²f/∂x∂y = cos(xy) - xy·sin(xy)

x_val, y_val = x
H_analytic = np.array([
    [-y_val**2 * np.sin(x_val*y_val) + 2, 
     np.cos(x_val*y_val) - x_val*y_val*np.sin(x_val*y_val)],
    [np.cos(x_val*y_val) - x_val*y_val*np.sin(x_val*y_val), 
     -x_val**2 * np.sin(x_val*y_val) + 6*y_val]
])

print("Analytical Hessian:")
print(H_analytic.round(4))
print(f"\nNumerical matches analytical: {np.allclose(H, H_analytic, atol=1e-4)}")
```

## Summary

### Key Matrices

| Matrix | Dimension | Definition |
|--------|-----------|------------|
| Jacobian | m × n | J_ij = ∂f_i/∂x_j |
| Hessian | n × n | H_ij = ∂²f/∂x_i∂x_j |

### Key Results

| Concept | Formula |
|---------|--------|
| Chain rule | J_{f∘g} = J_f · J_g |
| Taylor expansion | f(x+Δ) ≈ f + ∇fᵀΔ + ½ΔᵀHΔ |
| Newton's method | x_{k+1} = x_k - H⁻¹∇f |
| Backpropagation | ∂L/∂x = Jᵀ × ∂L/∂y |

### ML Applications

- **Jacobian**: Backpropagation, softmax gradient, normalizing flows
- **Hessian**: Loss landscape curvature, Newton's method, saddle point detection
