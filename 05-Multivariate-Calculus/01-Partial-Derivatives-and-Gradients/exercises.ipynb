## Exercise 5: Linear Regression Gradient

For MSE loss $L(w, b) = \frac{1}{n} \sum_i(wx_i + b - y_i)^2$

Derive $\frac{\partial L}{\partial w}$ and $\frac{\partial L}{\partial b}$, then implement gradient descent.

```python
# Your solution here

# Generate data
np.random.seed(42)
n = 50
x = np.random.randn(n)
y = 2*x + 3 + 0.1*np.random.randn(n)  # True: w=2, b=3

# Implement gradient descent
```

### Solution

```python
print("EXERCISE 5: Linear Regression Gradient")
print("="*60)

print("L(w, b) = (1/n) Σ(wx_i + b - y_i)²")

print("\nLet r_i = wx_i + b - y_i (residual)")
print("L = (1/n) Σ r_i²")

print("\n∂L/∂w = (1/n) Σ 2r_i · ∂r_i/∂w")
print("      = (1/n) Σ 2r_i · x_i")
print("      = (2/n) Σ (wx_i + b - y_i) · x_i")

print("\n∂L/∂b = (1/n) Σ 2r_i · ∂r_i/∂b")
print("      = (1/n) Σ 2r_i · 1")
print("      = (2/n) Σ (wx_i + b - y_i)")

# Example
np.random.seed(42)
n = 50
x = np.random.randn(n)
y = 2*x + 3 + 0.1*np.random.randn(n)

w, b = 0.0, 0.0
lr = 0.1

print("\n--- Implementation ---")
for i in range(100):
    pred = w*x + b
    residual = pred - y
    dw = (2/n) * np.sum(residual * x)
    db = (2/n) * np.sum(residual)
    w -= lr * dw
    b -= lr * db

print(f"Final w = {w:.4f} (true: 2)")
print(f"Final b = {b:.4f} (true: 3)")
```

## Exercise 6: Softmax Gradient

For softmax: $p_i = \frac{e^{z_i}}{\sum_j e^{z_j}}$

Show that $\frac{\partial p_i}{\partial z_j} = p_i(\delta_{ij} - p_j)$

where $\delta_{ij}$ is Kronecker delta (1 if i=j, 0 otherwise).

```python
# Your solution here

def softmax(z):
    exp_z = np.exp(z - np.max(z))  # Numerically stable
    return exp_z / np.sum(exp_z)

# Verify numerically
```

### Solution

```python
print("EXERCISE 6: Softmax Gradient")
print("="*60)

print("Softmax: p_i = exp(z_i) / S where S = Σ exp(z_j)")

print("\nCase 1: i = j")
print("∂p_i/∂z_i = [exp(z_i)·S - exp(z_i)·exp(z_i)] / S²")
print("         = exp(z_i)/S - exp(z_i)²/S²")
print("         = p_i - p_i²")
print("         = p_i(1 - p_i)")
print("         = p_i(δ_ii - p_i) ✓")

print("\nCase 2: i ≠ j")
print("∂p_i/∂z_j = [0 - exp(z_i)·exp(z_j)] / S²")
print("         = -exp(z_i)·exp(z_j) / S²")
print("         = -p_i · p_j")
print("         = p_i(0 - p_j)")
print("         = p_i(δ_ij - p_j) ✓")

print("\nCombined: ∂p_i/∂z_j = p_i(δ_ij - p_j)")

# Numerical verification
def softmax(z):
    exp_z = np.exp(z - np.max(z))
    return exp_z / np.sum(exp_z)

z = np.array([1.0, 2.0, 3.0])
p = softmax(z)

print("\n--- Numerical verification ---")
print(f"z = {z}, p = {p.round(4)}")

h = 1e-7
for i in range(3):
    for j in range(3):
        z_plus = z.copy()
        z_plus[j] += h
        z_minus = z.copy()
        z_minus[j] -= h
        numerical = (softmax(z_plus)[i] - softmax(z_minus)[i]) / (2*h)
        analytic = p[i] * ((1 if i==j else 0) - p[j])
        print(f"  ∂p_{i}/∂z_{j}: analytic = {analytic:.6f}, numerical = {numerical:.6f}")
```

## Exercise 7: Multivariate Chain Rule

Let $f(x, y) = x^2 + y^2$, where $x = r\cos(\theta)$, $y = r\sin(\theta)$

Find $\frac{\partial f}{\partial r}$ and $\frac{\partial f}{\partial \theta}$ using the chain rule.

```python
# Your solution here
```

### Solution

```python
print("EXERCISE 7: Multivariate Chain Rule")
print("="*60)

print("f(x, y) = x² + y²")
print("x = r cos(θ), y = r sin(θ)")

print("\nUsing chain rule:")
print("∂f/∂r = (∂f/∂x)(∂x/∂r) + (∂f/∂y)(∂y/∂r)")
print("      = 2x·cos(θ) + 2y·sin(θ)")
print("      = 2r·cos²(θ) + 2r·sin²(θ)")
print("      = 2r")

print("\n∂f/∂θ = (∂f/∂x)(∂x/∂θ) + (∂f/∂y)(∂y/∂θ)")
print("      = 2x·(-r·sin(θ)) + 2y·(r·cos(θ))")
print("      = -2r²·cos(θ)·sin(θ) + 2r²·sin(θ)·cos(θ)")
print("      = 0")

print("\nVerification: f = x² + y² = r², so:")
print("∂f/∂r = 2r ✓")
print("∂f/∂θ = 0 ✓ (f doesn't depend on θ)")
```

## Exercise 8: Gradient Checking

Implement gradient checking for:
$$f(x, y, z) = x \cdot \sin(y) + z^2$$

Compare analytical and numerical gradients at point $(2, \pi/4, 1)$.

```python
# Your solution here

def f(x):
    pass  # TODO

def grad_f_analytic(x):
    pass  # TODO

def grad_f_numerical(x, h=1e-7):
    pass  # TODO
```

### Solution

```python
print("EXERCISE 8: Gradient Checking")
print("="*60)

def f(x):
    return x[0] * np.sin(x[1]) + x[2]**2

def grad_f_analytic(x):
    return np.array([
        np.sin(x[1]),        # ∂f/∂x
        x[0] * np.cos(x[1]), # ∂f/∂y
        2 * x[2]             # ∂f/∂z
    ])

def grad_f_numerical(x, h=1e-7):
    grad = np.zeros(3)
    for i in range(3):
        x_plus = x.copy()
        x_plus[i] += h
        x_minus = x.copy()
        x_minus[i] -= h
        grad[i] = (f(x_plus) - f(x_minus)) / (2*h)
    return grad

x = np.array([2.0, np.pi/4, 1.0])

g_analytic = grad_f_analytic(x)
g_numerical = grad_f_numerical(x)

print(f"Point: x = {x}")
print(f"\nAnalytical gradient: {g_analytic}")
print(f"Numerical gradient:  {g_numerical}")

rel_error = np.linalg.norm(g_analytic - g_numerical) / (
    np.linalg.norm(g_analytic) + np.linalg.norm(g_numerical) + 1e-10
)
print(f"\nRelative error: {rel_error:.2e}")
print(f"Check {'PASSED' if rel_error < 1e-5 else 'FAILED'}")
```

Derive $\nabla_{\mathbf{w}} L$ using matrix calculus. Also find the closed-form solution.

```python
# Your solution here
```

### Solution

```python
print("EXERCISE 9: MSE Matrix Gradient")
print("="*60)

print("L(w) = ||Xw - y||² = (Xw - y)ᵀ(Xw - y)")
print("\nExpanding:")
print("L = wᵀXᵀXw - 2yᵀXw + yᵀy")

print("\nUsing matrix calculus rules:")
print("  ∂(wᵀAw)/∂w = (A + Aᵀ)w")
print("  ∂(bᵀw)/∂w = b")

print("\n∇_w L = 2XᵀXw - 2Xᵀy")
print("      = 2Xᵀ(Xw - y)")

print("\nSetting ∇_w L = 0:")
print("XᵀXw = Xᵀy")
print("w* = (XᵀX)⁻¹Xᵀy  (Normal equation)")

# Numerical verification
np.random.seed(42)
X = np.random.randn(100, 3)
w_true = np.array([1, 2, 3])
y = X @ w_true + 0.1*np.random.randn(100)

# Analytical solution
w_closed = np.linalg.solve(X.T @ X, X.T @ y)
print(f"\n--- Verification ---")
print(f"True w: {w_true}")
print(f"Estimated w (closed form): {w_closed.round(4)}")
```

## Exercise 10: Neural Network Layer Gradient

For a neural network layer:
- Forward: $\mathbf{z} = W\mathbf{x} + \mathbf{b}$, $\mathbf{y} = \sigma(\mathbf{z})$
- Loss: $L = \|\mathbf{y} - \mathbf{t}\|^2$

Derive $\frac{\partial L}{\partial W}$ and $\frac{\partial L}{\partial \mathbf{b}}$ using chain rule.

```python
# Your solution here
```

### Solution

```python
print("EXERCISE 10: Neural Network Layer Gradient")
print("="*60)

print("Forward: z = Wx + b")
print("         y = σ(z)")
print("         L = ||y - t||²")

print("\nBackward (chain rule):")
print("∂L/∂y = 2(y - t)")
print("∂y/∂z = σ'(z) = σ(z)(1 - σ(z))  [for sigmoid]")
print("∂L/∂z = ∂L/∂y ⊙ ∂y/∂z = 2(y - t) ⊙ σ'(z)")

print("\nLet δ = ∂L/∂z (error signal)")

print("\n∂z/∂W: z_i = Σ_j W_ij x_j + b_i")
print("       ∂z_i/∂W_ij = x_j")
print("∂L/∂W_ij = δ_i · x_j")
print("∂L/∂W = δ · xᵀ  (outer product)")

print("\n∂L/∂b = δ")

# Implementation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

np.random.seed(42)
x = np.array([1.0, 2.0])
t = np.array([0.8, 0.2])
W = np.random.randn(2, 2) * 0.1
b = np.zeros(2)

# Forward
z = W @ x + b
y = sigmoid(z)
L = np.sum((y - t)**2)

# Backward
dL_dy = 2 * (y - t)
dy_dz = sigmoid(z) * (1 - sigmoid(z))
delta = dL_dy * dy_dz

dL_dW = np.outer(delta, x)
dL_db = delta

print("\n--- Implementation ---")
print(f"x = {x}")
print(f"y = {y.round(4)}")
print(f"t = {t}")
print(f"Loss = {L:.4f}")
print(f"∂L/∂W =\n{dL_dW.round(4)}")
print(f"∂L/∂b = {dL_db.round(4)}")
```

## Summary

| Concept | Key Formula | Use Case |
|---------|-------------|----------|
| Partial derivative | $\partial f/\partial x_i$ | One variable at a time |
| Gradient | $\nabla f$ | All partial derivatives |
| Directional derivative | $D_u f = \nabla f \cdot u$ | Rate in direction $u$ |
| Gradient descent | $x_{k+1} = x_k - \eta \nabla f$ | Optimization |
| Chain rule | $\frac{dL}{dw} = \frac{dL}{dy}\frac{dy}{dz}\frac{dz}{dw}$ | Backpropagation |
| Softmax gradient | $\partial p_i/\partial z_j = p_i(\delta_{ij} - p_j)$ | Classification |
| Normal equation | $w^* = (X^TX)^{-1}X^Ty$ | Closed-form linear regression |
