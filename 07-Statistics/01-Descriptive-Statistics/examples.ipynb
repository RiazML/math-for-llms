```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.style.use('seaborn-v0_8-whitegrid')
np.set_printoptions(precision=6, suppress=True)
```

## 1. Measures of Central Tendency

**Mean**: $\bar{x} = \frac{1}{n}\sum_{i=1}^n x_i$ - Sensitive to outliers

**Median**: Middle value when sorted - Robust to outliers

**Mode**: Most frequent value - Used for categorical data

```python
print("MEASURES OF CENTRAL TENDENCY")
print("="*60)

np.random.seed(42)

# Symmetric data
symmetric = np.random.normal(50, 10, 1000)

print("\nSymmetric data (Normal):")
print(f"  Mean:   {symmetric.mean():.2f}")
print(f"  Median: {np.median(symmetric):.2f}")
print(f"  Mode:   {stats.mode(symmetric.round(), keepdims=True)[0][0]:.2f}")
print("  → Mean ≈ Median for symmetric data")

# Right-skewed data
skewed = np.random.exponential(10, 1000)

print("\nRight-skewed data (Exponential):")
print(f"  Mean:   {skewed.mean():.2f}")
print(f"  Median: {np.median(skewed):.2f}")
print("  → Mean > Median for right-skewed data")

# Data with outliers
with_outliers = np.concatenate([np.random.normal(50, 5, 95), [200, 250, 300, 350, 400]])

print("\nData with outliers:")
print(f"  Mean:   {with_outliers.mean():.2f}")
print(f"  Median: {np.median(with_outliers):.2f}")
print("  → Median is robust to outliers!")
```

## 2. Weighted, Geometric, Harmonic Means

**Weighted Mean**: $\bar{x}_w = \frac{\sum w_i x_i}{\sum w_i}$ - For importance weighting

**Geometric Mean**: $\bar{x}_g = (\prod x_i)^{1/n}$ - For growth rates

**Harmonic Mean**: $\bar{x}_h = \frac{n}{\sum 1/x_i}$ - For rates (F1-score!)

```python
print("SPECIAL MEANS")
print("="*60)

# Weighted Mean - Exam scores
grades = np.array([85, 90, 78, 92])
weights = np.array([0.1, 0.3, 0.2, 0.4])

print("\n📊 Weighted Mean (Exam Scores):")
print(f"  Grades: {grades}")
print(f"  Weights: {weights}")
print(f"  Simple Mean: {grades.mean():.2f}")
print(f"  Weighted Mean: {np.average(grades, weights=weights):.2f}")

# Geometric Mean - Growth rates
growth_rates = np.array([1.10, 0.95, 1.20, 1.05])  # +10%, -5%, +20%, +5%

print("\n📈 Geometric Mean (Investment Returns):")
print(f"  Annual multipliers: {growth_rates}")
print(f"  Arithmetic Mean: {growth_rates.mean():.4f}")
print(f"  Geometric Mean: {stats.gmean(growth_rates):.4f}")
print(f"  $100 after 4 years: ${100 * np.prod(growth_rates):.2f}")

# Harmonic Mean - F1 Score
precision, recall = 0.9, 0.6

print("\n🎯 Harmonic Mean (F1 Score):")
print(f"  Precision: {precision}, Recall: {recall}")
print(f"  Arithmetic Mean: {(precision + recall)/2:.4f}")
print(f"  F1 (Harmonic Mean): {stats.hmean([precision, recall]):.4f}")
print("  → Harmonic mean penalizes imbalance!")
```

## 3. Measures of Spread

**Variance**: $s^2 = \frac{1}{n-1}\sum(x_i - \bar{x})^2$

**IQR**: $Q_3 - Q_1$ - Robust to outliers

**CV**: $\frac{s}{\bar{x}}$ - Normalized (unitless)

```python
print("MEASURES OF SPREAD")
print("="*60)

np.random.seed(42)
data = np.random.normal(100, 15, 200)

print(f"\nData: n={len(data)}, from Normal(100, 15²)")

print(f"\n📏 Range: {data.max() - data.min():.2f}")
print(f"   Min: {data.min():.2f}, Max: {data.max():.2f}")

print(f"\n📊 Variance (sample): {data.var(ddof=1):.2f}")
print(f"   Std Dev: {data.std(ddof=1):.2f} (True σ = 15)")

q1, q3 = np.percentile(data, [25, 75])
print(f"\n📦 IQR: {q3 - q1:.2f}")
print(f"   Q1: {q1:.2f}, Q3: {q3:.2f}")

print(f"\n📉 CV: {data.std()/data.mean():.4f} ({data.std()/data.mean()*100:.2f}%)")
```

## 4. Percentiles and Five-Number Summary

```python
print("PERCENTILES")
print("="*60)

np.random.seed(42)
data = np.random.exponential(10, 1000)

print("\nPercentiles of exponential data:")
for p in [10, 25, 50, 75, 90, 95, 99]:
    print(f"  {p}th percentile: {np.percentile(data, p):.2f}")

five_num = np.percentile(data, [0, 25, 50, 75, 100])
print(f"\nFive-number summary:")
print(f"  Min={five_num[0]:.2f}, Q1={five_num[1]:.2f}, Med={five_num[2]:.2f}, Q3={five_num[3]:.2f}, Max={five_num[4]:.2f}")
```

## 5. Outlier Detection (IQR Method)

**Fences**:
- Lower: $Q_1 - 1.5 \times IQR$
- Upper: $Q_3 + 1.5 \times IQR$

```python
print("OUTLIER DETECTION (IQR METHOD)")
print("="*60)

np.random.seed(42)
clean = np.random.normal(50, 5, 100)
outliers = np.array([10, 15, 85, 90, 95])
data = np.concatenate([clean, outliers])

print(f"\nData: {len(clean)} normal + {len(outliers)} outliers")

q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print(f"\nQ1={q1:.2f}, Q3={q3:.2f}, IQR={iqr:.2f}")
print(f"Lower fence: {lower:.2f}")
print(f"Upper fence: {upper:.2f}")

detected = data[(data < lower) | (data > upper)]
print(f"\n🎯 Detected outliers: {detected}")

# Visualize
fig, ax = plt.subplots(figsize=(10, 3))
ax.boxplot(data, vert=False)
ax.scatter(detected, [1]*len(detected), color='red', s=100, zorder=5, label='Outliers')
ax.set_title('Box Plot with Outliers')
ax.legend()
plt.show()
```

## 6. Skewness and Kurtosis

**Skewness**: Measures asymmetry (0 = symmetric)

**Kurtosis**: Measures tail weight (0 = normal tails)

```python
print("SKEWNESS AND KURTOSIS")
print("="*60)

np.random.seed(42)
n = 10000

normal = np.random.normal(0, 1, n)
right_skewed = np.random.exponential(1, n)
left_skewed = -np.random.exponential(1, n)
heavy_tailed = np.random.standard_t(3, n)

dists = [
    (normal, "Normal"),
    (right_skewed, "Right-Skewed"),
    (left_skewed, "Left-Skewed"),
    (heavy_tailed, "Heavy-Tailed (t)")
]

print(f"\n{'Distribution':<20} {'Skewness':>10} {'Excess Kurt':>12}")
print("-"*45)
for data, name in dists:
    print(f"{name:<20} {stats.skew(data):>10.4f} {stats.kurtosis(data):>12.4f}")

# Visualization
fig, axes = plt.subplots(1, 4, figsize=(16, 3))
for ax, (data, name) in zip(axes, dists):
    ax.hist(data[np.abs(data) < 5], bins=50, density=True, alpha=0.7)
    ax.set_title(f"{name}\nskew={stats.skew(data):.2f}")
plt.tight_layout()
plt.show()
```

## 7. Correlation: Pearson vs Spearman

**Pearson**: Linear relationships

**Spearman**: Monotonic relationships (robust)

```python
print("CORRELATION COMPARISON")
print("="*60)

np.random.seed(42)
n = 100

# Linear
x1 = np.linspace(0, 10, n)
y1 = 2*x1 + np.random.normal(0, 2, n)

# Monotonic non-linear
x2 = np.linspace(0.1, 10, n)
y2 = np.log(x2) + np.random.normal(0, 0.3, n)

# Quadratic (non-monotonic)
x3 = np.linspace(-5, 5, n)
y3 = x3**2 + np.random.normal(0, 2, n)

cases = [(x1, y1, "Linear"), (x2, y2, "Logarithmic"), (x3, y3, "Quadratic")]

print(f"\n{'Relationship':<15} {'Pearson':>10} {'Spearman':>10}")
print("-"*40)
for x, y, name in cases:
    r, _ = stats.pearsonr(x, y)
    rho, _ = stats.spearmanr(x, y)
    print(f"{name:<15} {r:>10.4f} {rho:>10.4f}")

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, (x, y, name) in zip(axes, cases):
    ax.scatter(x, y, alpha=0.6)
    r, _ = stats.pearsonr(x, y)
    rho, _ = stats.spearmanr(x, y)
    ax.set_title(f"{name}\nPearson={r:.3f}, Spearman={rho:.3f}")
plt.tight_layout()
plt.show()

print("\n⚠️ Pearson only captures LINEAR relationships!")
print("💡 For quadratic, both fail → ALWAYS visualize!")
```

## 8. Data Normalization

**Z-Score**: $(x - \mu) / \sigma$ → mean=0, std=1

**Min-Max**: $(x - min) / (max - min)$ → [0, 1]

**Robust**: $(x - median) / IQR$ → Robust to outliers

```python
print("NORMALIZATION METHODS")
print("="*60)

data = np.array([10, 12, 14, 16, 18, 20, 22, 24, 26, 100])  # 100 is outlier
print(f"\nOriginal: {data}")

# Z-Score
z = (data - data.mean()) / data.std()
print(f"\nZ-Score: {np.round(z, 2)}")

# Min-Max
mm = (data - data.min()) / (data.max() - data.min())
print(f"Min-Max: {np.round(mm, 2)}")

# Robust
med = np.median(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
rob = (data - med) / iqr
print(f"Robust:  {np.round(rob, 2)}")

print("\n💡 Min-Max: Outlier compresses all normal data near 0")
print("💡 Robust: Handles outlier while keeping normal data spread")
```

## 9. ML Preprocessing Pipeline Example

```python
print("ML PREPROCESSING PIPELINE")
print("="*60)

np.random.seed(42)
n = 1000

# Simulate features
income = np.concatenate([np.random.exponential(50000, 950), np.random.uniform(500000, 1000000, 50)])
age = np.clip(np.random.normal(40, 12, n), 18, 80)
score = np.random.beta(5, 2, n) * 100

features = [(income, "Income"), (age, "Age"), (score, "Score")]

print("\n📊 ANALYSIS:")
print(f"{'Feature':<10} {'Mean':>12} {'Median':>12} {'Skewness':>10} {'Action':>18}")
print("-"*65)

for feat, name in features:
    skew = stats.skew(feat)
    action = "Log transform" if abs(skew) > 1 else "Standardize"
    print(f"{name:<10} {feat.mean():>12.1f} {np.median(feat):>12.1f} {skew:>10.2f} {action:>18}")

# Apply and visualize
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for i, (feat, name) in enumerate(features):
    axes[0, i].hist(feat, bins=50, alpha=0.7)
    axes[0, i].set_title(f"{name}\nskew={stats.skew(feat):.2f}")

# Transform
income_t = np.log1p(income)
age_t = (age - age.mean()) / age.std()
score_t = (score - score.mean()) / score.std()

transformed = [(income_t, "Log(Income)"), (age_t, "Std(Age)"), (score_t, "Std(Score)")]

for i, (feat, name) in enumerate(transformed):
    axes[1, i].hist(feat, bins=50, alpha=0.7, color='green')
    axes[1, i].set_title(f"{name}\nskew={stats.skew(feat):.2f}")

axes[0, 0].set_ylabel('Before', fontsize=12)
axes[1, 0].set_ylabel('After', fontsize=12)
plt.tight_layout()
plt.show()

print("\n✅ Log transform fixed income skewness!")
print("✅ All features now ready for ML algorithms")
```
