```python
print("EXERCISE 4: Taylor Series for 1/(1+x)")
print("="*60)

print("f(x) = 1/(1+x) = (1+x)^(-1)")
print("\n--- Derivatives ---")
print("f(x)   = (1+x)^(-1)           → f(0) = 1")
print("f'(x)  = -(1+x)^(-2)          → f'(0) = -1")
print("f''(x) = 2(1+x)^(-3)          → f''(0) = 2")
print("f'''(x)= -6(1+x)^(-4)         → f'''(0) = -6")
print("f^(n)(x) = (-1)^n n! (1+x)^(-(n+1))")
print("f^(n)(0) = (-1)^n n!")

print("\n--- Taylor Series ---")
print("f(x) = Σ f^(n)(0)/n! x^n")
print("     = Σ (-1)^n x^n")
print("     = 1 - x + x² - x³ + ...")
print("(This is the geometric series with r = -x)")

def taylor_one_over_one_plus_x(x, n_terms):
    """Taylor series for 1/(1+x): Σ(-1)^n x^n"""
    return sum((-1)**n * x**n for n in range(n_terms))

print("\n--- Verification ---")
for x in [0.1, 0.3, 0.5, 0.9]:
    exact = 1 / (1 + x)
    approx = taylor_one_over_one_plus_x(x, 20)
    print(f"  x = {x}: exact = {exact:.6f}, taylor = {approx:.6f}")

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(-0.9, 0.9, 200)

ax.plot(x, 1/(1+x), 'k-', linewidth=3, label='1/(1+x) exact')
for n in [1, 2, 5, 10]:
    taylor = sum((-1)**k * x**k for k in range(n))
    ax.plot(x, taylor, '--', linewidth=1.5, label=f'{n} terms')

ax.set_xlim(-1, 1)
ax.set_ylim(0, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Taylor Series for 1/(1+x)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
print("EXERCISE 8: Power Series Operations")
print("="*60)

print("f(x) = e^x = 1 + x + x²/2! + x³/3! + ...")
print("g(x) = 1/(1-x) = 1 + x + x² + x³ + ...")

# Coefficients
N = 6
a = [1/factorial(n) for n in range(N)]  # e^x coefficients
b = [1 for _ in range(N)]               # 1/(1-x) coefficients

print(f"\na_n (e^x): {[f'{c:.4f}' for c in a]}")
print(f"b_n (1/(1-x)): {b}")

print("\n--- Part 1: Addition f(x) + g(x) ---")
c_sum = [a[n] + b[n] for n in range(N)]
print(f"c_n = a_n + b_n: {[f'{c:.4f}' for c in c_sum]}")
print("f + g = 2 + 2x + 1.5x² + 1.167x³ + ...")

print("\n--- Part 2: Multiplication f(x)·g(x) (Cauchy product) ---")
print("c_n = Σ_{k=0}^{n} a_k b_{n-k}")
c_prod = []
for n in range(N):
    cn = sum(a[k] * b[n-k] for k in range(n+1))
    c_prod.append(cn)
    
print(f"Product coefficients: {[f'{c:.4f}' for c in c_prod]}")

# Verification
x = 0.3
exact_prod = np.exp(x) / (1 - x)
series_prod = sum(c_prod[n] * x**n for n in range(N))
print(f"\nAt x = {x}:")
print(f"  Exact: e^x/(1-x) = {exact_prod:.6f}")
print(f"  Series (6 terms): {series_prod:.6f}")

print("\n--- Part 3: Composition f(g(x) - 1) ---")
# g(x) - 1 = x/(1-x)
# e^(x/(1-x)) at x = 0.1
x = 0.1
inner = x / (1 - x)  # g(x) - 1 = x/(1-x)
exact_comp = np.exp(inner)
print(f"At x = {x}: x/(1-x) = {inner:.6f}")
print(f"e^(x/(1-x)) = e^{inner:.4f} = {exact_comp:.6f}")

# Series computation
series_comp = sum(inner**n / factorial(n) for n in range(10))
print(f"Series approximation: {series_comp:.6f}")
```
