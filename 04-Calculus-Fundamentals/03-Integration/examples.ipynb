## 1. Definite Integral as Area

The definite integral represents the signed area under a curve:
$$\int_a^b f(x)\, dx = \text{Area between } f(x) \text{ and the x-axis}$$

```python
print("DEFINITE INTEGRAL AS AREA")
print("="*60)

print("\nEvaluate: ∫[0 to 2] x² dx")
print("\nUsing Fundamental Theorem of Calculus:")
print("Antiderivative of x² is x³/3")
print("∫[0 to 2] x² dx = [x³/3] from 0 to 2")
print("              = 2³/3 - 0³/3")
print("              = 8/3 - 0")
print("              = 8/3 ≈ 2.667")

# Numerical verification
result, error = integrate.quad(lambda x: x**2, 0, 2)
print(f"\nNumerical verification: {result:.6f}")
```

```python
# Visualize the area under the curve
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(0, 2.5, 100)
y = x**2

ax.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^2$')

# Fill the area under the curve from 0 to 2
x_fill = np.linspace(0, 2, 100)
y_fill = x_fill**2
ax.fill_between(x_fill, 0, y_fill, alpha=0.3, color='blue', label='Area = 8/3 ≈ 2.667')

ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
ax.axvline(x=2, color='red', linestyle='--', alpha=0.5, label='x = 2')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title(r'Definite Integral: $\int_0^2 x^2\, dx$', fontsize=14)
ax.legend()
ax.set_xlim(-0.2, 2.5)
ax.set_ylim(-0.5, 5)

plt.tight_layout()
plt.show()
```

## 2. Riemann Sum Approximation

$$\int_a^b f(x)\, dx = \lim_{n \to \infty} \sum_{i=1}^n f(x_i^*) \Delta x$$

where $\Delta x = \frac{b-a}{n}$

```python
print("RIEMANN SUM APPROXIMATION")
print("="*60)

def f(x):
    return x**2

a, b = 0, 2
exact = 8/3

print(f"\nApproximate ∫[{a} to {b}] x² dx using Riemann sums")
print(f"Exact value: {exact:.6f}")
print("\n{:>8} | {:>12} | {:>12}".format('n', 'Riemann Sum', 'Error'))
print("-" * 40)

for n in [4, 10, 100, 1000, 10000]:
    dx = (b - a) / n
    # Right endpoint Riemann sum
    x_points = np.linspace(a + dx, b, n)
    riemann_sum = np.sum(f(x_points)) * dx
    error = abs(riemann_sum - exact)
    print(f"{n:8d} | {riemann_sum:12.6f} | {error:12.6f}")
```

```python
# Visualize Riemann sums
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, n in zip(axes, [4, 8, 20]):
    x = np.linspace(a, b, 100)
    y = x**2
    ax.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^2$')
    
    # Draw rectangles
    dx = (b - a) / n
    for i in range(n):
        x_left = a + i * dx
        x_right = x_left + dx
        height = f(x_right)  # Right endpoint
        rect = plt.Rectangle((x_left, 0), dx, height, 
                             fill=True, facecolor='lightblue', 
                             edgecolor='blue', alpha=0.6)
        ax.add_patch(rect)
    
    # Calculate sum
    x_points = np.linspace(a + dx, b, n)
    riemann_sum = np.sum(f(x_points)) * dx
    
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('f(x)', fontsize=12)
    ax.set_title(f'n = {n}, Sum = {riemann_sum:.4f}', fontsize=12)
    ax.set_xlim(-0.1, 2.2)
    ax.set_ylim(-0.2, 4.5)

plt.suptitle('Riemann Sum Convergence (Exact = 2.6667)', fontsize=14)
plt.tight_layout()
plt.show()
```

## 3. Basic Antiderivatives

| Function | Antiderivative |
|----------|----------------|
| $x^n$ | $\frac{x^{n+1}}{n+1} + C$ |
| $e^x$ | $e^x + C$ |
| $1/x$ | $\ln|x| + C$ |

```python
print("BASIC ANTIDERIVATIVES")
print("="*60)

print("\nPower Rule: ∫ x^n dx = x^(n+1)/(n+1) + C  (n ≠ -1)")

examples = [
    ("∫ x³ dx", "x⁴/4 + C"),
    ("∫ x^(-2) dx", "-x^(-1) + C = -1/x + C"),
    ("∫ √x dx = ∫ x^(1/2) dx", "x^(3/2)/(3/2) + C = (2/3)x^(3/2) + C"),
    ("∫ 1 dx", "x + C"),
]

for integral, result in examples:
    print(f"\n  {integral}")
    print(f"    = {result}")

print("\n" + "-"*60)
print("Special cases:")
print("  ∫ e^x dx = e^x + C")
print("  ∫ 1/x dx = ln|x| + C")
print("  ∫ sin(x) dx = -cos(x) + C")
print("  ∫ cos(x) dx = sin(x) + C")
```

## 4. U-Substitution

For $\int f(g(x))g'(x)\, dx$, let $u = g(x)$, then $du = g'(x)\,dx$

```python
print("U-SUBSTITUTION")
print("="*60)

print("\nEvaluate: ∫ 2x·e^(x²) dx")
print("\n--- Method ---")
print("Let u = x²")
print("Then du = 2x dx")
print("\nSubstitute:")
print("∫ 2x·e^(x²) dx = ∫ e^u du = e^u + C = e^(x²) + C")

print("\n--- Verification by differentiation ---")
print("d/dx[e^(x²)] = e^(x²)·2x = 2x·e^(x²) ✓")

# Numerical check
print("\n--- Definite integral: ∫[0 to 1] 2x·e^(x²) dx ---")
print("= [e^(x²)] from 0 to 1")
print(f"= e¹ - e⁰ = e - 1 = {np.e - 1:.6f}")

result, _ = integrate.quad(lambda x: 2*x*np.exp(x**2), 0, 1)
print(f"Numerical: {result:.6f}")
```

## 5. Integration by Parts

$$\int u\, dv = uv - \int v\, du$$

Choose $u$ and $dv$ using **LIATE** priority: Logarithmic, Inverse trig, Algebraic, Trigonometric, Exponential

```python
print("INTEGRATION BY PARTS")
print("="*60)

print("\nFormula: ∫ u dv = uv - ∫ v du")

print("\n--- Example: ∫ x·e^x dx ---")
print("Using LIATE rule: u = x (Algebraic), dv = e^x dx")
print("Then: du = dx, v = e^x")
print("\n∫ x·e^x dx = x·e^x - ∫ e^x dx")
print("          = x·e^x - e^x + C")
print("          = e^x(x - 1) + C")

print("\n--- Verification by differentiation ---")
print("d/dx[e^x(x-1)] = e^x(x-1) + e^x·1 = e^x·x - e^x + e^x = x·e^x ✓")

print("\n--- Definite integral: ∫[0 to 1] x·e^x dx ---")
print("= [e^x(x-1)] from 0 to 1")
print("= e¹(1-1) - e⁰(0-1) = 0 - (-1) = 1")

result, _ = integrate.quad(lambda x: x*np.exp(x), 0, 1)
print(f"Numerical: {result:.6f}")
```

```python
# More integration by parts examples
print("\n--- Example 2: ∫ ln(x) dx ---")
print("u = ln(x), dv = dx")
print("du = 1/x dx, v = x")
print("\n∫ ln(x) dx = x·ln(x) - ∫ x·(1/x) dx")
print("          = x·ln(x) - ∫ 1 dx")
print("          = x·ln(x) - x + C")
print("          = x(ln(x) - 1) + C")

# Verify
result, _ = integrate.quad(lambda x: np.log(x), 1, np.e)
analytical = np.e*(1 - 1) - 1*(0 - 1)  # = 0 - (-1) = 1
print(f"\n∫[1 to e] ln(x) dx = {result:.6f} (analytical: 1)")
```

## 6. Gaussian Integral

One of the most important integrals in probability and ML:
$$\int_{-\infty}^{\infty} e^{-x^2}\, dx = \sqrt{\pi}$$

```python
print("GAUSSIAN INTEGRAL")
print("="*60)

print("\nThe Gaussian integral:")
print("∫[-∞ to ∞] e^(-x²) dx = √π")

print(f"\n√π = {np.sqrt(np.pi):.10f}")

# Numerical approximation
result, error = integrate.quad(lambda x: np.exp(-x**2), -np.inf, np.inf)
print(f"Numerical: {result:.10f}")

print("\n" + "-"*60)
print("Standard Normal Distribution:")
print("∫[-∞ to ∞] (1/√(2π)) e^(-x²/2) dx = 1")

result_normal, _ = integrate.quad(
    lambda x: np.exp(-x**2/2) / np.sqrt(2*np.pi), 
    -np.inf, np.inf
)
print(f"Numerical: {result_normal:.10f}")
```

```python
# Visualize Gaussian
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-4, 4, 1000)

# Gaussian e^(-x^2)
y1 = np.exp(-x**2)
axes[0].plot(x, y1, 'b-', linewidth=2, label=r'$e^{-x^2}$')
axes[0].fill_between(x, 0, y1, alpha=0.3, label=f'Area = √π ≈ {np.sqrt(np.pi):.4f}')
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('Gaussian Integral', fontsize=14)
axes[0].legend()

# Standard Normal
y2 = np.exp(-x**2/2) / np.sqrt(2*np.pi)
axes[1].plot(x, y2, 'r-', linewidth=2, label=r'$\frac{1}{\sqrt{2\pi}}e^{-x^2/2}$')
axes[1].fill_between(x, 0, y2, alpha=0.3, color='red', label='Area = 1')
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('p(x)', fontsize=12)
axes[1].set_title('Standard Normal PDF', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 7. Expected Value (Probability)

For a continuous random variable with PDF $p(x)$:
$$\mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot p(x)\, dx$$

```python
print("EXPECTED VALUE (PROBABILITY)")
print("="*60)

print("\nFor continuous random variable X with PDF p(x):")
print("E[X] = ∫ x·p(x) dx")

print("\n--- Example: Uniform(0, 1) ---")
print("p(x) = 1 for x ∈ [0, 1], 0 otherwise")
print("\nE[X] = ∫[0 to 1] x·1 dx")
print("     = [x²/2] from 0 to 1")
print("     = 1/2 - 0 = 0.5")

print("\n--- Example: Exponential(λ=1) ---")
print("p(x) = e^(-x) for x ≥ 0")
print("\nE[X] = ∫[0 to ∞] x·e^(-x) dx")
print("(Using integration by parts)")
print("     = 1")

result, _ = integrate.quad(lambda x: x * np.exp(-x), 0, np.inf)
print(f"Numerical: {result:.6f}")

print("\n--- E[X²] for Uniform(0, 1) ---")
print("E[X²] = ∫[0 to 1] x²·1 dx")
print("      = [x³/3] from 0 to 1")
print("      = 1/3")

result_x2, _ = integrate.quad(lambda x: x**2, 0, 1)
print(f"Numerical: {result_x2:.6f}")
```

## 8. Variance Calculation

$$\text{Var}(X) = \mathbb{E}[(X - \mu)^2] = \mathbb{E}[X^2] - (\mathbb{E}[X])^2$$

```python
print("VARIANCE CALCULATION")
print("="*60)

print("\nVar(X) = E[(X - μ)²] = E[X²] - (E[X])²")

print("\n--- Uniform(0, 1) ---")
print("E[X] = 1/2")
print("E[X²] = 1/3")
print("Var(X) = 1/3 - (1/2)² = 1/3 - 1/4")
print("       = 4/12 - 3/12 = 1/12")
print(f"       = {1/12:.6f}")

print("\n--- Exponential(λ=1) ---")
print("E[X] = 1")
print("E[X²] = ∫[0 to ∞] x²·e^(-x) dx = 2 (using gamma function)")

E_X2, _ = integrate.quad(lambda x: x**2 * np.exp(-x), 0, np.inf)
print(f"E[X²] numerical: {E_X2:.6f}")
print(f"Var(X) = E[X²] - (E[X])² = 2 - 1 = 1")
```

## 9. Monte Carlo Integration

$$\int_a^b f(x)\, dx \approx (b-a) \cdot \frac{1}{n} \sum_{i=1}^n f(x_i)$$

where $x_i$ are random samples from $[a, b]$.

```python
print("MONTE CARLO INTEGRATION")
print("="*60)

def f(x):
    return np.exp(-x**2)

a, b = 0, 1

print(f"\nEstimate: ∫[{a} to {b}] e^(-x²) dx")

# Exact value
exact, _ = integrate.quad(f, a, b)
print(f"Exact value: {exact:.10f}")

print("\nMonte Carlo estimates:")
np.random.seed(42)

print("{:>10} | {:>14} | {:>12}".format('n', 'Estimate', 'Error'))
print("-" * 45)

for n in [100, 1000, 10000, 100000]:
    x_samples = np.random.uniform(a, b, n)
    mc_estimate = (b - a) * np.mean(f(x_samples))
    error = abs(mc_estimate - exact)
    print(f"{n:10d} | {mc_estimate:14.8f} | {error:12.8f}")

print("\nMonte Carlo: O(1/√n) convergence regardless of dimension!")
print("This is crucial for high-dimensional integrals in ML.")
```

```python
# Visualize Monte Carlo convergence
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

np.random.seed(42)
n_samples = 10000
samples = np.random.uniform(0, 1, n_samples)

# Running estimate
running_estimates = np.cumsum(f(samples)) / np.arange(1, n_samples + 1)

axes[0].plot(range(1, n_samples + 1), running_estimates, 'b-', linewidth=1, alpha=0.7)
axes[0].axhline(y=exact, color='red', linestyle='--', linewidth=2, label=f'Exact = {exact:.6f}')
axes[0].set_xlabel('Number of samples', fontsize=12)
axes[0].set_ylabel('Estimate', fontsize=12)
axes[0].set_title('Monte Carlo Convergence', fontsize=14)
axes[0].set_xscale('log')
axes[0].legend()

# Error convergence
errors = np.abs(running_estimates - exact)
n_range = np.arange(1, n_samples + 1)

axes[1].loglog(n_range, errors, 'b-', linewidth=0.5, alpha=0.5, label='Error')
axes[1].loglog(n_range, 1/np.sqrt(n_range), 'r--', linewidth=2, label=r'$O(1/\sqrt{n})$')
axes[1].set_xlabel('Number of samples', fontsize=12)
axes[1].set_ylabel('Error', fontsize=12)
axes[1].set_title('Monte Carlo Error Scaling', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 10. Numerical Integration Methods

| Method | Formula | Error Order |
|--------|---------|-------------|
| Rectangle | $\sum f(x_i) \Delta x$ | $O(\Delta x)$ |
| Trapezoidal | $\frac{\Delta x}{2}[f(x_0) + 2f(x_1) + ... + f(x_n)]$ | $O(\Delta x^2)$ |
| Simpson | Complex weights | $O(\Delta x^4)$ |

```python
print("NUMERICAL INTEGRATION METHODS")
print("="*60)

def f(x):
    return np.sin(x)

a, b = 0, np.pi
exact = 2.0  # ∫[0 to π] sin(x) dx = 2

print(f"\n∫[0 to π] sin(x) dx = 2 (exact)")

for n in [4, 8, 16, 32]:
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    
    # Rectangle rule (left endpoint)
    rect = np.sum(f(x[:-1])) * dx
    
    # Trapezoidal rule
    trap = np.trapz(f(x), x)
    
    # Simpson's rule (requires even number of intervals)
    if n % 2 == 0:
        simp = integrate.simpson(f(x), x=x)
    else:
        simp = float('nan')
    
    print(f"\nn = {n}:")
    print(f"  Rectangle:   {rect:.8f}, Error: {abs(rect - exact):.2e}")
    print(f"  Trapezoidal: {trap:.8f}, Error: {abs(trap - exact):.2e}")
    if not np.isnan(simp):
        print(f"  Simpson:     {simp:.8f}, Error: {abs(simp - exact):.2e}")
```

```python
# Visualize error convergence for different methods
fig, ax = plt.subplots(figsize=(10, 6))

n_values = [4, 8, 16, 32, 64, 128]
rect_errors = []
trap_errors = []
simp_errors = []

for n in n_values:
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    
    rect = np.sum(f(x[:-1])) * dx
    rect_errors.append(abs(rect - exact))
    
    trap = np.trapz(f(x), x)
    trap_errors.append(abs(trap - exact))
    
    simp = integrate.simpson(f(x), x=x)
    simp_errors.append(abs(simp - exact))

ax.loglog(n_values, rect_errors, 'ro-', linewidth=2, markersize=8, label='Rectangle O(Δx)')
ax.loglog(n_values, trap_errors, 'bs-', linewidth=2, markersize=8, label='Trapezoidal O(Δx²)')
ax.loglog(n_values, simp_errors, 'g^-', linewidth=2, markersize=8, label='Simpson O(Δx⁴)')

ax.set_xlabel('Number of intervals (n)', fontsize=12)
ax.set_ylabel('Error', fontsize=12)
ax.set_title('Numerical Integration Error Comparison', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 11. Improper Integrals

Integrals with infinite limits or discontinuous integrands.

```python
print("IMPROPER INTEGRALS")
print("="*60)

print("\nType 1: Infinite limits")
print("\n∫[1 to ∞] 1/x² dx")
print("= lim(t→∞) ∫[1 to t] x^(-2) dx")
print("= lim(t→∞) [-1/x] from 1 to t")
print("= lim(t→∞) (-1/t - (-1))")
print("= 0 + 1 = 1")

result, _ = integrate.quad(lambda x: 1/x**2, 1, np.inf)
print(f"Numerical: {result:.6f}")

print("\n" + "-"*60)
print("Divergent example:")
print("∫[1 to ∞] 1/x dx = lim(t→∞) ln(t) = ∞ (diverges)")

print("\n" + "-"*60)
print("p-test: ∫[1 to ∞] 1/x^p dx converges iff p > 1")

for p in [0.5, 1.0, 1.5, 2.0]:
    try:
        result, _ = integrate.quad(lambda x, p=p: 1/x**p, 1, 1000)
        if p > 1:
            expected = 1 / (p - 1)
            print(f"  p = {p}: ≈ {result:.4f} (converges to {expected:.4f})")
        else:
            print(f"  p = {p}: ≈ {result:.4f} (diverges)")
    except Exception as e:
        print(f"  p = {p}: computation error")
```

```python
# Visualize convergent vs divergent integrals
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(0.01, 5, 1000)

# Convergent: 1/x^2
y1 = 1/x**2
x_fill = np.linspace(1, 5, 100)
y_fill = 1/x_fill**2

axes[0].plot(x, y1, 'b-', linewidth=2, label=r'$1/x^2$')
axes[0].fill_between(x_fill, 0, y_fill, alpha=0.3, label='Converges to 1')
axes[0].axvline(x=1, color='red', linestyle='--', alpha=0.5)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title(r'Convergent: $\int_1^\infty 1/x^2\, dx = 1$', fontsize=14)
axes[0].legend()
axes[0].set_ylim(0, 2)

# Divergent: 1/x
y2 = 1/x
x_fill = np.linspace(1, 5, 100)
y_fill = 1/x_fill

axes[1].plot(x, y2, 'r-', linewidth=2, label=r'$1/x$')
axes[1].fill_between(x_fill, 0, y_fill, alpha=0.3, color='red', label='Diverges (area → ∞)')
axes[1].axvline(x=1, color='blue', linestyle='--', alpha=0.5)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('f(x)', fontsize=12)
axes[1].set_title(r'Divergent: $\int_1^\infty 1/x\, dx = \infty$', fontsize=14)
axes[1].legend()
axes[1].set_ylim(0, 2)

plt.tight_layout()
plt.show()
```

## Summary

| Concept | Formula | ML Application |
|---------|---------|----------------|
| Definite Integral | $\int_a^b f(x)\,dx = F(b)-F(a)$ | Area computations |
| Expected Value | $E[X] = \int x \cdot p(x)\,dx$ | Model predictions |
| Variance | $\text{Var}(X) = E[X^2] - E[X]^2$ | Uncertainty quantification |
| Gaussian | $\int e^{-x^2}dx = \sqrt{\pi}$ | Normal distributions |
| Monte Carlo | Average of random samples | High-dimensional integration |
