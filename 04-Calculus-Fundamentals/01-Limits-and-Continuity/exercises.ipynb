## Exercise 1: Basic Limits

Evaluate the following limits:

a) $\displaystyle \lim_{x \to 3} \frac{x^2 - 9}{x - 3}$

b) $\displaystyle \lim_{x \to 0} \frac{\sin(5x)}{x}$

c) $\displaystyle \lim_{x \to \infty} \frac{3x^2 + 2x}{x^2 + 1}$

```python
# SOLUTION
print("Exercise 1 Solution: Basic Limits")
print("="*60)

# Part a
print("\na) lim(x→3) (x² - 9)/(x - 3)")
print("-" * 40)
print("Factor: x² - 9 = (x-3)(x+3)")
print("= lim(x→3) (x-3)(x+3)/(x-3)")
print("= lim(x→3) (x+3)")
print("= 3 + 3 = 6")

def f_a(x):
    return (x**2 - 9) / (x - 3)

print(f"\nNumerical check: f(3.0001) = {f_a(3.0001):.6f}")

# Part b
print("\nb) lim(x→0) sin(5x)/x")
print("-" * 40)
print("Rewrite: = lim(x→0) 5 · sin(5x)/(5x)")
print("Let u = 5x, as x → 0, u → 0")
print("= 5 · lim(u→0) sin(u)/u")
print("= 5 · 1 = 5")

def f_b(x):
    return np.sin(5*x) / x

print(f"\nNumerical check: f(0.0001) = {f_b(0.0001):.6f}")

# Part c
print("\nc) lim(x→∞) (3x² + 2x)/(x² + 1)")
print("-" * 40)
print("Divide numerator and denominator by x²:")
print("= lim(x→∞) (3 + 2/x)/(1 + 1/x²)")
print("As x → ∞: 2/x → 0 and 1/x² → 0")
print("= (3 + 0)/(1 + 0) = 3")

def f_c(x):
    return (3*x**2 + 2*x) / (x**2 + 1)

print(f"\nNumerical check: f(10000) = {f_c(10000):.8f}")
```

## Exercise 2: One-Sided Limits

For $f(x) = \frac{x - 1}{|x - 1|}$:

a) Find $\displaystyle \lim_{x \to 1^-} f(x)$

b) Find $\displaystyle \lim_{x \to 1^+} f(x)$

c) Does $\displaystyle \lim_{x \to 1} f(x)$ exist?

```python
# SOLUTION
print("Exercise 2 Solution: One-Sided Limits")
print("="*60)

def f(x):
    return (x - 1) / np.abs(x - 1)

print("\nf(x) = (x - 1)/|x - 1|")

# Part a: Left limit
print("\na) lim(x→1⁻) f(x):")
print("-" * 40)
print("For x < 1: (x - 1) is negative")
print("So |x - 1| = -(x - 1)")
print("f(x) = (x-1)/(-(x-1)) = -1")

print("\nNumerical verification:")
for x in [0.9, 0.99, 0.999, 0.9999]:
    print(f"  f({x}) = {f(x)}")
print("\nLeft-hand limit = -1")

# Part b: Right limit
print("\nb) lim(x→1⁺) f(x):")
print("-" * 40)
print("For x > 1: (x - 1) is positive")
print("So |x - 1| = (x - 1)")
print("f(x) = (x-1)/(x-1) = 1")

print("\nNumerical verification:")
for x in [1.1, 1.01, 1.001, 1.0001]:
    print(f"  f({x}) = {f(x)}")
print("\nRight-hand limit = +1")

# Part c: Two-sided limit
print("\nc) Does lim(x→1) f(x) exist?")
print("-" * 40)
print("Left limit = -1")
print("Right limit = +1")
print("\nSince left limit ≠ right limit,")
print("lim(x→1) f(x) does NOT exist ✗")
```

## Exercise 3: L'Hôpital's Rule

Use L'Hôpital's Rule to evaluate:

a) $\displaystyle \lim_{x \to 0} \frac{1 - \cos(x)}{x^2}$

b) $\displaystyle \lim_{x \to \infty} x \cdot e^{-x}$

c) $\displaystyle \lim_{x \to 0^+} x \ln(x)$

```python
# SOLUTION
print("Exercise 3 Solution: L'Hôpital's Rule")
print("="*60)

# Part a
print("\na) lim(x→0) (1 - cos(x))/x²")
print("-" * 40)
print("At x = 0: numerator = 1 - 1 = 0, denominator = 0")
print("Form: 0/0 → Apply L'Hôpital")
print("\nf(x) = 1 - cos(x)  →  f'(x) = sin(x)")
print("g(x) = x²          →  g'(x) = 2x")
print("\n= lim(x→0) sin(x)/(2x) → still 0/0")
print("\nApply L'Hôpital again:")
print("f'(x) = sin(x)  →  f''(x) = cos(x)")
print("g'(x) = 2x      →  g''(x) = 2")
print("\n= lim(x→0) cos(x)/2 = 1/2")

def f_a(x):
    return (1 - np.cos(x)) / x**2

print(f"\nNumerical check: f(0.001) = {f_a(0.001):.8f}")

# Part b
print("\nb) lim(x→∞) x·e^(-x)")
print("-" * 40)
print("Rewrite as: lim(x→∞) x/eˣ")
print("Form: ∞/∞ → Apply L'Hôpital")
print("\nf(x) = x   →  f'(x) = 1")
print("g(x) = eˣ  →  g'(x) = eˣ")
print("\n= lim(x→∞) 1/eˣ = 0")

def f_b(x):
    return x * np.exp(-x)

print(f"\nNumerical check: f(50) = {f_b(50):.2e}")

# Part c
print("\nc) lim(x→0⁺) x·ln(x)")
print("-" * 40)
print("Form: 0·(-∞) → need to rewrite")
print("Rewrite as: lim(x→0⁺) ln(x)/(1/x)")
print("Form: -∞/∞ → Apply L'Hôpital")
print("\nf(x) = ln(x)  →  f'(x) = 1/x")
print("g(x) = 1/x    →  g'(x) = -1/x²")
print("\n= lim(x→0⁺) (1/x)/(-1/x²)")
print("= lim(x→0⁺) -x = 0")

def f_c(x):
    return x * np.log(x)

print(f"\nNumerical check: f(0.0001) = {f_c(0.0001):.6f}")
```

## Exercise 4: Continuity Analysis

Determine where each function is discontinuous and classify the discontinuity:

a) $f(x) = \frac{x^2 - 4}{x + 2}$

b) $f(x) = \frac{1}{x^2 - 1}$

c) $f(x) = \lfloor x \rfloor$ (floor function)

```python
# SOLUTION
print("Exercise 4 Solution: Continuity Analysis")
print("="*60)

# Part a
print("\na) f(x) = (x² - 4)/(x + 2)")
print("-" * 40)
print("Undefined when x + 2 = 0, i.e., x = -2")
print("\nFactor: x² - 4 = (x-2)(x+2)")
print("f(x) = (x-2)(x+2)/(x+2) = x - 2 (for x ≠ -2)")
print("\nlim(x→-2) f(x) = lim(x→-2) (x - 2) = -4")
print("\nThe limit exists! This is a REMOVABLE discontinuity.")
print("(A 'hole' in the graph at (-2, -4))")

# Part b
print("\nb) f(x) = 1/(x² - 1)")
print("-" * 40)
print("Undefined when x² - 1 = 0")
print("x² = 1 → x = ±1")
print("\nAt x = 1:")
print("  lim(x→1⁻) 1/(x²-1) = -∞")
print("  lim(x→1⁺) 1/(x²-1) = +∞")
print("\nAt x = -1:")
print("  lim(x→-1⁻) 1/(x²-1) = +∞")
print("  lim(x→-1⁺) 1/(x²-1) = -∞")
print("\nBoth are INFINITE discontinuities (vertical asymptotes).")

# Part c
print("\nc) f(x) = floor(x)")
print("-" * 40)
print("At every integer n:")
print("  lim(x→n⁻) floor(x) = n - 1")
print("  lim(x→n⁺) floor(x) = n")
print("  f(n) = n")
print("\nLeft limit ≠ Right limit at every integer.")
print("These are JUMP discontinuities.")

# Visualize floor function
fig, ax = plt.subplots(figsize=(10, 4))
for n in range(-3, 4):
    ax.hlines(y=n, xmin=n, xmax=n+1, color='blue', linewidth=2)
    ax.plot(n, n, 'bo', markersize=8)  # Closed circle
    ax.plot(n+1, n, 'bo', markersize=8, fillstyle='none')  # Open circle
ax.set_xlabel('x')
ax.set_ylabel('floor(x)')
ax.set_title('Floor Function: Jump Discontinuities at Every Integer')
ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 5: Squeeze Theorem

Use the Squeeze Theorem to find:

$$\lim_{x \to \infty} \frac{\sin(x)}{x}$$

```python
# SOLUTION
print("Exercise 5 Solution: Squeeze Theorem")
print("="*60)

print("\nFind: lim(x→∞) sin(x)/x")

print("\n--- Step 1: Establish bounds ---")
print("We know: -1 ≤ sin(x) ≤ 1 for all x")

print("\n--- Step 2: Divide by x (positive for large x) ---")
print("-1/x ≤ sin(x)/x ≤ 1/x")

print("\n--- Step 3: Take limits ---")
print("lim(x→∞) (-1/x) = 0")
print("lim(x→∞) (1/x) = 0")

print("\n--- Step 4: Apply Squeeze Theorem ---")
print("Since the lower and upper bounds both approach 0,")
print("by the Squeeze Theorem:")
print("\nlim(x→∞) sin(x)/x = 0 ✓")

# Numerical verification
print("\n--- Numerical Verification ---")
def f(x):
    return np.sin(x) / x

for x in [10, 100, 1000, 10000]:
    val = f(x)
    bound = 1/x
    print(f"  x = {x:5d}: sin(x)/x = {val:10.6f}  bounds = ±{bound:.6f}")

# Visualization
fig, ax = plt.subplots(figsize=(12, 5))
x = np.linspace(1, 50, 1000)
ax.plot(x, np.sin(x)/x, 'b-', linewidth=2, label=r'$\sin(x)/x$')
ax.plot(x, 1/x, 'r--', linewidth=1.5, label=r'$1/x$')
ax.plot(x, -1/x, 'r--', linewidth=1.5, label=r'$-1/x$')
ax.fill_between(x, -1/x, 1/x, alpha=0.2, color='green')
ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(r'Squeeze Theorem: $\lim_{x \to \infty} \frac{\sin x}{x} = 0$')
ax.legend()
plt.tight_layout()
plt.show()
```

## Exercise 6: Piecewise Function Continuity

For what value of $k$ is the following function continuous everywhere?

$$f(x) = \begin{cases} x^2 + k & \text{if } x < 2 \\ 3x & \text{if } x \geq 2 \end{cases}$$

```python
# SOLUTION
print("Exercise 6 Solution: Piecewise Continuity")
print("="*60)

print("\nf(x) = { x² + k,  if x < 2")
print("       { 3x,      if x ≥ 2")

print("\n--- Continuity requirements at x = 2 ---")
print("For f to be continuous at x = 2:")
print("lim(x→2⁻) f(x) = lim(x→2⁺) f(x) = f(2)")

print("\n--- Compute each piece ---")
print("Left limit:  lim(x→2⁻) (x² + k) = 4 + k")
print("Right limit: lim(x→2⁺) 3x = 6")
print("Function value: f(2) = 3(2) = 6")

print("\n--- Solve for k ---")
print("For continuity: 4 + k = 6")
print("Therefore: k = 2")

# Verification
def f(x, k=2):
    if x < 2:
        return x**2 + k
    else:
        return 3*x

print("\n--- Verification with k = 2 ---")
print(f"  f(1.999) = {f(1.999):.6f}")
print(f"  f(2.000) = {f(2.0):.6f}")
print(f"  f(2.001) = {f(2.001):.6f}")
print("\nAll values match at x = 2 → continuous ✓")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Discontinuous (k = 0)
x_left = np.linspace(0, 1.99, 100)
x_right = np.linspace(2, 4, 100)
axes[0].plot(x_left, x_left**2 + 0, 'b-', linewidth=2, label='x² + 0')
axes[0].plot(x_right, 3*x_right, 'b-', linewidth=2, label='3x')
axes[0].plot(2, 4, 'bo', markersize=8, fillstyle='none')
axes[0].plot(2, 6, 'bo', markersize=8)
axes[0].set_title('Discontinuous (k = 0)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')

# Continuous (k = 2)
axes[1].plot(x_left, x_left**2 + 2, 'b-', linewidth=2, label='x² + 2')
axes[1].plot(x_right, 3*x_right, 'b-', linewidth=2, label='3x')
axes[1].plot(2, 6, 'bo', markersize=8)
axes[1].set_title('Continuous (k = 2)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')

for ax in axes:
    ax.axvline(x=2, color='gray', linestyle='--', alpha=0.5)
    ax.legend()

plt.tight_layout()
plt.show()
```

## Exercise 7: ReLU Continuity (ML)

Show that the ReLU function $\text{ReLU}(x) = \max(0, x)$ is continuous everywhere.

```python
# SOLUTION
print("Exercise 7 Solution: ReLU Continuity")
print("="*60)

print("\nReLU(x) = max(0, x) = { 0,  if x < 0")
print("                      { x,  if x ≥ 0")

print("\n--- Check continuity at x = 0 (critical point) ---")
print("\n1. f(0) defined?")
print("   f(0) = max(0, 0) = 0 ✓")

print("\n2. Left limit exists?")
print("   lim(x→0⁻) 0 = 0 ✓")

print("\n3. Right limit exists?")
print("   lim(x→0⁺) x = 0 ✓")

print("\n4. Limits equal f(0)?")
print("   Left = Right = f(0) = 0 ✓")

print("\n--- Check everywhere else ---")
print("For x < 0: f(x) = 0 (constant, continuous)")
print("For x > 0: f(x) = x (identity, continuous)")

print("\n" + "="*60)
print("Conclusion: ReLU is continuous on all of ℝ ✓")
print("\nNote: ReLU is NOT differentiable at x = 0")
print("(left derivative = 0, right derivative = 1)")

# Numerical verification
def relu(x):
    return np.maximum(0, x)

print("\n--- Numerical Verification near x = 0 ---")
for x in [-0.001, -0.0001, 0, 0.0001, 0.001]:
    print(f"  ReLU({x:8.4f}) = {relu(x):.6f}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = np.linspace(-3, 3, 1000)

# ReLU
axes[0].plot(x, relu(x), 'b-', linewidth=2)
axes[0].plot(0, 0, 'bo', markersize=8)
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('ReLU(x)')
axes[0].set_title('ReLU: Continuous Everywhere')

# Derivative (not defined at 0, but typically uses 0)
relu_deriv = np.where(x > 0, 1, np.where(x < 0, 0, 0.5))  # 0.5 shown for visual
axes[1].plot(x[x < 0], np.zeros_like(x[x < 0]), 'g-', linewidth=2)
axes[1].plot(x[x > 0], np.ones_like(x[x > 0]), 'g-', linewidth=2)
axes[1].plot(0, 0, 'go', markersize=8, fillstyle='none')
axes[1].plot(0, 1, 'go', markersize=8, fillstyle='none')
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[1].set_xlabel('x')
axes[1].set_ylabel("ReLU'(x)")
axes[1].set_title("ReLU Derivative: NOT Differentiable at x=0")
axes[1].set_ylim(-0.5, 1.5)

plt.tight_layout()
plt.show()
```

## Exercise 8: Exponential Limit

Prove that $\displaystyle \lim_{n \to \infty} \left(1 + \frac{r}{n}\right)^n = e^r$

Verify numerically for $r = 2$.

```python
# SOLUTION
print("Exercise 8 Solution: Exponential Limit")
print("="*60)

print("\nProve: lim(n→∞) (1 + r/n)^n = e^r")

print("\n--- Proof Sketch ---")
print("Let y = (1 + r/n)^n")
print("\nTake natural log:")
print("ln(y) = n · ln(1 + r/n)")

print("\nAs n → ∞, r/n → 0")
print("Use Taylor expansion: ln(1 + x) ≈ x - x²/2 + x³/3 - ... for small x")
print("\nln(1 + r/n) ≈ r/n - (r/n)²/2 + O(1/n³)")
print("           ≈ r/n - r²/(2n²) + O(1/n³)")

print("\nMultiply by n:")
print("n · ln(1 + r/n) ≈ r - r²/(2n) + O(1/n²)")

print("\nAs n → ∞:")
print("ln(y) → r")
print("y → e^r ✓")

# Numerical verification for r = 2
r = 2
print(f"\n--- Numerical Verification for r = {r} ---")
print(f"Target: e^{r} = {np.exp(r):.12f}")
print()

for n in [10, 100, 1000, 10000, 100000, 1000000]:
    val = (1 + r/n)**n
    error = abs(val - np.exp(r))
    print(f"  n = {n:7d}: (1 + {r}/n)^n = {val:.12f}  error = {error:.2e}")

# Visualization
fig, ax = plt.subplots(figsize=(10, 5))

n_vals = np.logspace(0.5, 5, 100).astype(int)
y_vals = [(1 + r/n)**n for n in n_vals]

ax.semilogx(n_vals, y_vals, 'b-', linewidth=2, label=r'$(1 + 2/n)^n$')
ax.axhline(y=np.exp(r), color='r', linestyle='--', label=f'$e^2 = {np.exp(2):.4f}$')
ax.set_xlabel('n')
ax.set_ylabel('Value')
ax.set_title(r'Convergence of $(1 + r/n)^n$ to $e^r$ for $r=2$')
ax.legend()
plt.tight_layout()
plt.show()
```

## Exercise 9: Gradient as Limit (ML)

The derivative is defined as:
$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

For $f(x) = x^3$, compute $f'(2)$ using:
1. The limit definition numerically
2. The analytical formula

```python
# SOLUTION
print("Exercise 9 Solution: Gradient as Limit")
print("="*60)

def f(x):
    return x**3

x = 2
print(f"\nf(x) = x³, find f'({x})")

print("\n--- Part 1: Numerical Approximation ---")
print("Using the limit definition:")
print("f'(x) = lim(h→0) [f(x+h) - f(x)]/h")
print()

for h in [0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]:
    deriv_approx = (f(x + h) - f(x)) / h
    error = abs(deriv_approx - 12)
    print(f"  h = {h:.0e}: [f({x}+h) - f({x})]/h = {deriv_approx:.10f}  error = {error:.2e}")

print("\n--- Part 2: Analytical Formula ---")
print("f(x) = x³")
print("Using the power rule: d/dx(x^n) = n·x^(n-1)")
print("f'(x) = 3x²")
print(f"f'({x}) = 3·{x}² = 3·4 = 12")

print(f"\n--- Comparison ---")
print(f"Numerical limit approaches: 12")
print(f"Analytical result: 12")
print("They match! ✓")

# Central difference (more accurate)
print("\n--- Bonus: Central Difference (more accurate) ---")
print("f'(x) ≈ [f(x+h) - f(x-h)]/(2h)")
for h in [0.1, 0.01, 0.001, 0.0001]:
    central = (f(x + h) - f(x - h)) / (2*h)
    error = abs(central - 12)
    print(f"  h = {h:.0e}: central diff = {central:.12f}  error = {error:.2e}")
```

## Exercise 10: Cross-Entropy Limit (ML)

The binary cross-entropy loss is:
$$H(y, p) = -y \log(p) - (1-y) \log(1-p)$$

Find $\displaystyle \lim_{p \to 0^+} H(1, p)$ and interpret the result.

```python
# SOLUTION
print("Exercise 10 Solution: Cross-Entropy Limit")
print("="*60)

print("\nBinary cross-entropy:")
print("H(y, p) = -y·log(p) - (1-y)·log(1-p)")

print("\n--- Simplify for y = 1 ---")
print("H(1, p) = -1·log(p) - (1-1)·log(1-p)")
print("       = -log(p) - 0")
print("       = -log(p)")

print("\n--- Compute the limit ---")
print("lim(p→0⁺) H(1, p) = lim(p→0⁺) -log(p)")
print("\nAs p → 0⁺:")
print("  log(p) → -∞")
print("  -log(p) → +∞")
print("\nTherefore: lim(p→0⁺) H(1, p) = +∞")

# Numerical verification
def cross_entropy(y, p):
    # Add small epsilon to avoid log(0)
    eps = 1e-15
    p = np.clip(p, eps, 1 - eps)
    return -y * np.log(p) - (1-y) * np.log(1-p)

print("\n--- Numerical Verification ---")
for p in [0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001]:
    H = cross_entropy(1, p)
    print(f"  p = {p:.5f}: H(1, p) = {H:.4f}")

print("\n" + "="*60)
print("INTERPRETATION:")
print("-" * 60)
print("When the true label is y = 1 (positive class):")
print("")
print("• If model predicts p ≈ 1 → Loss ≈ 0 (correct, low penalty)")
print("• If model predicts p ≈ 0 → Loss → ∞ (wrong, severe penalty!)")
print("")
print("This is why cross-entropy is effective for classification:")
print("It heavily penalizes confident wrong predictions!")

# Visualization
fig, ax = plt.subplots(figsize=(10, 5))

p = np.linspace(0.001, 0.999, 1000)
H_y1 = cross_entropy(1, p)
H_y0 = cross_entropy(0, p)

ax.plot(p, H_y1, 'b-', linewidth=2, label='H(y=1, p) = -log(p)')
ax.plot(p, H_y0, 'r-', linewidth=2, label='H(y=0, p) = -log(1-p)')
ax.set_xlabel('Predicted probability p', fontsize=12)
ax.set_ylabel('Cross-Entropy Loss', fontsize=12)
ax.set_title('Cross-Entropy: Severe Penalty for Confident Wrong Predictions', fontsize=14)
ax.legend()
ax.set_ylim(0, 7)

# Annotations
ax.annotate('p→0: H(1,p)→∞\n(predicting 0 when true is 1)', 
            xy=(0.05, 5), fontsize=10, color='blue')
ax.annotate('p→1: H(0,p)→∞\n(predicting 1 when true is 0)', 
            xy=(0.7, 5), fontsize=10, color='red')

plt.tight_layout()
plt.show()
```
