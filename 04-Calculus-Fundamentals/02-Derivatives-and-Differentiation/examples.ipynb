## 1. Derivative from Definition

The derivative is defined as:
$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

```python
print("DERIVATIVE FROM DEFINITION")
print("="*60)
print("\nf(x) = x², find f'(x) using the limit definition")
print("f'(x) = lim(h→0) [f(x+h) - f(x)] / h")

def f(x):
    return x**2

x = 3
print(f"\nCompute f'({x}) numerically:")

for h in [0.1, 0.01, 0.001, 0.0001, 0.00001]:
    derivative_approx = (f(x + h) - f(x)) / h
    error = abs(derivative_approx - 2*x)
    print(f"  h = {h:.5f}: derivative ≈ {derivative_approx:.10f}  (error: {error:.2e})")

print(f"\nAnalytical: f'(x) = 2x, so f'({x}) = {2*x}")
```

```python
# Visualize derivative as tangent line slope
fig, ax = plt.subplots(figsize=(10, 6))

x_vals = np.linspace(0, 5, 100)
y_vals = x_vals**2

# Plot the function
ax.plot(x_vals, y_vals, 'b-', linewidth=2, label=r'$f(x) = x^2$')

# Tangent line at x=3
x0 = 3
slope = 2 * x0  # f'(3) = 6
y0 = x0**2
tangent_x = np.linspace(1, 5, 50)
tangent_y = slope * (tangent_x - x0) + y0

ax.plot(tangent_x, tangent_y, 'r--', linewidth=2, label=f"Tangent at x={x0} (slope={slope})")
ax.plot(x0, y0, 'ro', markersize=10)

# Secant lines for different h values
for h, color in [(1.5, 'lightgreen'), (0.75, 'green')]:
    x1 = x0 + h
    y1 = x1**2
    secant_slope = (y1 - y0) / h
    secant_y = secant_slope * (tangent_x - x0) + y0
    ax.plot(tangent_x, secant_y, '--', color=color, linewidth=1.5, alpha=0.7,
           label=f'Secant h={h} (slope={secant_slope:.2f})')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title('Derivative as Limit of Secant Lines', fontsize=14)
ax.legend()
ax.set_ylim(0, 25)
plt.tight_layout()
plt.show()
```

## 2. Power Rule

$$\frac{d}{dx}x^n = nx^{n-1}$$

Works for any real $n$.

```python
print("POWER RULE")
print("="*60)
print("\nPower Rule: d/dx[x^n] = n·x^(n-1)")

examples = [
    ("x³", 3, lambda x: x**3, lambda x: 3*x**2),
    ("x⁵", 5, lambda x: x**5, lambda x: 5*x**4),
    ("x^(-1) = 1/x", -1, lambda x: x**(-1), lambda x: -x**(-2)),
    ("x^(1/2) = √x", 0.5, lambda x: x**0.5, lambda x: 0.5*x**(-0.5)),
]

x = 2
print(f"\nEvaluate at x = {x}:")
print("-" * 50)

for name, n, f, df in examples:
    h = 0.0001
    numerical = (f(x + h) - f(x)) / h
    analytical = df(x)
    print(f"  f(x) = {name:15} → f'(x) = {n}·x^({n-1})")
    print(f"    f'({x}) = {analytical:.6f}  (numerical: {numerical:.6f})\n")
```

## 3. Chain Rule

For composite functions $f(g(x))$:
$$\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)$$

```python
print("CHAIN RULE")
print("="*60)
print("\nChain Rule: d/dx[f(g(x))] = f'(g(x)) · g'(x)")

print("\n--- Example: f(x) = (3x + 1)⁴ ---")
print("Let u = g(x) = 3x + 1")
print("Then f(x) = u⁴")
print("\nf'(x) = d/du[u⁴] · d/dx[3x + 1]")
print("     = 4u³ · 3")
print("     = 12(3x + 1)³")

def f(x):
    return (3*x + 1)**4

def df(x):
    return 12 * (3*x + 1)**3

x = 1
h = 0.0001
numerical = (f(x + h) - f(x)) / h

print(f"\nAt x = {x}:")
print(f"  f({x}) = (3·{x} + 1)⁴ = 4⁴ = {f(x)}")
print(f"  f'({x}) = 12·(3·{x} + 1)³ = 12·64 = {df(x)}")
print(f"  Numerical verification: {numerical:.4f}")
```

```python
# More chain rule examples
print("\nMore Chain Rule Examples:")
print("="*60)

# e^(x²)
print("\n1. f(x) = e^(x²)")
print("   Outer: e^u, derivative = e^u")
print("   Inner: u = x², derivative = 2x")
print("   f'(x) = e^(x²) · 2x = 2x·e^(x²)")

def f1(x): return np.exp(x**2)
def df1(x): return 2*x*np.exp(x**2)

x = 1
h = 0.0001
print(f"   At x={x}: analytical={df1(x):.6f}, numerical={(f1(x+h)-f1(x))/h:.6f}")

# sin(3x)
print("\n2. f(x) = sin(3x)")
print("   f'(x) = cos(3x) · 3 = 3cos(3x)")

def f2(x): return np.sin(3*x)
def df2(x): return 3*np.cos(3*x)

x = np.pi/6
print(f"   At x=π/6: analytical={df2(x):.6f}, numerical={(f2(x+h)-f2(x))/h:.6f}")

# ln(x² + 1)
print("\n3. f(x) = ln(x² + 1)")
print("   f'(x) = [1/(x²+1)] · 2x = 2x/(x² + 1)")

def f3(x): return np.log(x**2 + 1)
def df3(x): return 2*x / (x**2 + 1)

x = 2
print(f"   At x={x}: analytical={df3(x):.6f}, numerical={(f3(x+h)-f3(x))/h:.6f}")
```

## 4. Product Rule

$$\frac{d}{dx}[f(x) \cdot g(x)] = f'(x)g(x) + f(x)g'(x)$$

```python
print("PRODUCT RULE")
print("="*60)
print("\nProduct Rule: d/dx[f(x)·g(x)] = f'(x)g(x) + f(x)g'(x)")

print("\n--- Example: h(x) = x² · eˣ ---")
print("f(x) = x², f'(x) = 2x")
print("g(x) = eˣ, g'(x) = eˣ")
print("\nh'(x) = 2x·eˣ + x²·eˣ = eˣ(2x + x²) = eˣ·x(2 + x)")

def h(x):
    return x**2 * np.exp(x)

def dh(x):
    return np.exp(x) * (2*x + x**2)

x = 2
h_val = 0.0001
numerical = (h(x + h_val) - h(x)) / h_val

print(f"\nAt x = {x}:")
print(f"  h({x}) = {h(x):.4f}")
print(f"  h'({x}) = e^{x}·(2·{x} + {x}²) = {dh(x):.4f}")
print(f"  Numerical: {numerical:.4f}")
```

## 5. Sigmoid Derivative

The sigmoid function and its derivative:
$$\sigma(x) = \frac{1}{1 + e^{-x}}, \quad \sigma'(x) = \sigma(x)(1 - \sigma(x))$$

```python
print("SIGMOID DERIVATIVE")
print("="*60)

print("\nσ(x) = 1/(1 + e^(-x))")
print("\n--- Derivation ---")
print("Let u = 1 + e^(-x)")
print("σ(x) = u^(-1)")
print("\nσ'(x) = -u^(-2) · d/dx[1 + e^(-x)]")
print("      = -1/(1 + e^(-x))² · (-e^(-x))")
print("      = e^(-x)/(1 + e^(-x))²")

print("\n--- Simplify to σ(1-σ) ---")
print("σ(x) = 1/(1 + e^(-x))")
print("1 - σ(x) = e^(-x)/(1 + e^(-x))")
print("\nσ(x)(1-σ(x)) = [1/(1+e^(-x))] · [e^(-x)/(1+e^(-x))]")
print("            = e^(-x)/(1 + e^(-x))²")
print("            = σ'(x) ✓")

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

print("\n--- Numerical verification ---")
h = 0.0001
for x in [-2, -1, 0, 1, 2]:
    numerical = (sigmoid(x + h) - sigmoid(x)) / h
    analytical = sigmoid_derivative(x)
    print(f"  x = {x:2d}: σ'(x) = {analytical:.6f}, numerical = {numerical:.6f}")
```

```python
# Visualize sigmoid and its derivative
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-6, 6, 1000)
y = sigmoid(x)
dy = sigmoid_derivative(x)

# Sigmoid function
axes[0].plot(x, y, 'b-', linewidth=2, label='σ(x)')
axes[0].axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)
axes[0].axhline(y=1, color='gray', linestyle=':', alpha=0.5)
axes[0].axhline(y=0, color='gray', linestyle=':', alpha=0.5)
axes[0].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('σ(x)', fontsize=12)
axes[0].set_title('Sigmoid Function', fontsize=14)
axes[0].legend()

# Sigmoid derivative
axes[1].plot(x, dy, 'r-', linewidth=2, label="σ'(x) = σ(x)(1-σ(x))")
axes[1].axhline(y=0.25, color='gray', linestyle='--', alpha=0.5, label='Max = 0.25')
axes[1].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[1].fill_between(x, 0, dy, alpha=0.3, color='red')
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel("σ'(x)", fontsize=12)
axes[1].set_title('Sigmoid Derivative (Max at x=0)', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()

print("\nNote: Maximum derivative is only 0.25 at x=0.")
print("This contributes to vanishing gradients in deep networks!")
```

## 6. ReLU Derivative

$$\text{ReLU}(x) = \max(0, x)$$

$$\text{ReLU}'(x) = \begin{cases} 1 & x > 0 \\ 0 & x < 0 \\ \text{undefined} & x = 0 \end{cases}$$

```python
print("RELU DERIVATIVE")
print("="*60)

print("\nReLU(x) = max(0, x)")
print("\nReLU'(x) = { 0 if x < 0")
print("          { 1 if x > 0")
print("          { undefined if x = 0")

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

print("\n--- Values ---")
x_vals = np.array([-2, -1, 0, 1, 2])
for x in x_vals:
    print(f"  x = {x:2d}: ReLU(x) = {relu(x)}, ReLU'(x) = {relu_derivative(np.array([x]))[0]:.0f}")

print("\n--- At x = 0 (special case) ---")
print("Left derivative: lim(h→0⁻) [ReLU(0+h) - ReLU(0)]/h = 0/h = 0")
print("Right derivative: lim(h→0⁺) [ReLU(0+h) - ReLU(0)]/h = h/h = 1")
print("Left ≠ Right → ReLU is not differentiable at x = 0")
print("\nIn practice, we use subgradient (often 0 or 0.5 at x = 0)")
```

```python
# Visualize ReLU and its derivative
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-3, 3, 1000)

# ReLU
axes[0].plot(x, relu(x), 'b-', linewidth=2, label='ReLU(x)')
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[0].plot(0, 0, 'ko', markersize=8)  # Point at origin
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('ReLU(x)', fontsize=12)
axes[0].set_title('ReLU Function', fontsize=14)
axes[0].legend()

# ReLU derivative (step function)
x_neg = np.linspace(-3, 0, 100)
x_pos = np.linspace(0, 3, 100)
axes[1].plot(x_neg, np.zeros_like(x_neg), 'g-', linewidth=2, label="ReLU'(x)")
axes[1].plot(x_pos, np.ones_like(x_pos), 'g-', linewidth=2)
axes[1].plot(0, 0, 'go', markersize=8, fillstyle='none', markeredgewidth=2)
axes[1].plot(0, 1, 'go', markersize=8, fillstyle='none', markeredgewidth=2)
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel("ReLU'(x)", fontsize=12)
axes[1].set_title('ReLU Derivative (Step Function)', fontsize=14)
axes[1].legend()
axes[1].set_ylim(-0.5, 1.5)

plt.tight_layout()
plt.show()
```

## 7. Tanh Derivative

$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

$$\tanh'(x) = 1 - \tanh^2(x) = \text{sech}^2(x)$$

```python
print("TANH DERIVATIVE")
print("="*60)

print("\ntanh(x) = (eˣ - e^(-x))/(eˣ + e^(-x))")
print("tanh'(x) = 1 - tanh²(x) = sech²(x)")

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

print("\n--- Numerical verification ---")
h = 0.0001
for x in [-2, -1, 0, 1, 2]:
    numerical = (np.tanh(x + h) - np.tanh(x)) / h
    analytical = tanh_derivative(x)
    print(f"  x = {x:2d}: tanh'(x) = {analytical:.6f}, numerical = {numerical:.6f}")

print("\nNote: tanh'(0) = 1 (maximum), tanh'(±∞) → 0 (saturation)")
```

```python
# Compare activation derivatives
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-4, 4, 1000)

# Activation functions
axes[0].plot(x, sigmoid(x), 'b-', linewidth=2, label='Sigmoid')
axes[0].plot(x, np.tanh(x), 'g-', linewidth=2, label='Tanh')
axes[0].plot(x, relu(x)/4, 'r-', linewidth=2, label='ReLU/4 (scaled)')  # Scaled for comparison
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('Activation Functions', fontsize=14)
axes[0].legend()

# Derivatives
axes[1].plot(x, sigmoid_derivative(x), 'b-', linewidth=2, label="Sigmoid' (max=0.25)")
axes[1].plot(x, tanh_derivative(x), 'g-', linewidth=2, label="Tanh' (max=1)")
axes[1].plot(x, relu_derivative(x), 'r-', linewidth=2, label="ReLU' (0 or 1)")
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel("f'(x)", fontsize=12)
axes[1].set_title('Activation Derivatives', fontsize=14)
axes[1].legend()
axes[1].set_ylim(-0.1, 1.1)

plt.tight_layout()
plt.show()
```

## 8. Softplus (Smooth ReLU)

Softplus is a smooth approximation to ReLU:
$$\text{softplus}(x) = \ln(1 + e^x)$$
$$\text{softplus}'(x) = \frac{e^x}{1 + e^x} = \sigma(x)$$

```python
print("SOFTPLUS (SMOOTH RELU)")
print("="*60)

print("\nsoftplus(x) = ln(1 + eˣ)")
print("softplus'(x) = eˣ/(1 + eˣ) = σ(x)")
print("\nSoftplus is a smooth approximation to ReLU!")

def softplus(x):
    return np.log1p(np.exp(np.clip(x, -500, 500)))

print("\nComparison:")
print("  x   | ReLU(x) | softplus(x) | softplus'(x)=σ(x)")
print("  " + "-" * 50)
for x_val in [-2, -1, 0, 1, 2, 5]:
    r = relu(x_val)
    sp = softplus(x_val)
    dsp = sigmoid(x_val)
    print(f"  {x_val:2d}  |  {r:5.2f}  |   {sp:6.3f}   |     {dsp:.4f}")
```

```python
# Compare ReLU and Softplus
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-4, 4, 1000)

# Functions
axes[0].plot(x, relu(x), 'b-', linewidth=2, label='ReLU(x)')
axes[0].plot(x, softplus(x), 'r--', linewidth=2, label='Softplus(x)')
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('ReLU vs Softplus', fontsize=14)
axes[0].legend()

# Derivatives
axes[1].plot(x, relu_derivative(x), 'b-', linewidth=2, label="ReLU'(x)")
axes[1].plot(x, sigmoid(x), 'r--', linewidth=2, label="Softplus'(x) = σ(x)")
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel("f'(x)", fontsize=12)
axes[1].set_title('Derivatives', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()

print("Note: Softplus has smooth derivative (no discontinuity at x=0)")
```

## 9. Higher-Order Derivatives

The second derivative measures the rate of change of the slope (curvature).

```python
print("HIGHER-ORDER DERIVATIVES")
print("="*60)

print("\nf(x) = x⁴ - 2x³ + x²")
print("\nFirst derivative:")
print("f'(x) = 4x³ - 6x² + 2x")
print("\nSecond derivative:")
print("f''(x) = 12x² - 12x + 2")
print("\nThird derivative:")
print("f'''(x) = 24x - 12")
print("\nFourth derivative:")
print("f''''(x) = 24")
print("\nFifth derivative:")
print("f'''''(x) = 0")

def f(x): return x**4 - 2*x**3 + x**2
def df(x): return 4*x**3 - 6*x**2 + 2*x
def d2f(x): return 12*x**2 - 12*x + 2
def d3f(x): return 24*x - 12
def d4f(x): return 24 * np.ones_like(x)

x = 1
print(f"\nAt x = {x}:")
print(f"  f({x}) = {f(x)}")
print(f"  f'({x}) = {df(x)}")
print(f"  f''({x}) = {d2f(x)}")
print(f"  f'''({x}) = {d3f(x)}")
print(f"  f''''({x}) = {d4f(x)}")
```

## 10. Critical Points and Extrema

Critical points occur where $f'(x) = 0$ or $f'(x)$ is undefined.

```python
print("CRITICAL POINTS AND EXTREMA")
print("="*60)

print("\nf(x) = x³ - 3x")
print("\n--- Find critical points ---")
print("f'(x) = 3x² - 3 = 3(x² - 1) = 3(x-1)(x+1)")
print("f'(x) = 0 when x = ±1")

print("\n--- Classify using second derivative ---")
print("f''(x) = 6x")
print("\nAt x = -1: f''(-1) = -6 < 0 → Local MAXIMUM")
print("At x = 1:  f''(1) = 6 > 0 → Local MINIMUM")

def f(x): return x**3 - 3*x
def df(x): return 3*x**2 - 3
def d2f(x): return 6*x

print("\n--- Values ---")
for x in [-1, 1]:
    print(f"  x = {x:2d}: f(x) = {f(x):2d}, f'(x) = {df(x)}, f''(x) = {d2f(x):2d}")
```

```python
# Visualize critical points
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(-2.5, 2.5, 1000)
y = f(x)

ax.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^3 - 3x$')

# Mark critical points
ax.plot(-1, f(-1), 'ro', markersize=12, label='Local max at x=-1')
ax.plot(1, f(1), 'go', markersize=12, label='Local min at x=1')

# Tangent lines at critical points (horizontal)
ax.axhline(y=f(-1), color='red', linestyle='--', alpha=0.5, xmin=0.2, xmax=0.5)
ax.axhline(y=f(1), color='green', linestyle='--', alpha=0.5, xmin=0.5, xmax=0.8)

ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='gray', linestyle='-', alpha=0.3)

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title('Critical Points: Where f\'(x) = 0', fontsize=14)
ax.legend()
plt.tight_layout()
plt.show()
```

## 11. Gradient Descent in 1D

The gradient descent update rule:
$$x_{t+1} = x_t - \alpha \cdot f'(x_t)$$

```python
print("GRADIENT DESCENT IN 1D")
print("="*60)

print("\nMinimize: f(x) = (x - 3)²")
print("f'(x) = 2(x - 3)")
print("\nGradient descent: x_{t+1} = x_t - α·f'(x_t)")

def f(x):
    return (x - 3)**2

def df(x):
    return 2 * (x - 3)

# Gradient descent
x = 0  # Starting point
alpha = 0.1  # Learning rate

print(f"\nStarting at x = {x}, learning rate α = {alpha}")
print("\nIteration | x       | f(x)    | f'(x)")
print("-" * 45)

history = [x]
for i in range(12):
    print(f"    {i:2d}    | {x:7.4f} | {f(x):7.4f} | {df(x):7.4f}")
    x = x - alpha * df(x)
    history.append(x)

print(f"\nConverged to x ≈ 3 (the minimum)")
```

```python
# Visualize gradient descent
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x_vals = np.linspace(-1, 5, 100)
y_vals = (x_vals - 3)**2

# Function with GD steps
axes[0].plot(x_vals, y_vals, 'b-', linewidth=2, label=r'$f(x) = (x-3)^2$')
for i, x in enumerate(history[:-1]):
    axes[0].plot(x, f(x), 'ro', markersize=8, alpha=0.5 + 0.5*i/len(history))
    if i < len(history) - 2:
        axes[0].annotate('', xy=(history[i+1], f(history[i+1])), xytext=(x, f(x)),
                        arrowprops=dict(arrowstyle='->', color='red', alpha=0.5))
axes[0].plot(3, 0, 'g*', markersize=15, label='Minimum at x=3')
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('Gradient Descent Path', fontsize=14)
axes[0].legend()

# Convergence
axes[1].plot(range(len(history)), [abs(h - 3) for h in history], 'b-o', linewidth=2)
axes[1].set_xlabel('Iteration', fontsize=12)
axes[1].set_ylabel('|x - 3|', fontsize=12)
axes[1].set_title('Convergence to Minimum', fontsize=14)
axes[1].set_yscale('log')

plt.tight_layout()
plt.show()
```

## 12. Numerical Differentiation Methods

| Method | Formula | Error |
|--------|---------|-------|
| Forward | $(f(x+h) - f(x))/h$ | $O(h)$ |
| Central | $(f(x+h) - f(x-h))/(2h)$ | $O(h^2)$ |

```python
print("NUMERICAL DIFFERENTIATION METHODS")
print("="*60)

def f(x):
    return np.sin(x)

def df_analytical(x):
    return np.cos(x)

x = 1.0
true_derivative = df_analytical(x)

print(f"\nf(x) = sin(x), f'(x) = cos(x)")
print(f"True f'({x}) = cos({x}) = {true_derivative:.12f}")

print("\n--- Forward Difference: [f(x+h) - f(x)] / h ---")
print(f"{'h':>12} | {'Approx':>14} | {'Error':>12}")
print("-" * 45)
for h in [0.1, 0.01, 0.001, 0.0001, 0.00001]:
    forward = (f(x + h) - f(x)) / h
    error = abs(forward - true_derivative)
    print(f"{h:12.5f} | {forward:14.10f} | {error:12.2e}")

print("\n--- Central Difference: [f(x+h) - f(x-h)] / (2h) ---")
print(f"{'h':>12} | {'Approx':>14} | {'Error':>12}")
print("-" * 45)
for h in [0.1, 0.01, 0.001, 0.0001, 0.00001]:
    central = (f(x + h) - f(x - h)) / (2 * h)
    error = abs(central - true_derivative)
    print(f"{h:12.5f} | {central:14.10f} | {error:12.2e}")

print("\nCentral difference is more accurate (O(h²) vs O(h))!")
```

```python
# Visualize error convergence
fig, ax = plt.subplots(figsize=(10, 6))

h_values = np.logspace(-1, -8, 50)
forward_errors = []
central_errors = []

for h in h_values:
    forward = (f(x + h) - f(x)) / h
    central = (f(x + h) - f(x - h)) / (2 * h)
    forward_errors.append(abs(forward - true_derivative))
    central_errors.append(abs(central - true_derivative))

ax.loglog(h_values, forward_errors, 'b-', linewidth=2, label='Forward diff O(h)')
ax.loglog(h_values, central_errors, 'r-', linewidth=2, label='Central diff O(h²)')

# Reference lines
ax.loglog(h_values, h_values, 'b--', alpha=0.5, label='O(h) reference')
ax.loglog(h_values, h_values**2, 'r--', alpha=0.5, label='O(h²) reference')

ax.set_xlabel('Step size h', fontsize=12)
ax.set_ylabel('Error', fontsize=12)
ax.set_title('Numerical Differentiation Error Convergence', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Note: For very small h, roundoff errors dominate (error increases).")
print("Optimal h is typically around 10^-8 for central difference.")
```
