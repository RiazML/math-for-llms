## Exercise 2: Chain Rule Application

Find the derivative of:
- $f(x) = \sin(x^2)$
- $g(x) = e^{\cos(x)}$
- $h(x) = \ln(\sqrt{x^2 + 1})$

```python
# Your work here
```

```python
# Solution
print("CHAIN RULE APPLICATION")
print("="*60)

h = 0.0001

# f(x) = sin(x^2)
print("\n1. f(x) = sin(x²)")
print("   Outer: sin(u), derivative = cos(u)")
print("   Inner: u = x², derivative = 2x")
print("   f'(x) = cos(x²) · 2x = 2x·cos(x²)")

def f(x): return np.sin(x**2)
def df(x): return 2*x*np.cos(x**2)

x = np.pi/4
print(f"   At x=π/4: f'(π/4) = {df(x):.6f}, numerical = {(f(x+h)-f(x))/h:.6f}")

# g(x) = e^(cos(x))
print("\n2. g(x) = e^(cos(x))")
print("   Outer: e^u, derivative = e^u")
print("   Inner: u = cos(x), derivative = -sin(x)")
print("   g'(x) = e^(cos(x)) · (-sin(x)) = -sin(x)·e^(cos(x))")

def g(x): return np.exp(np.cos(x))
def dg(x): return -np.sin(x)*np.exp(np.cos(x))

x = np.pi/6
print(f"   At x=π/6: g'(π/6) = {dg(x):.6f}, numerical = {(g(x+h)-g(x))/h:.6f}")

# h(x) = ln(sqrt(x^2 + 1))
print("\n3. h(x) = ln(√(x² + 1)) = (1/2)ln(x² + 1)")
print("   h'(x) = (1/2) · [1/(x²+1)] · 2x = x/(x² + 1)")

def phi(x): return np.log(np.sqrt(x**2 + 1))
def dphi(x): return x/(x**2 + 1)

x = 2
print(f"   At x={x}: h'({x}) = {dphi(x):.6f}, numerical = {(phi(x+h)-phi(x))/h:.6f}")
```

## Exercise 3: Product and Quotient Rules

Find the derivative of:
- $f(x) = x^2 \cdot \ln(x)$
- $g(x) = \frac{e^x}{x}$
- $h(x) = \frac{x^2 - 1}{x^2 + 1}$

```python
# Your work here
```

```python
# Solution
print("PRODUCT AND QUOTIENT RULES")
print("="*60)

h = 0.0001

# f(x) = x^2 · ln(x)
print("\n1. f(x) = x² · ln(x) (Product Rule)")
print("   f = x², f' = 2x")
print("   g = ln(x), g' = 1/x")
print("   f'(x) = 2x·ln(x) + x²·(1/x) = 2x·ln(x) + x = x(2ln(x) + 1)")

def f(x): return x**2 * np.log(x)
def df(x): return x * (2*np.log(x) + 1)

x = 2
print(f"   At x={x}: f'({x}) = {df(x):.6f}, numerical = {(f(x+h)-f(x))/h:.6f}")

# g(x) = e^x / x
print("\n2. g(x) = eˣ/x (Quotient Rule)")
print("   f = eˣ, f' = eˣ")
print("   g = x, g' = 1")
print("   g'(x) = (eˣ·x - eˣ·1)/x² = eˣ(x-1)/x²")

def g(x): return np.exp(x) / x
def dg(x): return np.exp(x) * (x - 1) / x**2

x = 2
print(f"   At x={x}: g'({x}) = {dg(x):.6f}, numerical = {(g(x+h)-g(x))/h:.6f}")

# h(x) = (x^2 - 1) / (x^2 + 1)
print("\n3. h(x) = (x² - 1)/(x² + 1) (Quotient Rule)")
print("   f = x²-1, f' = 2x")
print("   g = x²+1, g' = 2x")
print("   h'(x) = [2x(x²+1) - (x²-1)·2x] / (x²+1)²")
print("        = 2x[(x²+1) - (x²-1)] / (x²+1)²")
print("        = 2x · 2 / (x²+1)² = 4x/(x²+1)²")

def phi(x): return (x**2 - 1) / (x**2 + 1)
def dphi(x): return 4*x / (x**2 + 1)**2

x = 1
print(f"   At x={x}: h'({x}) = {dphi(x):.6f}, numerical = {(phi(x+h)-phi(x))/h:.6f}")
```

## Exercise 4: Sigmoid Derivative Derivation

Show that $\sigma'(x) = \sigma(x)(1 - \sigma(x))$ where $\sigma(x) = \frac{1}{1+e^{-x}}$.

Then verify numerically and show the gradient saturation problem.

```python
# Your work here
```

```python
# Solution
print("SIGMOID DERIVATIVE DERIVATION")
print("="*60)

print("\nStep 1: Write σ(x) = (1 + e^(-x))^(-1)")
print("\nStep 2: Apply chain rule")
print("σ'(x) = -1·(1 + e^(-x))^(-2) · d/dx[1 + e^(-x)]")
print("      = -1·(1 + e^(-x))^(-2) · (-e^(-x))")
print("      = e^(-x)/(1 + e^(-x))²")

print("\nStep 3: Show this equals σ(x)(1 - σ(x))")
print("σ(x) = 1/(1 + e^(-x))")
print("1 - σ(x) = 1 - 1/(1 + e^(-x))")
print("        = (1 + e^(-x) - 1)/(1 + e^(-x))")
print("        = e^(-x)/(1 + e^(-x))")
print("\nσ(x)(1-σ(x)) = [1/(1+e^(-x))] · [e^(-x)/(1+e^(-x))]")
print("            = e^(-x)/(1 + e^(-x))² = σ'(x) ✓")

def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -500, 500)))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

print("\n--- Numerical Verification ---")
h = 0.0001
for x in [-3, -1, 0, 1, 3]:
    numerical = (sigmoid(x + h) - sigmoid(x)) / h
    analytical = sigmoid_derivative(x)
    print(f"x = {x:2d}: σ'(x) = {analytical:.6f}, numerical = {numerical:.6f}")
```

```python
# Visualize gradient saturation
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(-6, 6, 1000)
y = sigmoid_derivative(x)

ax.plot(x, y, 'b-', linewidth=2)
ax.fill_between(x, 0, y, alpha=0.3)
ax.axhline(y=0.25, color='red', linestyle='--', label='Max gradient = 0.25')
ax.axhline(y=0.01, color='orange', linestyle='--', label='Threshold for "saturated"')

# Shade saturation regions
ax.axvspan(-6, -3, alpha=0.2, color='red', label='Saturated (|x| > 3)')
ax.axvspan(3, 6, alpha=0.2, color='red')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel("σ'(x)", fontsize=12)
ax.set_title('Sigmoid Derivative: Gradient Saturation Problem', fontsize=14)
ax.legend()

plt.tight_layout()
plt.show()

print("When |x| is large, σ'(x) ≈ 0 (vanishing gradients).")
print("This makes deep networks hard to train with sigmoid activations.")
```

## Exercise 5: Implicit Differentiation

Find $\frac{dy}{dx}$ for:
- $x^2 + y^2 = 25$ (circle)
- $xy = 6$ (hyperbola)
- $e^y = x \cdot y$ (transcendental)

```python
# Your work here
```

```python
# Solution
print("IMPLICIT DIFFERENTIATION")
print("="*60)

# x^2 + y^2 = 25
print("\n1. x² + y² = 25 (circle)")
print("   Differentiate both sides with respect to x:")
print("   2x + 2y·(dy/dx) = 0")
print("   dy/dx = -x/y")

print("\n   Check at (3, 4) on the circle:")
x, y = 3, 4
dydx = -x/y
print(f"   Point on circle: {x}² + {y}² = {x**2 + y**2} ✓")
print(f"   dy/dx = -{x}/{y} = {dydx}")
print(f"   Tangent line: y - {y} = {dydx}(x - {x})")

# xy = 6
print("\n2. xy = 6 (hyperbola)")
print("   Use product rule: y + x·(dy/dx) = 0")
print("   dy/dx = -y/x")

print("\n   Check at (2, 3) on the hyperbola:")
x, y = 2, 3
dydx = -y/x
print(f"   xy = {x}·{y} = {x*y} ✓")
print(f"   dy/dx = -{y}/{x} = {dydx}")

# e^y = xy
print("\n3. e^y = xy (transcendental)")
print("   Left: d/dx[e^y] = e^y · (dy/dx)")
print("   Right: d/dx[xy] = y + x·(dy/dx)")
print("   \n   e^y·(dy/dx) = y + x·(dy/dx)")
print("   (dy/dx)·(e^y - x) = y")
print("   dy/dx = y/(e^y - x)")
```

```python
# Visualize implicit curves with tangent lines
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Circle with tangent
theta = np.linspace(0, 2*np.pi, 100)
x_circle = 5 * np.cos(theta)
y_circle = 5 * np.sin(theta)

axes[0].plot(x_circle, y_circle, 'b-', linewidth=2, label='$x^2 + y^2 = 25$')

# Tangent at (3, 4)
px, py = 3, 4
slope = -px/py
tangent_x = np.linspace(0, 6, 50)
tangent_y = slope * (tangent_x - px) + py
axes[0].plot(tangent_x, tangent_y, 'r--', linewidth=2, label=f'Tangent (slope={slope:.2f})')
axes[0].plot(px, py, 'ro', markersize=10)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('y', fontsize=12)
axes[0].set_title('Circle: Implicit Differentiation', fontsize=14)
axes[0].legend()
axes[0].set_aspect('equal')
axes[0].set_xlim(-6, 7)
axes[0].set_ylim(-6, 7)

# Hyperbola with tangent
x_hyp = np.linspace(0.5, 6, 100)
y_hyp = 6 / x_hyp

axes[1].plot(x_hyp, y_hyp, 'b-', linewidth=2, label='$xy = 6$')

# Tangent at (2, 3)
px, py = 2, 3
slope = -py/px
tangent_x = np.linspace(0.5, 4, 50)
tangent_y = slope * (tangent_x - px) + py
axes[1].plot(tangent_x, tangent_y, 'r--', linewidth=2, label=f'Tangent (slope={slope:.2f})')
axes[1].plot(px, py, 'ro', markersize=10)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('y', fontsize=12)
axes[1].set_title('Hyperbola: Implicit Differentiation', fontsize=14)
axes[1].legend()
axes[1].set_xlim(0, 6)
axes[1].set_ylim(0, 8)

plt.tight_layout()
plt.show()
```

## Exercise 6: Finding Extrema

Find all local maxima and minima for:
$$f(x) = x^4 - 8x^2 + 16$$

Use both the first and second derivative tests.

```python
# Your work here
```

```python
# Solution
print("FINDING EXTREMA")
print("="*60)

print("\nf(x) = x⁴ - 8x² + 16")

print("\n--- Step 1: Find critical points ---")
print("f'(x) = 4x³ - 16x = 4x(x² - 4) = 4x(x-2)(x+2)")
print("f'(x) = 0 when x = 0, x = 2, x = -2")

print("\n--- Step 2: Second derivative test ---")
print("f''(x) = 12x² - 16")

def f(x): return x**4 - 8*x**2 + 16
def df(x): return 4*x**3 - 16*x
def d2f(x): return 12*x**2 - 16

critical_points = [-2, 0, 2]
print("\nClassification:")
for x in critical_points:
    second = d2f(x)
    if second > 0:
        classification = "LOCAL MINIMUM"
    elif second < 0:
        classification = "LOCAL MAXIMUM"
    else:
        classification = "INCONCLUSIVE"
    print(f"  x = {x:2d}: f(x) = {f(x):2d}, f'(x) = {df(x):2d}, f''(x) = {second:3d} → {classification}")

print("\n--- Summary ---")
print("Local maximum: x = 0, f(0) = 16")
print("Local minima: x = ±2, f(±2) = 0")
```

```python
# Visualize
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(-3, 3, 1000)
y = f(x)

ax.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^4 - 8x^2 + 16$')

# Mark critical points
ax.plot(0, f(0), 'ro', markersize=12, label='Local max at (0, 16)')
ax.plot(-2, f(-2), 'go', markersize=12, label='Local min at (-2, 0)')
ax.plot(2, f(2), 'go', markersize=12, label='Local min at (2, 0)')

ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title('Finding Extrema Using First and Second Derivative Tests', fontsize=14)
ax.legend()

plt.tight_layout()
plt.show()
```

## Exercise 7: Gradient Descent Implementation

Implement gradient descent to find the minimum of:
$$f(x) = x^4 - 3x^3 + 2$$

Try different learning rates and starting points.

```python
# Your work here
```

```python
# Solution
print("GRADIENT DESCENT IMPLEMENTATION")
print("="*60)

def f(x):
    return x**4 - 3*x**3 + 2

def df(x):
    return 4*x**3 - 9*x**2

def gradient_descent(x0, learning_rate, n_iters, verbose=True):
    """Run gradient descent and return history."""
    x = x0
    history = [x]
    
    for i in range(n_iters):
        grad = df(x)
        x = x - learning_rate * grad
        history.append(x)
        
        if verbose and i < 5:
            print(f"Iter {i+1}: x = {x:.6f}, f(x) = {f(x):.6f}, f'(x) = {grad:.6f}")
    
    return history

print("\nf(x) = x⁴ - 3x³ + 2")
print("f'(x) = 4x³ - 9x²")
print("\nCritical points: f'(x) = x²(4x - 9) = 0")
print("x = 0 (saddle point) or x = 9/4 = 2.25 (potential minimum)")

print("\n--- Gradient Descent with α = 0.01, x₀ = 3 ---")
history1 = gradient_descent(x0=3, learning_rate=0.01, n_iters=100)
print(f"...")
print(f"Final: x = {history1[-1]:.6f}, f(x) = {f(history1[-1]):.6f}")
```

```python
# Compare different learning rates
print("\n--- Comparing Learning Rates ---")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x_vals = np.linspace(-1, 4, 200)
y_vals = f(x_vals)

# Different learning rates
learning_rates = [0.001, 0.01, 0.02]
colors = ['red', 'green', 'blue']

axes[0].plot(x_vals, y_vals, 'k-', linewidth=2, label='f(x)')

for lr, color in zip(learning_rates, colors):
    hist = gradient_descent(x0=3.5, learning_rate=lr, n_iters=50, verbose=False)
    axes[0].plot(hist, [f(x) for x in hist], 'o-', color=color, alpha=0.5, label=f'α={lr}')

axes[0].plot(2.25, f(2.25), 'k*', markersize=15, label='Minimum at x=2.25')
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('GD Path with Different Learning Rates', fontsize=14)
axes[0].legend()

# Convergence comparison
for lr, color in zip(learning_rates, colors):
    hist = gradient_descent(x0=3.5, learning_rate=lr, n_iters=100, verbose=False)
    errors = [abs(x - 2.25) for x in hist]
    axes[1].semilogy(errors, '-', color=color, linewidth=2, label=f'α={lr}')

axes[1].set_xlabel('Iteration', fontsize=12)
axes[1].set_ylabel('|x - 2.25|', fontsize=12)
axes[1].set_title('Convergence Speed', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Exercise 8: Second Derivative and Concavity

Analyze the concavity of:
$$f(x) = x^3 - 6x^2 + 9x + 1$$

Find inflection points and intervals of concavity.

```python
# Your work here
```

```python
# Solution
print("SECOND DERIVATIVE AND CONCAVITY")
print("="*60)

def f(x): return x**3 - 6*x**2 + 9*x + 1
def df(x): return 3*x**2 - 12*x + 9
def d2f(x): return 6*x - 12

print("\nf(x) = x³ - 6x² + 9x + 1")
print("f'(x) = 3x² - 12x + 9 = 3(x² - 4x + 3) = 3(x-1)(x-3)")
print("f''(x) = 6x - 12 = 6(x - 2)")

print("\n--- Inflection Point ---")
print("f''(x) = 0 when x = 2")
print(f"Inflection point: (2, f(2)) = (2, {f(2)})")

print("\n--- Concavity Analysis ---")
print("For x < 2: f''(x) < 0 → Concave DOWN")
print("For x > 2: f''(x) > 0 → Concave UP")

print("\n--- Critical Points ---")
print("f'(x) = 0 when x = 1 or x = 3")
for x in [1, 3]:
    print(f"  x = {x}: f({x}) = {f(x)}, f''({x}) = {d2f(x)} → {'Local min' if d2f(x) > 0 else 'Local max'}")
```

```python
# Visualize concavity
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-0.5, 4.5, 1000)

# Function with concavity regions
axes[0].plot(x, f(x), 'b-', linewidth=2, label='f(x)')

# Shade concavity regions
x_down = x[x < 2]
x_up = x[x >= 2]
axes[0].fill_between(x_down, f(x_down), alpha=0.2, color='red', label='Concave down (x < 2)')
axes[0].fill_between(x_up, f(x_up), alpha=0.2, color='green', label='Concave up (x > 2)')

# Mark points
axes[0].plot(1, f(1), 'go', markersize=10, label=f'Local max (1, {f(1)})')
axes[0].plot(3, f(3), 'ro', markersize=10, label=f'Local min (3, {f(3)})')
axes[0].plot(2, f(2), 'ko', markersize=10, label=f'Inflection (2, {f(2)})')

axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('Concavity Analysis', fontsize=14)
axes[0].legend(fontsize=8)

# Plot first and second derivatives
axes[1].plot(x, df(x), 'g-', linewidth=2, label="f'(x) = 3(x-1)(x-3)")
axes[1].plot(x, d2f(x), 'r-', linewidth=2, label="f''(x) = 6(x-2)")
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.5)
axes[1].axvline(x=2, color='black', linestyle='--', alpha=0.5, label='Inflection x=2')
axes[1].plot([1, 3], [0, 0], 'go', markersize=8)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('Derivative', fontsize=12)
axes[1].set_title('First and Second Derivatives', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Exercise 9: Leaky ReLU Derivative

The Leaky ReLU is defined as:
$$\text{LeakyReLU}(x) = \begin{cases} x & x \geq 0 \\ \alpha x & x < 0 \end{cases}$$

where $\alpha$ is a small positive constant (typically 0.01).

Find the derivative and compare with standard ReLU.

```python
# Your work here
```

```python
# Solution
print("LEAKY RELU DERIVATIVE")
print("="*60)

alpha = 0.01

print(f"\nLeaky ReLU with α = {alpha}:")
print("LeakyReLU(x) = { x      if x ≥ 0")
print("              { αx     if x < 0")

print("\n--- Derivative ---")
print("LeakyReLU'(x) = { 1     if x > 0")
print("               { α     if x < 0")
print("               { undefined at x = 0 (but continuous if α = 1)")

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x >= 0, 1.0, alpha)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1.0, 0.0)

print("\n--- Comparison with ReLU ---")
print(f"{'x':>5} | {'ReLU':>8} | {'ReLU\'':>6} | {'LeakyReLU':>10} | {'LeakyReLU\'':>11}")
print("-" * 55)
for x in [-2, -1, 0, 1, 2]:
    print(f"{x:5d} | {relu(x):8.2f} | {relu_derivative(x):6.2f} | {leaky_relu(x):10.4f} | {leaky_relu_derivative(x):11.4f}")

print("\n--- Advantage ---")
print("Leaky ReLU has non-zero gradient for x < 0,")
print("which helps avoid 'dying ReLU' problem where gradients become zero.")
```

## Exercise 10: MSE Loss Derivative

The Mean Squared Error loss for predictions $\hat{y}$ and targets $y$ is:
$$\mathcal{L} = \frac{1}{n}\sum_{i=1}^{n}(\hat{y}_i - y_i)^2$$

If $\hat{y} = wx + b$ (linear model), find $\frac{\partial \mathcal{L}}{\partial w}$ and $\frac{\partial \mathcal{L}}{\partial b}$.

```python
# Your work here
```

```python
# Solution
print("MSE LOSS DERIVATIVE")
print("="*60)

print("\nFor linear model: ŷ = wx + b")
print("MSE Loss: L = (1/n) Σ(ŷᵢ - yᵢ)²")
print("            = (1/n) Σ(wxᵢ + b - yᵢ)²")

print("\n--- Gradient with respect to w ---")
print("∂L/∂w = (1/n) Σ 2(wxᵢ + b - yᵢ) · xᵢ")
print("      = (2/n) Σ (ŷᵢ - yᵢ) · xᵢ")

print("\n--- Gradient with respect to b ---")
print("∂L/∂b = (1/n) Σ 2(wxᵢ + b - yᵢ) · 1")
print("      = (2/n) Σ (ŷᵢ - yᵢ)")

print("\n--- Vector form ---")
print("∂L/∂w = (2/n) · (ŷ - y)ᵀ · x")
print("∂L/∂b = (2/n) · Σ(ŷ - y) = (2/n) · 1ᵀ(ŷ - y)")
```

```python
# Numerical verification
print("\n--- Numerical Verification ---")

# Generate sample data
np.random.seed(42)
n = 100
x = np.random.randn(n)
y_true = 2.5 * x + 1.0 + 0.1 * np.random.randn(n)  # True: w=2.5, b=1.0

def mse_loss(w, b):
    y_pred = w * x + b
    return np.mean((y_pred - y_true)**2)

def mse_gradients_analytical(w, b):
    """Compute analytical gradients."""
    y_pred = w * x + b
    error = y_pred - y_true
    dL_dw = (2/n) * np.sum(error * x)
    dL_db = (2/n) * np.sum(error)
    return dL_dw, dL_db

def mse_gradients_numerical(w, b, h=1e-5):
    """Compute numerical gradients."""
    dL_dw = (mse_loss(w + h, b) - mse_loss(w - h, b)) / (2 * h)
    dL_db = (mse_loss(w, b + h) - mse_loss(w, b - h)) / (2 * h)
    return dL_dw, dL_db

# Test at w=2, b=0.5
w, b = 2.0, 0.5

analytical = mse_gradients_analytical(w, b)
numerical = mse_gradients_numerical(w, b)

print(f"\nAt w={w}, b={b}:")
print(f"  Loss = {mse_loss(w, b):.6f}")
print(f"  \n  ∂L/∂w: analytical = {analytical[0]:10.6f}, numerical = {numerical[0]:10.6f}")
print(f"  ∂L/∂b: analytical = {analytical[1]:10.6f}, numerical = {numerical[1]:10.6f}")
```

```python
# Gradient descent on MSE
print("\n--- Gradient Descent to Find Optimal w, b ---")

w, b = 0.0, 0.0  # Start from zero
lr = 0.1
history = [(w, b, mse_loss(w, b))]

for i in range(100):
    dL_dw, dL_db = mse_gradients_analytical(w, b)
    w = w - lr * dL_dw
    b = b - lr * dL_db
    history.append((w, b, mse_loss(w, b)))

print(f"\nStarting: w=0, b=0, Loss={history[0][2]:.4f}")
print(f"After 100 iterations: w={w:.4f}, b={b:.4f}, Loss={mse_loss(w, b):.4f}")
print(f"\nTrue values: w=2.5, b=1.0")

# Visualize convergence
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss over time
losses = [h[2] for h in history]
axes[0].plot(losses, 'b-', linewidth=2)
axes[0].set_xlabel('Iteration', fontsize=12)
axes[0].set_ylabel('MSE Loss', fontsize=12)
axes[0].set_title('Loss Convergence', fontsize=14)

# Parameter trajectory
ws = [h[0] for h in history]
bs = [h[1] for h in history]
axes[1].plot(ws, bs, 'b.-', alpha=0.5, label='GD path')
axes[1].plot(ws[0], bs[0], 'go', markersize=10, label='Start (0, 0)')
axes[1].plot(ws[-1], bs[-1], 'ro', markersize=10, label=f'End ({ws[-1]:.2f}, {bs[-1]:.2f})')
axes[1].plot(2.5, 1.0, 'k*', markersize=15, label='True (2.5, 1.0)')
axes[1].set_xlabel('w', fontsize=12)
axes[1].set_ylabel('b', fontsize=12)
axes[1].set_title('Parameter Space Trajectory', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```
