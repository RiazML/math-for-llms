```python
class GCNLayerWithGradSolution:
    """Solution for Exercise 1."""
    
    def __init__(self, in_features: int, out_features: int):
        scale = np.sqrt(2.0 / (in_features + out_features))
        self.W = np.random.randn(in_features, out_features) * scale
        self.b = np.zeros(out_features)
        self.cache = {}
    
    def forward(self, A_hat: np.ndarray, H: np.ndarray) -> np.ndarray:
        Z = A_hat @ H @ self.W + self.b
        H_out = np.maximum(0, Z)  # ReLU
        
        self.cache = {'A_hat': A_hat, 'H': H, 'Z': Z}
        return H_out
    
    def backward(self, grad_output: np.ndarray):
        A_hat = self.cache['A_hat']
        H = self.cache['H']
        Z = self.cache['Z']
        
        # Backprop through ReLU
        grad_Z = grad_output * (Z > 0)
        
        # Gradients
        grad_W = (A_hat @ H).T @ grad_Z
        grad_b = grad_Z.sum(axis=0)
        grad_H = A_hat.T @ grad_Z @ self.W.T
        
        return grad_H, grad_W, grad_b

# Test solution
layer = GCNLayerWithGradSolution(d_in, d_out)
H_out = layer.forward(A_hat, H)
print(f"Forward output shape: {H_out.shape}")

# Test backward
grad_output = np.random.randn(*H_out.shape)
grad_H, grad_W, grad_b = layer.backward(grad_output)
print(f"Gradient shapes - H: {grad_H.shape}, W: {grad_W.shape}, b: {grad_b.shape}")
```

```python
def scaled_dot_product_attention_solution(Q: np.ndarray, K: np.ndarray,
                                         A: np.ndarray) -> np.ndarray:
    """Solution for scaled dot-product attention."""
    d = Q.shape[1]
    scores = Q @ K.T / np.sqrt(d)
    
    # Mask non-edges
    A_with_self = A + np.eye(len(A))
    scores = np.where(A_with_self > 0, scores, -1e9)
    
    # Softmax
    exp_scores = np.exp(scores - scores.max(axis=1, keepdims=True))
    attention = exp_scores / (exp_scores.sum(axis=1, keepdims=True) + 1e-10)
    
    return attention

def additive_attention_solution(Q: np.ndarray, K: np.ndarray,
                               A: np.ndarray, W_q: np.ndarray,
                               W_k: np.ndarray, a: np.ndarray) -> np.ndarray:
    """Solution for additive attention."""
    n = len(Q)
    hidden_dim = W_q.shape[1]
    
    # Transform Q and K
    Wq = Q @ W_q  # (n, hidden_dim)
    Wk = K @ W_k  # (n, hidden_dim)
    
    # Compute pairwise scores
    scores = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            scores[i, j] = a @ np.tanh(Wq[i] + Wk[j])
    
    # Mask and softmax
    A_with_self = A + np.eye(n)
    scores = np.where(A_with_self > 0, scores, -1e9)
    
    exp_scores = np.exp(scores - scores.max(axis=1, keepdims=True))
    attention = exp_scores / (exp_scores.sum(axis=1, keepdims=True) + 1e-10)
    
    return attention

def general_attention_solution(Q: np.ndarray, K: np.ndarray,
                              A: np.ndarray, W: np.ndarray) -> np.ndarray:
    """Solution for general bilinear attention."""
    scores = Q @ W @ K.T
    
    # Mask and softmax
    A_with_self = A + np.eye(len(A))
    scores = np.where(A_with_self > 0, scores, -1e9)
    
    exp_scores = np.exp(scores - scores.max(axis=1, keepdims=True))
    attention = exp_scores / (exp_scores.sum(axis=1, keepdims=True) + 1e-10)
    
    return attention

# Test solutions
att_dot = scaled_dot_product_attention_solution(Q, K, A)
print(f"Scaled dot-product attention shape: {att_dot.shape}")
print(f"Row sums (should be 1): {np.round(att_dot.sum(axis=1), 4)}")

W = np.random.randn(d, d)
att_general = general_attention_solution(Q, K, A, W)
print(f"General attention row sums: {np.round(att_general.sum(axis=1), 4)}")
```

### Solution 3

```python
def uniform_sample_solution(adj_list: Dict[int, List[int]],
                           node: int,
                           num_samples: int) -> List[int]:
    """Uniform random sampling without replacement."""
    neighbors = adj_list.get(node, [])
    if len(neighbors) <= num_samples:
        return neighbors
    return list(np.random.choice(neighbors, num_samples, replace=False))

def layer_wise_sampling_solution(adj_list, batch_nodes, fanouts):
    """Layer-wise sampling for multi-layer GNNs."""
    layers = []
    current_nodes = set(batch_nodes)
    
    for fanout in reversed(fanouts):
        sampled_adj = {}
        next_nodes = set()
        
        for v in current_nodes:
            neighbors = adj_list.get(v, [])
            if len(neighbors) <= fanout:
                sampled = neighbors
            else:
                sampled = list(np.random.choice(neighbors, fanout, replace=False))
            sampled_adj[v] = sampled
            next_nodes.update(sampled)
        
        layers.append({
            'target_nodes': list(current_nodes),
            'sampled_adj': sampled_adj,
            'source_nodes': list(next_nodes)
        })
        
        current_nodes = current_nodes.union(next_nodes)
    
    return layers[::-1]

# Test solutions
np.random.seed(42)
samples = uniform_sample_solution(adj_list, node=0, num_samples=3)
print(f"Sampled neighbors of node 0: {samples}")

layers = layer_wise_sampling_solution(adj_list, batch_nodes=[0, 5], fanouts=[3, 2])
print(f"\nLayer-wise sampling for nodes [0, 5]:")
for i, layer in enumerate(layers):
    print(f"  Layer {i}: {len(layer['target_nodes'])} target nodes")
```

### Solution 4

```python
class MLPSolution:
    """Simple MLP for GIN."""
    
    def __init__(self, layers: List[int]):
        self.weights = []
        self.biases = []
        
        for i in range(len(layers) - 1):
            scale = np.sqrt(2.0 / (layers[i] + layers[i+1]))
            W = np.random.randn(layers[i], layers[i+1]) * scale
            b = np.zeros(layers[i+1])
            self.weights.append(W)
            self.biases.append(b)
    
    def forward(self, x: np.ndarray) -> np.ndarray:
        for i, (W, b) in enumerate(zip(self.weights, self.biases)):
            x = x @ W + b
            if i < len(self.weights) - 1:
                x = np.maximum(0, x)  # ReLU
        return x

class GINConvSolution:
    """GIN convolution layer."""
    
    def __init__(self, in_dim: int, out_dim: int, 
                eps: float = 0.0, train_eps: bool = True):
        self.eps = eps
        self.train_eps = train_eps
        self.mlp = MLPSolution([in_dim, in_dim * 2, out_dim])
    
    def forward(self, A: np.ndarray, H: np.ndarray) -> np.ndarray:
        # Aggregate neighbors (sum)
        neighbor_sum = A @ H
        
        # Update: (1 + eps) * h_v + sum(h_u)
        H_combined = (1 + self.eps) * H + neighbor_sum
        
        # Apply MLP
        return np.maximum(0, self.mlp.forward(H_combined))

# Test solution - can GIN distinguish different graphs?
np.random.seed(42)

# Initial features (same for both graphs)
X1 = np.ones((4, 4))
X2 = np.ones((4, 4))

gin = GINConvSolution(in_dim=4, out_dim=4, eps=0.1)

# Run multiple layers
H1, H2 = X1.copy(), X2.copy()
for _ in range(3):
    H1 = gin.forward(A1, H1)
    H2 = gin.forward(A2, H2)

# Graph-level representation (sum pooling)
g1_repr = H1.sum(axis=0)
g2_repr = H2.sum(axis=0)

print(f"Graph 1 representation: {np.round(g1_repr, 3)}")
print(f"Graph 2 representation: {np.round(g2_repr, 3)}")
print(f"Difference: {np.linalg.norm(g1_repr - g2_repr):.4f}")
print("GIN can distinguish these non-isomorphic graphs!" if np.linalg.norm(g1_repr - g2_repr) > 0.1 else "Graphs have similar representations")
```

### Solution 5

```python
class DiffPoolSolution:
    """Differentiable Pooling solution."""
    
    def __init__(self, in_dim: int, out_dim: int, num_clusters: int):
        self.num_clusters = num_clusters
        
        # Embedding GNN weights
        scale = np.sqrt(2.0 / (in_dim + out_dim))
        self.W_embed = np.random.randn(in_dim, out_dim) * scale
        
        # Assignment GNN weights
        self.W_assign = np.random.randn(in_dim, num_clusters) * scale
    
    def _gcn(self, A: np.ndarray, H: np.ndarray, W: np.ndarray) -> np.ndarray:
        """Simple GCN layer."""
        n = len(A)
        A_tilde = A + np.eye(n)
        D_inv_sqrt = np.diag(1.0 / np.sqrt(A_tilde.sum(axis=1) + 1e-10))
        A_hat = D_inv_sqrt @ A_tilde @ D_inv_sqrt
        return np.maximum(0, A_hat @ H @ W)
    
    def forward(self, A: np.ndarray, H: np.ndarray):
        # Compute embeddings
        Z = self._gcn(A, H, self.W_embed)
        
        # Compute soft assignments
        S_logits = self._gcn(A, H, self.W_assign)
        exp_S = np.exp(S_logits - S_logits.max(axis=1, keepdims=True))
        S = exp_S / (exp_S.sum(axis=1, keepdims=True) + 1e-10)
        
        # Pool features: H' = S^T Z
        H_pooled = S.T @ Z
        
        # Pool adjacency: A' = S^T A S
        A_pooled = S.T @ A @ S
        
        self.S = S  # Save for loss computation
        return A_pooled, H_pooled
    
    def link_loss(self, A: np.ndarray, S: np.ndarray) -> float:
        A_reconstructed = S @ S.T
        return np.linalg.norm(A - A_reconstructed, 'fro') ** 2
    
    def entropy_loss(self, S: np.ndarray) -> float:
        return -np.sum(S * np.log(S + 1e-10))

class SAGPoolingSolution:
    """Self-Attention Graph Pooling solution."""
    
    def __init__(self, in_dim: int, ratio: float = 0.5):
        self.ratio = ratio
        scale = np.sqrt(2.0 / in_dim)
        self.W_score = np.random.randn(in_dim, 1) * scale
    
    def forward(self, A: np.ndarray, H: np.ndarray):
        n = len(H)
        k = max(1, int(n * self.ratio))
        
        # GCN-based scoring
        A_tilde = A + np.eye(n)
        D_inv_sqrt = np.diag(1.0 / np.sqrt(A_tilde.sum(axis=1) + 1e-10))
        A_hat = D_inv_sqrt @ A_tilde @ D_inv_sqrt
        
        H_scored = A_hat @ H
        scores = (H_scored @ self.W_score).flatten()
        
        # Select top-k
        top_indices = np.argsort(scores)[-k:]
        
        # Mask features with tanh(score)
        gate = np.tanh(scores[top_indices])
        H_pooled = H[top_indices] * gate.reshape(-1, 1)
        
        # Create pooled adjacency
        A_pooled = A[np.ix_(top_indices, top_indices)]
        
        return A_pooled, H_pooled, top_indices

# Test solutions
np.random.seed(42)

diffpool = DiffPoolSolution(in_dim=8, out_dim=8, num_clusters=4)
A_pooled, H_pooled = diffpool.forward(A, H)
print(f"DiffPool: {n} nodes → {len(H_pooled)} clusters")
print(f"  Link loss: {diffpool.link_loss(A, diffpool.S):.4f}")
print(f"  Entropy loss: {diffpool.entropy_loss(diffpool.S):.4f}")

sagpool = SAGPoolingSolution(in_dim=8, ratio=0.5)
A_pooled_sag, H_pooled_sag, indices = sagpool.forward(A, H)
print(f"\nSAGPooling (ratio=0.5): {n} nodes → {len(H_pooled_sag)} nodes")
print(f"  Selected nodes: {sorted(indices)}")
```

```python
class PairNormSolution:
    """PairNorm solution."""
    
    def forward(self, H: np.ndarray) -> np.ndarray:
        # Center
        H_centered = H - H.mean(axis=0, keepdims=True)
        
        # Scale
        norm_sq = (H_centered ** 2).sum() / len(H)
        H_normed = H_centered / (np.sqrt(norm_sq) + 1e-10)
        
        return H_normed

class JumpingKnowledgeSolution:
    """Jumping Knowledge solution."""
    
    def __init__(self, hidden_dim: int, num_layers: int, mode: str = 'concat'):
        self.mode = mode
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
    
    def aggregate(self, layer_outputs: List[np.ndarray]) -> np.ndarray:
        if self.mode == 'concat':
            return np.concatenate(layer_outputs, axis=1)
        elif self.mode == 'max':
            stacked = np.stack(layer_outputs, axis=0)
            return stacked.max(axis=0)
        elif self.mode == 'mean':
            stacked = np.stack(layer_outputs, axis=0)
            return stacked.mean(axis=0)
        else:
            raise ValueError(f"Unknown mode: {self.mode}")

class ResidualGCNSolution:
    """Residual GCN solution."""
    
    def __init__(self, in_dim: int, hidden_dim: int, out_dim: int, num_layers: int):
        self.num_layers = num_layers
        self.weights = []
        
        # First layer
        scale = np.sqrt(2.0 / (in_dim + hidden_dim))
        self.weights.append(np.random.randn(in_dim, hidden_dim) * scale)
        
        # Hidden layers (same dimension for residual)
        for _ in range(num_layers - 2):
            scale = np.sqrt(2.0 / (hidden_dim + hidden_dim))
            self.weights.append(np.random.randn(hidden_dim, hidden_dim) * scale)
        
        # Output layer
        scale = np.sqrt(2.0 / (hidden_dim + out_dim))
        self.weights.append(np.random.randn(hidden_dim, out_dim) * scale)
    
    def forward(self, A: np.ndarray, X: np.ndarray) -> np.ndarray:
        n = len(A)
        A_tilde = A + np.eye(n)
        D_inv_sqrt = np.diag(1.0 / np.sqrt(A_tilde.sum(axis=1) + 1e-10))
        A_hat = D_inv_sqrt @ A_tilde @ D_inv_sqrt
        
        H = X
        
        for i, W in enumerate(self.weights):
            H_new = np.maximum(0, A_hat @ H @ W)
            
            # Residual connection (only if dimensions match)
            if i > 0 and i < len(self.weights) - 1 and H.shape[1] == H_new.shape[1]:
                H = H + H_new
            else:
                H = H_new
        
        return H

# Test solutions - compare MAD with and without mitigation
def compute_mad(H):
    n = len(H)
    total = 0
    for i in range(n):
        for j in range(i+1, n):
            total += np.linalg.norm(H[i] - H[j])
    return total / (n * (n-1) / 2)

np.random.seed(42)

# Standard GCN (will over-smooth)
A_tilde = A + np.eye(n)
D_inv_sqrt = np.diag(1.0 / np.sqrt(A_tilde.sum(axis=1) + 1e-10))
A_hat = D_inv_sqrt @ A_tilde @ D_inv_sqrt

H_standard = X.copy()
H_pairnorm = X.copy()
pairnorm = PairNormSolution()

print("MAD after each layer:")
print("Layer | Standard | With PairNorm")
print("-" * 40)

for layer in range(10):
    H_standard = A_hat @ H_standard
    H_pairnorm = pairnorm.forward(A_hat @ H_pairnorm)
    
    mad_std = compute_mad(H_standard)
    mad_pn = compute_mad(H_pairnorm)
    print(f"{layer+1:5d} | {mad_std:8.4f} | {mad_pn:8.4f}")

print("\nPairNorm maintains feature diversity!")
```

```python
class HeteroGNNLayerSolution:
    """Heterogeneous GNN layer solution."""
    
    def __init__(self, node_types: List[str],
                edge_types: List[Tuple[str, str, str]],
                in_dims: Dict[str, int],
                out_dim: int):
        self.node_types = node_types
        self.edge_types = edge_types
        self.out_dim = out_dim
        
        # Per-node-type weights
        self.W_node = {}
        for ntype in node_types:
            scale = np.sqrt(2.0 / (in_dims[ntype] + out_dim))
            self.W_node[ntype] = np.random.randn(in_dims[ntype], out_dim) * scale
        
        # Per-edge-type weights
        self.W_edge = {}
        for src_type, etype, dst_type in edge_types:
            scale = np.sqrt(2.0 / (in_dims[src_type] + out_dim))
            self.W_edge[(src_type, etype, dst_type)] = np.random.randn(
                in_dims[src_type], out_dim) * scale
    
    def forward(self, H: Dict[str, np.ndarray],
               edges: Dict[Tuple[str, str, str], 
                          Tuple[np.ndarray, np.ndarray]]) -> Dict[str, np.ndarray]:
        
        # Initialize output with transformed self features
        H_out = {
            ntype: np.maximum(0, H[ntype] @ self.W_node[ntype])
            for ntype in self.node_types
        }
        
        # Message passing for each edge type
        for edge_key, (src_idx, dst_idx) in edges.items():
            src_type, etype, dst_type = edge_key
            W = self.W_edge[edge_key]
            
            # Transform source features
            src_feats = H[src_type][src_idx] @ W
            
            # Aggregate to destination nodes
            n_dst = len(H[dst_type])
            for i, (s, d) in enumerate(zip(src_idx, dst_idx)):
                H_out[dst_type][d] += src_feats[i]
        
        # Apply activation
        for ntype in self.node_types:
            H_out[ntype] = np.maximum(0, H_out[ntype])
        
        return H_out

# Test solution
np.random.seed(42)

# Create test data
num_users, num_items = 5, 8
H = {
    'user': np.random.randn(num_users, 8),
    'item': np.random.randn(num_items, 16)
}

# User-Item edges (user 0 buys items 1, 3; user 1 buys items 0, 2, etc.)
edges = {
    ('user', 'buys', 'item'): (
        np.array([0, 0, 1, 1, 2, 3, 4]),  # src (user)
        np.array([1, 3, 0, 2, 5, 6, 7])   # dst (item)
    ),
    ('item', 'bought_by', 'user'): (
        np.array([1, 3, 0, 2, 5, 6, 7]),  # src (item)
        np.array([0, 0, 1, 1, 2, 3, 4])   # dst (user)
    )
}

het_gnn = HeteroGNNLayerSolution(
    node_types=node_types,
    edge_types=edge_types,
    in_dims={'user': 8, 'item': 16},
    out_dim=32
)

H_out = het_gnn.forward(H, edges)
print(f"User features: {H['user'].shape} → {H_out['user'].shape}")
print(f"Item features: {H['item'].shape} → {H_out['item'].shape}")
```

```python
class TimeEncodeSolution:
    """Time encoding solution."""
    
    def __init__(self, dim: int):
        self.dim = dim
        # Learnable frequencies
        self.omega = np.random.randn(dim // 2) * 0.1
    
    def forward(self, t: np.ndarray) -> np.ndarray:
        t = t.reshape(-1, 1)  # (batch, 1)
        
        # Fourier features
        angles = t * self.omega  # (batch, dim//2)
        
        cos_feats = np.cos(angles)
        sin_feats = np.sin(angles)
        
        return np.concatenate([cos_feats, sin_feats], axis=1)

class TemporalGNNLayerSolution:
    """Temporal GNN layer solution."""
    
    def __init__(self, node_dim: int, time_dim: int, out_dim: int):
        self.time_encoder = TimeEncodeSolution(time_dim)
        self.decay_rate = 0.1  # Temporal decay factor
        
        # Combined transformation
        scale = np.sqrt(2.0 / (node_dim + time_dim + out_dim))
        self.W = np.random.randn(node_dim + time_dim, out_dim) * scale
        self.W_self = np.random.randn(node_dim, out_dim) * scale
    
    def forward(self, H: np.ndarray, edges: np.ndarray,
               timestamps: np.ndarray, current_time: float) -> np.ndarray:
        n = len(H)
        
        # Encode time differences
        time_diffs = current_time - timestamps
        time_feats = self.time_encoder.forward(time_diffs)
        
        # Temporal decay weights
        decay_weights = np.exp(-self.decay_rate * time_diffs)
        
        # Self transformation
        H_out = H @ self.W_self
        
        # Aggregate neighbor messages with temporal weighting
        for i, (src, dst) in enumerate(edges):
            # Concatenate source features with time encoding
            msg_input = np.concatenate([H[src], time_feats[i]])
            msg = msg_input @ self.W
            
            # Weight by temporal decay
            H_out[dst] += decay_weights[i] * msg
        
        return np.maximum(0, H_out)

# Test solution
np.random.seed(42)

time_enc = TimeEncodeSolution(dim=16)
t = np.array([0.0, 1.0, 2.0, 10.0])
encoded = time_enc.forward(t)
print(f"Time encoding shape: {encoded.shape}")
print(f"Time encoding for t=0: {np.round(encoded[0], 3)}")

# Test temporal GNN
n = 5
H = np.random.randn(n, 8)
edges = np.array([[0, 1], [1, 2], [2, 3], [3, 4], [0, 4]])
timestamps = np.array([0.0, 1.0, 2.0, 5.0, 8.0])

temp_gnn = TemporalGNNLayerSolution(node_dim=8, time_dim=16, out_dim=8)
H_out = temp_gnn.forward(H, edges, timestamps, current_time=10.0)
print(f"\nTemporal GNN output shape: {H_out.shape}")
```

### Solution 10

```python
class SimpleGCN:
    """Simple 2-layer GCN for training demo."""
    
    def __init__(self, in_dim, hidden_dim, out_dim):
        scale1 = np.sqrt(2.0 / (in_dim + hidden_dim))
        scale2 = np.sqrt(2.0 / (hidden_dim + out_dim))
        
        self.W1 = np.random.randn(in_dim, hidden_dim) * scale1
        self.W2 = np.random.randn(hidden_dim, out_dim) * scale2
        self.cache = {}
    
    def forward(self, A_hat, X):
        H1 = np.maximum(0, A_hat @ X @ self.W1)  # ReLU
        logits = A_hat @ H1 @ self.W2
        
        # Softmax
        exp_logits = np.exp(logits - logits.max(axis=1, keepdims=True))
        probs = exp_logits / (exp_logits.sum(axis=1, keepdims=True) + 1e-10)
        
        self.cache = {'A_hat': A_hat, 'X': X, 'H1': H1, 'logits': logits, 'probs': probs}
        return probs
    
    def backward(self, labels, mask):
        A_hat = self.cache['A_hat']
        X = self.cache['X']
        H1 = self.cache['H1']
        probs = self.cache['probs']
        
        n = len(X)
        num_classes = probs.shape[1]
        
        # Gradient of cross-entropy + softmax
        grad_logits = probs.copy()
        for i in range(n):
            if mask[i]:
                grad_logits[i, labels[i]] -= 1
            else:
                grad_logits[i] = 0
        grad_logits /= mask.sum()
        
        # Backprop through layer 2
        grad_W2 = (A_hat @ H1).T @ grad_logits
        grad_H1 = A_hat.T @ grad_logits @ self.W2.T
        
        # Backprop through ReLU
        grad_H1 = grad_H1 * (H1 > 0)
        
        # Backprop through layer 1
        grad_W1 = (A_hat @ X).T @ grad_H1
        
        return grad_W1, grad_W2

class GNNTrainerSolution:
    """Training loop solution."""
    
    def __init__(self, model, learning_rate=0.01, weight_decay=5e-4):
        self.model = model
        self.lr = learning_rate
        self.wd = weight_decay
    
    def compute_loss(self, probs, labels, mask):
        n_masked = mask.sum()
        if n_masked == 0:
            return 0.0
        
        log_probs = np.log(probs[mask] + 1e-10)
        loss = -log_probs[np.arange(n_masked), labels[mask]].mean()
        return loss
    
    def train_epoch(self, A_hat, X, labels, train_mask):
        # Forward
        probs = self.model.forward(A_hat, X)
        loss = self.compute_loss(probs, labels, train_mask)
        
        # Backward
        grad_W1, grad_W2 = self.model.backward(labels, train_mask)
        
        # Update with weight decay
        self.model.W1 -= self.lr * (grad_W1 + self.wd * self.model.W1)
        self.model.W2 -= self.lr * (grad_W2 + self.wd * self.model.W2)
        
        return loss
    
    def evaluate(self, A_hat, X, labels, mask):
        probs = self.model.forward(A_hat, X)
        preds = probs[mask].argmax(axis=1)
        return (preds == labels[mask]).mean()

# Test training
np.random.seed(42)

# Create synthetic dataset
n = 50
num_classes = 3
feature_dim = 8

# Community graph
A = np.zeros((n, n))
labels = np.repeat(np.arange(num_classes), n // num_classes + 1)[:n]

for i in range(n):
    for j in range(i+1, n):
        if labels[i] == labels[j]:
            if np.random.rand() < 0.3:
                A[i, j] = A[j, i] = 1
        else:
            if np.random.rand() < 0.05:
                A[i, j] = A[j, i] = 1

# Features
X = np.random.randn(n, feature_dim)
for i, c in enumerate(labels):
    X[i, :2] += c * 0.5

# Normalize adjacency
A_tilde = A + np.eye(n)
D_inv_sqrt = np.diag(1.0 / np.sqrt(A_tilde.sum(axis=1) + 1e-10))
A_hat = D_inv_sqrt @ A_tilde @ D_inv_sqrt

# Masks
train_mask = np.zeros(n, dtype=bool)
train_mask[:int(n*0.6)] = True
test_mask = ~train_mask

# Train
model = SimpleGCN(in_dim=feature_dim, hidden_dim=16, out_dim=num_classes)
trainer = GNNTrainerSolution(model, learning_rate=0.1, weight_decay=1e-4)

print("Training GCN for node classification:")
print("Epoch | Train Loss | Train Acc | Test Acc")
print("-" * 50)

for epoch in range(100):
    loss = trainer.train_epoch(A_hat, X, labels, train_mask)
    
    if epoch % 20 == 0 or epoch == 99:
        train_acc = trainer.evaluate(A_hat, X, labels, train_mask)
        test_acc = trainer.evaluate(A_hat, X, labels, test_mask)
        print(f"{epoch:5d} | {loss:10.4f} | {train_acc:9.2%} | {test_acc:8.2%}")

print("\nTraining complete!")
```
