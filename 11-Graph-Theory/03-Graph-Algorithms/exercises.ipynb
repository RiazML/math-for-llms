```python
def has_cycle_undirected_solution(adj_list: Dict[int, List[int]], 
                                   vertices: List[int]) -> bool:
    """Solution: Cycle detection in undirected graph."""
    visited = set()
    
    def dfs(u, parent):
        visited.add(u)
        for v in adj_list.get(u, []):
            if v not in visited:
                if dfs(v, u):
                    return True
            elif v != parent:  # Visited and not parent = cycle!
                return True
        return False
    
    for v in vertices:
        if v not in visited:
            if dfs(v, -1):
                return True
    return False


def has_cycle_directed_solution(adj_list: Dict[int, List[int]], 
                                 vertices: List[int]) -> bool:
    """Solution: Cycle detection in directed graph using 3-color DFS."""
    WHITE, GRAY, BLACK = 0, 1, 2
    color = {v: WHITE for v in vertices}
    
    def dfs(u):
        color[u] = GRAY  # Currently in recursion stack
        for v in adj_list.get(u, []):
            if color[v] == GRAY:  # Back edge = cycle
                return True
            if color[v] == WHITE and dfs(v):
                return True
        color[u] = BLACK  # Done processing
        return False
    
    for v in vertices:
        if color[v] == WHITE:
            if dfs(v):
                return True
    return False


# Test solutions
print(f"Undirected has cycle: {has_cycle_undirected_solution(adj_undirected, list(adj_undirected.keys()))}")
print(f"Directed DAG has cycle: {has_cycle_directed_solution(adj_dag, list(adj_dag.keys()))}")
print(f"Directed with cycle: {has_cycle_directed_solution(adj_cycle, list(adj_cycle.keys()))}")
```

### Solution

```python
def euclidean_heuristic_solution(positions: Dict[int, Tuple[float, float]],
                                  goal: int) -> Dict[int, float]:
    """Solution: Euclidean distance heuristic."""
    goal_pos = positions[goal]
    return {
        v: np.sqrt((pos[0] - goal_pos[0])**2 + (pos[1] - goal_pos[1])**2)
        for v, pos in positions.items()
    }


def astar_solution(adj_list: Dict[int, List[Tuple[int, float]]],
                   start: int,
                   goal: int,
                   heuristic: Dict[int, float]) -> Tuple[List[int], float]:
    """Solution: A* search algorithm."""
    # Priority queue: (f-value, g-value, vertex, path)
    pq = [(heuristic[start], 0, start, [start])]
    visited = set()
    
    while pq:
        f, g, u, path = heapq.heappop(pq)
        
        if u == goal:
            return path, g
        
        if u in visited:
            continue
        visited.add(u)
        
        for v, cost in adj_list.get(u, []):
            if v not in visited:
                new_g = g + cost
                new_f = new_g + heuristic[v]
                heapq.heappush(pq, (new_f, new_g, v, path + [v]))
    
    return [], float('inf')  # No path found


# Test solution
h = euclidean_heuristic_solution(positions, goal=5)
print(f"Heuristic values: {h}")

path, cost = astar_solution(astar_adj, start=0, goal=5, heuristic=h)
print(f"A* Path: {path}")
print(f"A* Cost: {cost}")
```

```python
def find_all_paths_solution(adj_list: Dict[int, List[int]],
                             start: int,
                             end: int,
                             max_length: Optional[int] = None) -> List[List[int]]:
    """Solution: Find all simple paths using DFS backtracking."""
    all_paths = []
    
    def dfs(current, path, visited):
        if max_length and len(path) > max_length:
            return
        
        if current == end:
            all_paths.append(path[:])
            return
        
        for neighbor in adj_list.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                path.append(neighbor)
                dfs(neighbor, path, visited)
                path.pop()  # Backtrack
                visited.remove(neighbor)
    
    dfs(start, [start], {start})
    return all_paths


def find_shortest_paths_solution(adj_list: Dict[int, List[int]],
                                  start: int,
                                  end: int) -> List[List[int]]:
    """Solution: Find all shortest paths using BFS."""
    if start == end:
        return [[start]]
    
    # BFS to find shortest distance first
    queue = deque([(start, [start])])
    visited_at_level = {start: 0}
    paths = []
    shortest_length = float('inf')
    
    while queue:
        current, path = queue.popleft()
        
        if len(path) > shortest_length:
            break
        
        for neighbor in adj_list.get(current, []):
            new_path = path + [neighbor]
            
            if neighbor == end:
                paths.append(new_path)
                shortest_length = len(new_path)
            elif neighbor not in visited_at_level or visited_at_level[neighbor] == len(new_path) - 1:
                visited_at_level[neighbor] = len(new_path) - 1
                queue.append((neighbor, new_path))
    
    return paths


# Test solutions
all_paths = find_all_paths_solution(paths_adj, 0, 3)
print(f"All paths from 0 to 3: {all_paths}")

shortest = find_shortest_paths_solution(paths_adj, 0, 3)
print(f"Shortest paths: {shortest}")
```

```python
def find_articulation_points_solution(adj_list: Dict[int, List[int]],
                                       vertices: List[int]) -> List[int]:
    """Solution: Find articulation points using Tarjan's algorithm."""
    disc = {}  # Discovery time
    low = {}   # Lowest reachable vertex
    parent = {}
    ap = set()
    time = [0]
    
    def dfs(u):
        children = 0
        disc[u] = low[u] = time[0]
        time[0] += 1
        
        for v in adj_list.get(u, []):
            if v not in disc:
                children += 1
                parent[v] = u
                dfs(v)
                low[u] = min(low[u], low[v])
                
                # u is articulation point if:
                # 1. u is root with 2+ children, OR
                # 2. u is not root and low[v] >= disc[u]
                if parent.get(u) is None and children > 1:
                    ap.add(u)
                if parent.get(u) is not None and low[v] >= disc[u]:
                    ap.add(u)
            elif v != parent.get(u):
                low[u] = min(low[u], disc[v])
    
    for v in vertices:
        if v not in disc:
            parent[v] = None
            dfs(v)
    
    return list(ap)


def find_bridges_solution(adj_list: Dict[int, List[int]],
                           vertices: List[int]) -> List[Tuple[int, int]]:
    """Solution: Find bridges using Tarjan's algorithm."""
    disc = {}
    low = {}
    parent = {}
    bridges = []
    time = [0]
    
    def dfs(u):
        disc[u] = low[u] = time[0]
        time[0] += 1
        
        for v in adj_list.get(u, []):
            if v not in disc:
                parent[v] = u
                dfs(v)
                low[u] = min(low[u], low[v])
                
                # Edge is bridge if low[v] > disc[u]
                if low[v] > disc[u]:
                    bridges.append((u, v))
            elif v != parent.get(u):
                low[u] = min(low[u], disc[v])
    
    for v in vertices:
        if v not in disc:
            parent[v] = None
            dfs(v)
    
    return bridges


# Test solutions
art_points = find_articulation_points_solution(ap_adj, list(ap_adj.keys()))
bridges = find_bridges_solution(ap_adj, list(ap_adj.keys()))
print(f"Articulation points: {art_points}")
print(f"Bridges: {bridges}")

# Visualize
G_ap = nx.Graph(ap_adj)
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_ap, seed=42)
colors = ['red' if n in art_points else 'lightblue' for n in G_ap.nodes()]
nx.draw(G_ap, pos, with_labels=True, node_color=colors, node_size=500, font_weight='bold')
plt.title("Articulation Points (red) and Bridges")
plt.show()
```

```python
def is_bipartite_solution(adj_list: Dict[int, List[int]], 
                           vertices: List[int]) -> Tuple[bool, Dict[int, int]]:
    """Solution: Bipartite check using BFS coloring."""
    color = {}
    
    for start in vertices:
        if start in color:
            continue
        
        # BFS coloring
        queue = deque([start])
        color[start] = 0
        
        while queue:
            u = queue.popleft()
            for v in adj_list.get(u, []):
                if v not in color:
                    color[v] = 1 - color[u]  # Alternate color
                    queue.append(v)
                elif color[v] == color[u]:  # Same color = not bipartite
                    return False, {}
    
    return True, color


# Test solutions
result, coloring = is_bipartite_solution(bipartite_adj, list(bipartite_adj.keys()))
print(f"Even cycle bipartite: {result}, coloring: {coloring}")

result, coloring = is_bipartite_solution(not_bipartite_adj, list(not_bipartite_adj.keys()))
print(f"Odd cycle (triangle) bipartite: {result}")

# Visualize bipartite graph
G_bi = nx.Graph(bipartite_adj)
_, coloring = is_bipartite_solution(bipartite_adj, list(bipartite_adj.keys()))
plt.figure(figsize=(6, 4))
pos = {0: (0, 1), 1: (1, 1), 2: (1, 0), 3: (0, 0)}
colors = ['lightcoral' if coloring[n] == 0 else 'lightblue' for n in G_bi.nodes()]
nx.draw(G_bi, pos, with_labels=True, node_color=colors, node_size=500, font_weight='bold')
plt.title("Bipartite Graph (2-colored)")
plt.show()
```

```python
def topological_sort_for_dag(adj_list: Dict[int, List[Tuple[int, float]]],
                              vertices: List[int]) -> List[int]:
    """Helper: Topological sort using Kahn's algorithm."""
    in_degree = {v: 0 for v in vertices}
    for u in adj_list:
        for v, _ in adj_list[u]:
            in_degree[v] += 1
    
    queue = deque([v for v in vertices if in_degree[v] == 0])
    result = []
    
    while queue:
        u = queue.popleft()
        result.append(u)
        for v, _ in adj_list.get(u, []):
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)
    
    return result


def shortest_path_dag_solution(adj_list: Dict[int, List[Tuple[int, float]]],
                                vertices: List[int],
                                source: int) -> Dict[int, float]:
    """Solution: Shortest path in DAG using topological sort."""
    # Get topological order
    topo_order = topological_sort_for_dag(adj_list, vertices)
    
    # Initialize distances
    distances = {v: float('inf') for v in vertices}
    distances[source] = 0
    
    # Process in topological order
    for u in topo_order:
        if distances[u] != float('inf'):
            for v, weight in adj_list.get(u, []):
                if distances[u] + weight < distances[v]:
                    distances[v] = distances[u] + weight
    
    return distances


# Test solution
distances = shortest_path_dag_solution(dag_adj, list(dag_adj.keys()), source=0)
print(f"Shortest distances from vertex 0:")
for v, d in sorted(distances.items()):
    print(f"  To {v}: {d}")
```

```python
def personalized_pagerank_solution(adj_list: Dict[int, List[int]],
                                    seed: int,
                                    alpha: float = 0.15,
                                    max_iter: int = 100,
                                    tol: float = 1e-6) -> Dict[int, float]:
    """Solution: Personalized PageRank using power iteration."""
    vertices = list(adj_list.keys())
    n = len(vertices)
    v_to_idx = {v: i for i, v in enumerate(vertices)}
    
    # Compute out-degrees
    out_degree = np.array([len(adj_list.get(v, [])) for v in vertices], dtype=float)
    out_degree[out_degree == 0] = 1  # Handle dangling nodes
    
    # Initialize PPR
    ppr = np.zeros(n)
    ppr[v_to_idx[seed]] = 1.0
    
    # Power iteration
    for _ in range(max_iter):
        ppr_new = np.zeros(n)
        ppr_new[v_to_idx[seed]] = alpha  # Teleport to seed
        
        # Propagate scores
        for u_idx, u in enumerate(vertices):
            for v in adj_list.get(u, []):
                v_idx = v_to_idx[v]
                ppr_new[v_idx] += (1 - alpha) * ppr[u_idx] / out_degree[u_idx]
        
        if np.abs(ppr_new - ppr).sum() < tol:
            break
        ppr = ppr_new
    
    return {v: ppr[i] for i, v in enumerate(vertices)}


# Test solution
ppr = personalized_pagerank_solution(ppr_adj, seed=0)
print(f"Personalized PageRank from node 0:")
for v, score in sorted(ppr.items(), key=lambda x: -x[1]):
    print(f"  Node {v}: {score:.4f}")

# Compare seeds
ppr_5 = personalized_pagerank_solution(ppr_adj, seed=5)
print(f"\nPPR from node 5:")
for v, score in sorted(ppr_5.items(), key=lambda x: -x[1]):
    print(f"  Node {v}: {score:.4f}")
```

```python
def compute_eccentricities_solution(adj_list: Dict[int, List[int]],
                                     vertices: List[int]) -> Dict[int, int]:
    """Solution: Compute eccentricity using BFS from each vertex."""
    def bfs_max_dist(start):
        distances = {start: 0}
        queue = deque([start])
        
        while queue:
            u = queue.popleft()
            for v in adj_list.get(u, []):
                if v not in distances:
                    distances[v] = distances[u] + 1
                    queue.append(v)
        
        return max(distances.values()) if distances else 0
    
    return {v: bfs_max_dist(v) for v in vertices}


def graph_diameter_solution(eccentricities: Dict[int, int]) -> int:
    """Solution: Graph diameter."""
    return max(eccentricities.values())


def graph_radius_solution(eccentricities: Dict[int, int]) -> int:
    """Solution: Graph radius."""
    return min(eccentricities.values())


def graph_center_solution(eccentricities: Dict[int, int]) -> List[int]:
    """Solution: Graph center vertices."""
    radius = min(eccentricities.values())
    return [v for v, e in eccentricities.items() if e == radius]


# Test solution
ecc = compute_eccentricities_solution(metrics_adj, list(metrics_adj.keys()))
print(f"Eccentricities: {ecc}")
print(f"Diameter: {graph_diameter_solution(ecc)}")
print(f"Radius: {graph_radius_solution(ecc)}")
print(f"Center: {graph_center_solution(ecc)}")

# Visualize
G_metrics = nx.Graph(metrics_adj)
center = graph_center_solution(ecc)
plt.figure(figsize=(8, 4))
pos = {i: (i, 0) for i in range(6)}
colors = ['gold' if n in center else 'lightblue' for n in G_metrics.nodes()]
nx.draw(G_metrics, pos, with_labels=True, node_color=colors, 
        node_size=500, font_weight='bold')
plt.title(f"Graph with Center (gold), Diameter={graph_diameter_solution(ecc)}")
plt.show()
```

### Solution

```python
class UnionFind:
    """Union-Find data structure for Kruskal's algorithm."""
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n
        self.components = n
    
    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]
    
    def union(self, x, y):
        px, py = self.find(x), self.find(y)
        if px == py:
            return False
        if self.rank[px] < self.rank[py]:
            px, py = py, px
        self.parent[py] = px
        if self.rank[px] == self.rank[py]:
            self.rank[px] += 1
        self.components -= 1
        return True


def min_cost_to_connect_solution(n: int, 
                                  edges: List[Tuple[int, int, int]]) -> int:
    """Solution: Minimum cost to connect using Kruskal's algorithm."""
    # Sort edges by cost
    sorted_edges = sorted(edges, key=lambda x: x[2])
    
    uf = UnionFind(n)
    total_cost = 0
    edges_used = 0
    
    for u, v, cost in sorted_edges:
        if uf.union(u, v):
            total_cost += cost
            edges_used += 1
            if edges_used == n - 1:
                break
    
    # Check if fully connected
    if uf.components > 1:
        return -1
    
    return total_cost


# Test solutions
cost = min_cost_to_connect_solution(4, edges_1)
print(f"Min cost to connect 4 nodes: {cost}")
print(f"Expected: 9 (edges: 1-2:1, 1-3:3, 0-1:5)")

cost = min_cost_to_connect_solution(4, edges_2)
print(f"\nDisconnected graph result: {cost}")
print("Expected: -1 (impossible to connect)")
```
