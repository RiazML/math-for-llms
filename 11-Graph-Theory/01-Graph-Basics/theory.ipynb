# Graph Theory Basics - Worked Examples

This notebook covers fundamental graph theory concepts essential for AI and machine learning:
- Vertices and edges
- Graph types (directed, undirected, weighted)
- Adjacency and degree
- Paths and cycles
- Connectivity and trees
- Bipartite graphs
- Graph isomorphism and subgraphs
- Special graph properties

We use the `networkx` library for graph operations.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

## Example 1: Vertices and Edges

A **graph** $G = (V, E)$ consists of:
- $V$: a set of **vertices** (or nodes)
- $E$: a set of **edges** connecting pairs of vertices

Let's create a simple graph and explore its structure.

```python
# Create an undirected graph
G = nx.Graph()

# Add vertices (nodes)
G.add_nodes_from([1, 2, 3, 4, 5])

# Add edges
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

print(f"Vertices V = {set(G.nodes())}")
print(f"Number of vertices |V| = {G.number_of_nodes()}")
print(f"\nEdges E = {set(G.edges())}")
print(f"Number of edges |E| = {G.number_of_edges()}")

# Visualize
plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='lightblue', 
        node_size=700, font_size=16, font_weight='bold')
plt.title("Simple Undirected Graph")
plt.show()
```

## Example 2: Directed Graphs

A **directed graph** (digraph) has edges with direction. Each edge $(u, v)$ goes *from* $u$ *to* $v$.

Key concepts:
- **In-degree**: number of edges coming into a vertex
- **Out-degree**: number of edges going out of a vertex
- **Source**: vertex with in-degree 0
- **Sink**: vertex with out-degree 0

```python
# Create a directed graph
D = nx.DiGraph()

# Add directed edges
D.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'B')])

print("Directed Graph Analysis")
print("=" * 40)
print(f"{'Vertex':<10} {'In-degree':<12} {'Out-degree'}")
print("-" * 35)

for v in sorted(D.nodes()):
    print(f"{v:<10} {D.in_degree(v):<12} {D.out_degree(v)}")

# Find sources and sinks
sources = [v for v in D.nodes() if D.in_degree(v) == 0]
sinks = [v for v in D.nodes() if D.out_degree(v) == 0]

print(f"\nSources (in-degree 0): {sources}")
print(f"Sinks (out-degree 0): {sinks}")

# Visualize
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(D, seed=42)
nx.draw(D, pos, with_labels=True, node_color='lightcoral', 
        node_size=700, font_size=16, arrows=True, arrowsize=20)
plt.title("Directed Graph")
plt.show()
```

## Example 3: Weighted Graphs

A **weighted graph** assigns a numerical value (weight) to each edge. Common applications:
- Road networks (distance/time)
- Neural networks (connection strengths)
- Social networks (interaction frequency)

```python
# Create weighted graph
W = nx.Graph()

# Add weighted edges (node1, node2, weight)
weighted_edges = [
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'C', 1),
    ('B', 'D', 5), ('C', 'D', 8), ('C', 'E', 10),
    ('D', 'E', 2)
]

W.add_weighted_edges_from(weighted_edges)

print("Weighted Graph Edges:")
for u, v, data in W.edges(data=True):
    print(f"  {u} -- {v}: weight = {data['weight']}")

# Compute total weight
total_weight = sum(d['weight'] for u, v, d in W.edges(data=True))
print(f"\nTotal edge weight: {total_weight}")

# Visualize
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(W, seed=42)
nx.draw(W, pos, with_labels=True, node_color='lightgreen',
        node_size=700, font_size=14)
edge_labels = nx.get_edge_attributes(W, 'weight')
nx.draw_networkx_edge_labels(W, pos, edge_labels=edge_labels, font_size=12)
plt.title("Weighted Graph")
plt.show()
```

## Example 4: Adjacency and Degree

Two vertices are **adjacent** if connected by an edge. The **degree** of a vertex is the number of edges incident to it.

**Handshaking Lemma**: The sum of all degrees equals twice the number of edges:
$$\sum_{v \in V} \deg(v) = 2|E|$$

```python
# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (3, 4), (4, 5)])

print("Adjacency Analysis")
print("=" * 40)

# Check adjacency
print(f"Is 1 adjacent to 2? {G.has_edge(1, 2)}")
print(f"Is 2 adjacent to 5? {G.has_edge(2, 5)}")

# Neighbors (adjacent vertices)
print(f"\nNeighbors of vertex 1: {list(G.neighbors(1))}")
print(f"Neighbors of vertex 3: {list(G.neighbors(3))}")

# Degree of each vertex
print(f"\nDegree of each vertex:")
for v in sorted(G.nodes()):
    print(f"  deg({v}) = {G.degree(v)}")

# Verify handshaking lemma
sum_degrees = sum(dict(G.degree()).values())
print(f"\nSum of degrees: {sum_degrees}")
print(f"2 × |E| = 2 × {G.number_of_edges()} = {2 * G.number_of_edges()}")
print(f"Handshaking lemma verified: {sum_degrees == 2 * G.number_of_edges()}")
```

## Example 5: Paths

A **path** is a sequence of vertices where consecutive vertices are adjacent. 

- **Simple path**: no repeated vertices
- **Path length**: number of edges in the path
- **Shortest path**: path with minimum length (or minimum total weight)

```python
# Create a graph
G = nx.Graph()
G.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'),
    ('C', 'F'), ('D', 'G'), ('E', 'G'), ('F', 'G')
])

print("Path Analysis")
print("=" * 40)

# Check if path exists
print(f"Path exists A → G? {nx.has_path(G, 'A', 'G')}")

# Find shortest path
shortest = nx.shortest_path(G, 'A', 'G')
print(f"\nShortest path A → G: {' → '.join(shortest)}")
print(f"Path length: {len(shortest) - 1} edges")

# All simple paths
all_paths = list(nx.all_simple_paths(G, 'A', 'G'))
print(f"\nAll simple paths A → G ({len(all_paths)} total):")
for path in all_paths:
    print(f"  {' → '.join(path)} (length {len(path)-1})")

# Visualize with path highlighted
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700)
path_edges = list(zip(shortest[:-1], shortest[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)
plt.title("Graph with Shortest Path Highlighted")
plt.show()
```

## Example 6: Cycles

A **cycle** is a path that starts and ends at the same vertex (with at least 3 vertices).

- **Acyclic graph**: contains no cycles
- **DAG**: Directed Acyclic Graph (important for neural network computation graphs)

```python
# Graph with cycles
G = nx.Graph()
G.add_edges_from([
    (1, 2), (2, 3), (3, 1),  # Triangle
    (3, 4), (4, 5), (5, 3)   # Another triangle sharing vertex 3
])

print("Cycle Analysis - Undirected Graph")
print("=" * 40)

# Find cycle basis
cycles = nx.cycle_basis(G)
print(f"Cycle basis (fundamental cycles): {cycles}")

# Number of independent cycles
# For connected graph: |E| - |V| + 1
cyclomatic = G.number_of_edges() - G.number_of_nodes() + 1
print(f"\nCyclomatic number: {cyclomatic}")

# Directed Acyclic Graph (DAG)
print("\n" + "=" * 40)
print("DAG Analysis")

DAG = nx.DiGraph()
DAG.add_edges_from([('Input', 'Hidden1'), ('Input', 'Hidden2'),
                    ('Hidden1', 'Output'), ('Hidden2', 'Output')])

print(f"Is DAG? {nx.is_directed_acyclic_graph(DAG)}")
print(f"Topological order: {list(nx.topological_sort(DAG))}")

# Add cycle and check
DAG_with_cycle = DAG.copy()
DAG_with_cycle.add_edge('Output', 'Input')  # Creates cycle
print(f"\nAfter adding Output→Input:")
print(f"Is DAG? {nx.is_directed_acyclic_graph(DAG_with_cycle)}")
```

## Example 7: Connectivity

A graph is **connected** if there's a path between every pair of vertices.

**Connected components**: maximal connected subgraphs.

For directed graphs:
- **Strongly connected**: path exists in both directions between all pairs
- **Weakly connected**: connected when ignoring edge directions

```python
# Graph with multiple components
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3)])  # Component 1
G.add_edges_from([(4, 5), (5, 6), (6, 4)])  # Component 2
G.add_node(7)  # Component 3 (isolated)

print("Connectivity Analysis - Undirected Graph")
print("=" * 40)
print(f"Is connected? {nx.is_connected(G)}")
print(f"Number of components: {nx.number_connected_components(G)}")

# List components
components = list(nx.connected_components(G))
for i, comp in enumerate(components, 1):
    print(f"  Component {i}: {comp}")

# Strongly connected components in digraph
print("\n" + "=" * 40)
print("Strongly Connected Components - Directed Graph")

D = nx.DiGraph()
D.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A'),  # SCC 1
                  ('C', 'D'), ('D', 'E'), ('E', 'D')])  # SCC 2

print(f"Strongly connected? {nx.is_strongly_connected(D)}")
print(f"Weakly connected? {nx.is_weakly_connected(D)}")

sccs = list(nx.strongly_connected_components(D))
print(f"\nStrongly connected components:")
for i, scc in enumerate(sccs, 1):
    print(f"  SCC {i}: {scc}")
```

## Example 8: Trees

A **tree** is a connected acyclic graph. Properties:
- $|E| = |V| - 1$
- Exactly one path between any two vertices
- Removing any edge disconnects the graph
- Adding any edge creates a cycle

Special trees:
- **Rooted tree**: one vertex designated as root
- **Binary tree**: each node has at most 2 children
- **Spanning tree**: tree containing all vertices of a graph

```python
# Create a tree
T = nx.Graph()
T.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'),
    ('C', 'F'), ('C', 'G')
])

print("Tree Properties")
print("=" * 40)
print(f"Is tree? {nx.is_tree(T)}")
print(f"Vertices: {T.number_of_nodes()}")
print(f"Edges: {T.number_of_edges()}")
print(f"|E| = |V| - 1? {T.number_of_edges() == T.number_of_nodes() - 1}")

# Spanning tree of a graph
print("\n" + "=" * 40)
print("Minimum Spanning Tree")

G = nx.Graph()
G.add_weighted_edges_from([
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'C', 1),
    ('B', 'D', 5), ('C', 'D', 3)
])

mst = nx.minimum_spanning_tree(G)
print(f"Original edges: {G.number_of_edges()}")
print(f"MST edges: {list(mst.edges(data=True))}")
mst_weight = sum(d['weight'] for u, v, d in mst.edges(data=True))
print(f"MST total weight: {mst_weight}")

# Visualize tree
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(T, seed=42)
nx.draw(T, pos, with_labels=True, node_color='lightyellow',
        node_size=700, font_size=14)
plt.title("Tree Structure")
plt.show()
```

## Example 9: Bipartite Graphs

A graph is **bipartite** if vertices can be divided into two disjoint sets such that every edge connects vertices from different sets.

**Theorem**: A graph is bipartite iff it contains no odd-length cycles.

Applications: matching problems, recommendation systems, user-item interactions.

```python
# Create a bipartite graph (users and items)
B = nx.Graph()

# Two sets of nodes
users = ['User1', 'User2', 'User3']
items = ['Item_A', 'Item_B', 'Item_C', 'Item_D']

# Edges only between users and items
B.add_edges_from([
    ('User1', 'Item_A'), ('User1', 'Item_B'),
    ('User2', 'Item_B'), ('User2', 'Item_C'),
    ('User3', 'Item_C'), ('User3', 'Item_D')
])

print("Bipartite Graph Analysis")
print("=" * 40)
print(f"Is bipartite? {nx.is_bipartite(B)}")

# Get the two partitions
if nx.is_bipartite(B):
    left, right = nx.bipartite.sets(B)
    print(f"\nPartition 1: {left}")
    print(f"Partition 2: {right}")

# Non-bipartite example (triangle)
T = nx.Graph()
T.add_edges_from([(1, 2), (2, 3), (3, 1)])
print(f"\nTriangle is bipartite? {nx.is_bipartite(T)}")
print("(Contains odd cycle of length 3)")

# Visualize bipartite graph
plt.figure(figsize=(10, 6))
pos = {}
for i, u in enumerate(users):
    pos[u] = (0, i)
for i, item in enumerate(items):
    pos[item] = (1, i)

colors = ['lightblue' if n in users else 'lightcoral' for n in B.nodes()]
nx.draw(B, pos, with_labels=True, node_color=colors, node_size=1500, font_size=10)
plt.title("Bipartite Graph (User-Item Interactions)")
plt.show()
```

## Example 10: Graph Isomorphism

Two graphs $G$ and $H$ are **isomorphic** if there's a bijection between their vertices that preserves adjacency.

Isomorphic graphs have identical:
- Number of vertices and edges
- Degree sequence
- Number of cycles of each length
- Connectivity properties

```python
# Two isomorphic graphs (same structure, different labels)
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

G2 = nx.Graph()
G2.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')])

print("Graph Isomorphism")
print("=" * 40)

# Check isomorphism
print(f"G1 and G2 isomorphic? {nx.is_isomorphic(G1, G2)}")

# Properties that must match
print(f"\nG1: {G1.number_of_nodes()} nodes, {G1.number_of_edges()} edges")
print(f"G2: {G2.number_of_nodes()} nodes, {G2.number_of_edges()} edges")

print(f"\nDegree sequence G1: {sorted([d for n, d in G1.degree()], reverse=True)}")
print(f"Degree sequence G2: {sorted([d for n, d in G2.degree()], reverse=True)}")

# Non-isomorphic example
G3 = nx.Graph()
G3.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])  # Add diagonal

print(f"\nG1 and G3 isomorphic? {nx.is_isomorphic(G1, G3)}")
print(f"G1 edges: {G1.number_of_edges()}, G3 edges: {G3.number_of_edges()}")

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
nx.draw(G1, ax=axes[0], with_labels=True, node_color='lightblue', node_size=700)
axes[0].set_title("G1 (Square)")
nx.draw(G2, ax=axes[1], with_labels=True, node_color='lightgreen', node_size=700)
axes[1].set_title("G2 (Isomorphic to G1)")
plt.suptitle("Isomorphic Graphs")
plt.show()
```

## Example 11: Subgraphs

A **subgraph** $H$ of $G$ is a graph where $V(H) \subseteq V(G)$ and $E(H) \subseteq E(G)$.

Types:
- **Induced subgraph**: includes all edges between selected vertices
- **Spanning subgraph**: includes all vertices, subset of edges

```python
# Original graph
G = nx.Graph()
G.add_edges_from([
    (1, 2), (1, 3), (1, 4), (2, 3), (2, 4),
    (3, 4), (3, 5), (4, 5)
])

print("Subgraph Analysis")
print("=" * 40)
print(f"Original G: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

# Induced subgraph (vertices 1, 2, 3)
H_induced = G.subgraph([1, 2, 3]).copy()
print(f"\nInduced subgraph on {{1, 2, 3}}:")
print(f"  Nodes: {list(H_induced.nodes())}")
print(f"  Edges: {list(H_induced.edges())}")

# Edge subgraph
selected_edges = [(1, 2), (2, 3), (3, 4), (4, 5)]
H_edge = G.edge_subgraph(selected_edges).copy()
print(f"\nEdge subgraph (path):")
print(f"  Nodes: {list(H_edge.nodes())}")
print(f"  Edges: {list(H_edge.edges())}")

# Check if K3 (triangle) is subgraph
K3 = nx.complete_graph(3)
GM = nx.algorithms.isomorphism.GraphMatcher(G, K3)
print(f"\nDoes G contain K3 (triangle)? {GM.subgraph_is_isomorphic()}")

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, ax=axes[0], with_labels=True, node_color='lightblue', node_size=700)
axes[0].set_title("Original Graph G")

pos_sub = {k: pos[k] for k in [1, 2, 3]}
nx.draw(H_induced, pos_sub, ax=axes[1], with_labels=True, node_color='lightgreen', node_size=700)
axes[1].set_title("Induced Subgraph {1,2,3}")

nx.draw(H_edge, pos, ax=axes[2], with_labels=True, node_color='lightyellow', node_size=700)
axes[2].set_title("Edge Subgraph (Path)")
plt.show()
```

## Example 12: Special Graph Properties

### Complete Graph $K_n$
Every pair of vertices is connected. Has $\binom{n}{2} = \frac{n(n-1)}{2}$ edges.

### Regular Graph
Every vertex has the same degree.

### Planar Graph
Can be drawn in a plane without edge crossings. By Euler's formula: $V - E + F = 2$.

```python
print("Special Graph Properties")
print("=" * 40)

# Complete graph K5
K5 = nx.complete_graph(5)
print("Complete Graph K5:")
print(f"  Vertices: {K5.number_of_nodes()}")
print(f"  Edges: {K5.number_of_edges()}")
print(f"  Expected edges n(n-1)/2: {5*4//2}")
print(f"  Is regular? All degrees = {K5.degree(0)}: {all(d == 4 for n, d in K5.degree())}")

# Regular graph (cycle)
print("\nCycle Graph C6 (2-regular):")
C6 = nx.cycle_graph(6)
degrees = [d for n, d in C6.degree()]
print(f"  Degrees: {degrees}")
print(f"  Is 2-regular: {all(d == 2 for d in degrees)}")

# Planar graph check
print("\nPlanarity:")
K4 = nx.complete_graph(4)
is_planar_K4, _ = nx.check_planarity(K4)
print(f"  K4 is planar: {is_planar_K4}")

is_planar_K5, _ = nx.check_planarity(K5)
print(f"  K5 is planar: {is_planar_K5}")

# Euler's formula for K4
print(f"\nEuler's formula for K4:")
V, E = K4.number_of_nodes(), K4.number_of_edges()
F = 2 - V + E  # Faces
print(f"  V = {V}, E = {E}, F = {F}")
print(f"  V - E + F = {V - E + F}")

# Graph density
print("\nGraph Density (ratio of actual to max possible edges):")
for name, graph in [('K5', K5), ('C6', C6)]:
    density = nx.density(graph)
    print(f"  {name}: {density:.3f}")

# Visualize special graphs
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
nx.draw(K5, ax=axes[0], with_labels=True, node_color='lightcoral', node_size=700)
axes[0].set_title("Complete Graph K5")

nx.draw_circular(C6, ax=axes[1], with_labels=True, node_color='lightgreen', node_size=700)
axes[1].set_title("Cycle Graph C6 (2-regular)")

petersen = nx.petersen_graph()
nx.draw(petersen, ax=axes[2], with_labels=True, node_color='lightyellow', node_size=500)
axes[2].set_title("Petersen Graph (3-regular)")
plt.show()
```
