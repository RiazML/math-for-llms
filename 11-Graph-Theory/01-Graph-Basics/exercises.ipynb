### Solution 1

```python
# Solution 1: Graph Implementation

class Graph:
    def __init__(self, directed=False):
        self.directed = directed
        self.adj = defaultdict(set)
    
    def add_vertex(self, v):
        """Add a vertex (creates empty adjacency list if not exists)"""
        if v not in self.adj:
            self.adj[v] = set()
    
    def add_edge(self, u, v):
        """Add edge between u and v (adds vertices if needed)"""
        self.add_vertex(u)
        self.add_vertex(v)
        self.adj[u].add(v)
        if not self.directed:
            self.adj[v].add(u)
    
    def remove_edge(self, u, v):
        """Remove edge between u and v"""
        self.adj[u].discard(v)
        if not self.directed:
            self.adj[v].discard(u)
    
    def has_edge(self, u, v):
        """Check if edge (u, v) exists"""
        return v in self.adj.get(u, set())
    
    def neighbors(self, v):
        """Return set of neighbors of v"""
        return self.adj.get(v, set())
    
    def degree(self, v):
        """Return degree of v"""
        return len(self.adj.get(v, set()))
    
    def vertices(self):
        """Return all vertices"""
        return set(self.adj.keys())
    
    def edges(self):
        """Return all edges (avoid duplicates for undirected)"""
        seen = set()
        result = []
        for u in self.adj:
            for v in self.adj[u]:
                if self.directed or (v, u) not in seen:
                    result.append((u, v))
                    seen.add((u, v))
        return result

# Test
g = Graph(directed=False)
for u, v in [(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)]:
    g.add_edge(u, v)

print(f"Vertices: {g.vertices()}")
print(f"Edges: {g.edges()}")
print(f"Neighbors of 1: {g.neighbors(1)}")
print(f"Degree of 1: {g.degree(1)}")
print(f"Has edge (1,2): {g.has_edge(1, 2)}")
print(f"Has edge (2,4): {g.has_edge(2, 4)}")
```

```python
# Solution 2: Adjacency Matrix Operations

def list_to_matrix(adj_list, vertices):
    """Convert adjacency list to adjacency matrix."""
    n = len(vertices)
    vertex_idx = {v: i for i, v in enumerate(vertices)}
    
    A = np.zeros((n, n), dtype=int)
    for u in adj_list:
        for v in adj_list[u]:
            A[vertex_idx[u], vertex_idx[v]] = 1
    return A

def count_paths_length_k(A, k):
    """Count paths of length k using matrix power."""
    return np.linalg.matrix_power(A, k)

def count_triangles(A):
    """Count triangles using trace(A^3) / 6."""
    A3 = np.linalg.matrix_power(A, 3)
    # Divide by 6: each triangle counted 3 times (starting vertices) × 2 (directions)
    return np.trace(A3) // 6

# Test
adj_list = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D'],
    'D': ['B', 'C']
}
vertices = ['A', 'B', 'C', 'D']

A = list_to_matrix(adj_list, vertices)
print("Adjacency Matrix:")
print(A)

A2 = count_paths_length_k(A, 2)
print(f"\nPaths of length 2 (A^2):")
print(A2)
print(f"Paths from A to D (length 2): {A2[0, 3]}")

triangles = count_triangles(A)
print(f"\nNumber of triangles: {triangles}")
```

### Solution 3

```python
# Solution 3: Graph Traversal

def bfs(graph, start):
    """BFS returning distances and parents."""
    distances = {start: 0}
    parents = {start: None}
    queue = deque([start])
    
    while queue:
        v = queue.popleft()
        for neighbor in graph.get(v, []):
            if neighbor not in distances:
                distances[neighbor] = distances[v] + 1
                parents[neighbor] = v
                queue.append(neighbor)
    
    return distances, parents

def dfs_with_times(graph, start):
    """DFS with discovery and finish times."""
    discovery = {}
    finish = {}
    time = [0]  # Use list for mutability in nested function
    
    def dfs_visit(v):
        time[0] += 1
        discovery[v] = time[0]
        
        for neighbor in graph.get(v, []):
            if neighbor not in discovery:
                dfs_visit(neighbor)
        
        time[0] += 1
        finish[v] = time[0]
    
    dfs_visit(start)
    return discovery, finish

def reconstruct_path(parents, target):
    """Reconstruct path using parent pointers."""
    path = []
    current = target
    while current is not None:
        path.append(current)
        current = parents.get(current)
    return path[::-1]

# Test
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

distances, parents = bfs(graph, 'A')
print("BFS from A:")
print(f"  Distances: {distances}")
print(f"  Path A → F: {reconstruct_path(parents, 'F')}")

discovery, finish = dfs_with_times(graph, 'A')
print(f"\nDFS from A:")
print(f"  Discovery times: {discovery}")
print(f"  Finish times: {finish}")
```

```python
# Solution 4: Cycle Detection

def has_cycle_undirected(graph):
    """Detect cycle in undirected graph using DFS."""
    visited = set()
    
    def dfs(v, parent):
        visited.add(v)
        for neighbor in graph.get(v, []):
            if neighbor not in visited:
                if dfs(neighbor, v):
                    return True
            elif neighbor != parent:
                # Found back edge to non-parent = cycle
                return True
        return False
    
    # Check all components
    for start in graph:
        if start not in visited:
            if dfs(start, None):
                return True
    return False

def has_cycle_directed(graph):
    """Detect cycle in directed graph using coloring."""
    WHITE, GRAY, BLACK = 0, 1, 2
    color = {v: WHITE for v in graph}
    
    def dfs(v):
        color[v] = GRAY
        for neighbor in graph.get(v, []):
            if neighbor not in color:
                color[neighbor] = WHITE
            if color[neighbor] == GRAY:
                return True  # Back edge = cycle
            if color[neighbor] == WHITE:
                if dfs(neighbor):
                    return True
        color[v] = BLACK
        return False
    
    for v in graph:
        if color.get(v, WHITE) == WHITE:
            if dfs(v):
                return True
    return False

# Test
# Undirected with cycle
graph_cycle = {
    'A': ['B', 'D'],
    'B': ['A', 'C'],
    'C': ['B', 'D'],
    'D': ['A', 'C']
}
print(f"Undirected graph (A-B-C-D-A) has cycle: {has_cycle_undirected(graph_cycle)}")

# Tree (no cycle)
tree = {
    'A': ['B', 'C'],
    'B': ['A'],
    'C': ['A']
}
print(f"Tree has cycle: {has_cycle_undirected(tree)}")

# Directed with cycle
digraph_cycle = {
    'A': ['B'],
    'B': ['C'],
    'C': ['A']
}
print(f"Directed cycle A→B→C→A has cycle: {has_cycle_directed(digraph_cycle)}")

# DAG
dag = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': []
}
print(f"DAG has cycle: {has_cycle_directed(dag)}")
```

```python
# Solution 5: Dijkstra's Algorithm

import heapq

def dijkstra(graph, start):
    """Dijkstra's shortest path algorithm."""
    distances = {v: float('inf') for v in graph}
    distances[start] = 0
    parents = {start: None}
    
    # Priority queue: (distance, vertex)
    pq = [(0, start)]
    visited = set()
    
    while pq:
        d, u = heapq.heappop(pq)
        
        if u in visited:
            continue
        visited.add(u)
        
        for v, weight in graph.get(u, []):
            if v not in visited:
                new_dist = distances[u] + weight
                if new_dist < distances[v]:
                    distances[v] = new_dist
                    parents[v] = u
                    heapq.heappush(pq, (new_dist, v))
    
    return distances, parents

# Test
weighted_graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('A', 4), ('C', 1), ('D', 5)],
    'C': [('A', 2), ('B', 1), ('D', 8), ('E', 10)],
    'D': [('B', 5), ('C', 8), ('E', 2)],
    'E': [('C', 10), ('D', 2)]
}

distances, parents = dijkstra(weighted_graph, 'A')

print("Shortest distances from A:")
for v in sorted(distances):
    print(f"  d(A, {v}) = {distances[v]}")

# Path to E
path = reconstruct_path(parents, 'E')
print(f"\nShortest path A → E: {' → '.join(path)}")
```

```python
# Solution 6: Connected Components

def find_connected_components(graph):
    """Find all connected components using BFS."""
    visited = set()
    components = []
    
    for start in graph:
        if start not in visited:
            # BFS from this vertex
            component = set()
            queue = deque([start])
            visited.add(start)
            
            while queue:
                v = queue.popleft()
                component.add(v)
                
                for neighbor in graph.get(v, []):
                    if neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
            
            components.append(component)
    
    return components

# Test
graph = {
    'A': ['B'],
    'B': ['A', 'C'],
    'C': ['B'],
    'D': ['E'],
    'E': ['D'],
    'F': []  # Isolated
}

components = find_connected_components(graph)
print(f"Number of components: {len(components)}")
for i, comp in enumerate(components, 1):
    print(f"  Component {i}: {comp}")
```

### Solution 7

```python
# Solution 7: Bipartite Graph Verification

def is_bipartite(graph):
    """Check if graph is bipartite using BFS 2-coloring."""
    color = {}
    
    for start in graph:
        if start in color:
            continue
        
        # BFS and try to 2-color
        queue = deque([start])
        color[start] = 0
        
        while queue:
            v = queue.popleft()
            for neighbor in graph.get(v, []):
                if neighbor not in color:
                    color[neighbor] = 1 - color[v]
                    queue.append(neighbor)
                elif color[neighbor] == color[v]:
                    # Same color = not bipartite
                    return False, (set(), set())
    
    # Build partition
    set0 = {v for v in color if color[v] == 0}
    set1 = {v for v in color if color[v] == 1}
    return True, (set0, set1)

# Test
# Bipartite graph
bipartite = {
    1: [4, 5],
    2: [4, 5],
    3: [5, 6],
    4: [1, 2],
    5: [1, 2, 3],
    6: [3]
}

is_bip, partition = is_bipartite(bipartite)
print(f"Graph is bipartite: {is_bip}")
if is_bip:
    print(f"Partition: {partition}")

# Non-bipartite (triangle)
triangle = {
    'A': ['B', 'C'],
    'B': ['A', 'C'],
    'C': ['A', 'B']
}

is_bip, _ = is_bipartite(triangle)
print(f"\nTriangle is bipartite: {is_bip}")
```

### Solution 8

```python
# Solution 8: Tree Properties

def is_tree(graph):
    """Check if graph is a tree."""
    if not graph:
        return True
    
    # Count vertices and edges
    n = len(graph)
    m = sum(len(neighbors) for neighbors in graph.values()) // 2
    
    # Tree must have |E| = |V| - 1
    if m != n - 1:
        return False
    
    # Must be connected
    start = next(iter(graph))
    visited = set()
    queue = deque([start])
    visited.add(start)
    
    while queue:
        v = queue.popleft()
        for neighbor in graph.get(v, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    
    return len(visited) == n

def tree_diameter(graph):
    """Find diameter using two BFS."""
    if not graph:
        return 0
    
    def bfs_farthest(start):
        """Find farthest node from start."""
        distances = {start: 0}
        queue = deque([start])
        farthest = start
        max_dist = 0
        
        while queue:
            v = queue.popleft()
            for neighbor in graph.get(v, []):
                if neighbor not in distances:
                    distances[neighbor] = distances[v] + 1
                    queue.append(neighbor)
                    if distances[neighbor] > max_dist:
                        max_dist = distances[neighbor]
                        farthest = neighbor
        
        return farthest, max_dist
    
    # First BFS from any node
    start = next(iter(graph))
    farthest, _ = bfs_farthest(start)
    
    # Second BFS from farthest node
    _, diameter = bfs_farthest(farthest)
    
    return diameter

# Test
tree = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['E']
}

print(f"Is tree: {is_tree(tree)}")
print(f"Diameter: {tree_diameter(tree)}")

# Add edge to create cycle
not_tree = tree.copy()
not_tree['C'] = ['A', 'D']
not_tree['D'] = ['B', 'C']
print(f"\nWith added edge C-D, is tree: {is_tree(not_tree)}")
```

```python
# Solution 9: Graph Isomorphism Invariants

def compute_invariants(G):
    """Compute graph invariants."""
    invariants = {
        'num_vertices': G.number_of_nodes(),
        'num_edges': G.number_of_edges(),
        'degree_sequence': tuple(sorted([d for n, d in G.degree()], reverse=True)),
        'num_triangles': sum(nx.triangles(G).values()) // 3,
        'density': nx.density(G),
        'is_connected': nx.is_connected(G)
    }
    return invariants

def could_be_isomorphic(G1, G2):
    """Check if graphs could be isomorphic based on invariants."""
    inv1 = compute_invariants(G1)
    inv2 = compute_invariants(G2)
    
    for key in inv1:
        if inv1[key] != inv2[key]:
            print(f"Different {key}: {inv1[key]} vs {inv2[key]}")
            return False
    return True

# Test
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

G2 = nx.Graph()
G2.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')])

print("G1 (4-cycle) vs G2 (4-cycle):")
print(f"Invariants match: {could_be_isomorphic(G1, G2)}")
print(f"Actually isomorphic: {nx.is_isomorphic(G1, G2)}")

# Different graph
G3 = nx.Graph()
G3.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])  # With diagonal

print("\nG1 (4-cycle) vs G3 (4-cycle with diagonal):")
print(f"Invariants match: {could_be_isomorphic(G1, G3)}")
```

### Solution 10

```python
# Solution 10: Graph Metrics and Centrality

def degree_centrality(graph):
    """Compute normalized degree centrality."""
    n = len(graph)
    if n <= 1:
        return {v: 0.0 for v in graph}
    
    return {v: len(neighbors) / (n - 1) for v, neighbors in graph.items()}

def closeness_centrality(graph):
    """Compute closeness centrality using BFS."""
    n = len(graph)
    centrality = {}
    
    for start in graph:
        # BFS for distances
        distances = {start: 0}
        queue = deque([start])
        
        while queue:
            v = queue.popleft()
            for neighbor in graph.get(v, []):
                if neighbor not in distances:
                    distances[neighbor] = distances[v] + 1
                    queue.append(neighbor)
        
        total_dist = sum(distances.values())
        centrality[start] = (n - 1) / total_dist if total_dist > 0 else 0.0
    
    return centrality

def local_clustering(graph, v):
    """Compute local clustering coefficient."""
    neighbors = list(graph.get(v, []))
    k = len(neighbors)
    
    if k < 2:
        return 0.0
    
    # Count edges between neighbors
    edges = 0
    for i in range(len(neighbors)):
        for j in range(i + 1, len(neighbors)):
            if neighbors[j] in graph.get(neighbors[i], []):
                edges += 1
    
    max_edges = k * (k - 1) / 2
    return edges / max_edges

# Test
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C', 'E'],
    'E': ['D']
}

print("Degree Centrality:")
dc = degree_centrality(graph)
for v in sorted(dc):
    print(f"  {v}: {dc[v]:.3f}")

print("\nCloseness Centrality:")
cc = closeness_centrality(graph)
for v in sorted(cc):
    print(f"  {v}: {cc[v]:.3f}")

print("\nLocal Clustering Coefficient:")
for v in sorted(graph):
    lc = local_clustering(graph, v)
    print(f"  {v}: {lc:.3f}")

# Verify with NetworkX
G = nx.Graph(graph)
print("\n--- Verification with NetworkX ---")
print(f"Degree centrality matches: {all(abs(dc[v] - nx.degree_centrality(G)[v]) < 0.001 for v in graph)}")
print(f"Closeness centrality matches: {all(abs(cc[v] - nx.closeness_centrality(G)[v]) < 0.001 for v in graph)}")
```
