## Exercise 2: Compute Eigenvectors (Basic)

For $A = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}$:

a) Find eigenvalues using the characteristic equation

b) Find eigenvectors for each eigenvalue by solving $(A - \lambda I)\mathbf{v} = 0$

c) Verify $A\mathbf{v} = \lambda\mathbf{v}$ for each pair

````python
# Your work here
````

### Solution 2

````python
print("Exercise 2 Solution: Compute Eigenvectors\n")

A = np.array([[4, 2], [1, 3]])
print(f"A = \n{A}")

print("\na) Finding eigenvalues:")
print("   det(A - λI) = (4-λ)(3-λ) - 2 = λ² - 7λ + 10 = 0")
print("   (λ - 5)(λ - 2) = 0")
print("   λ₁ = 5, λ₂ = 2")

print("\nb) Finding eigenvectors:")

print("\n   For λ₁ = 5:")
print("   (A - 5I)v = 0")
print("   [[-1, 2], [1, -2]]v = 0")
print("   -v₁ + 2v₂ = 0 → v₁ = 2v₂")
print("   v₁ = [2, 1]ᵀ (or any scalar multiple)")

print("\n   For λ₂ = 2:")
print("   (A - 2I)v = 0")
print("   [[2, 2], [1, 1]]v = 0")
print("   2v₁ + 2v₂ = 0 → v₁ = -v₂")
print("   v₂ = [1, -1]ᵀ (or any scalar multiple)")

print("\nc) Verification:")
eigenvalues, eigenvectors = eig(A)
for i in range(2):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    Av = A @ v
    lam_v = lam * v
    print(f"\n   λ = {lam.real:.1f}:")
    print(f"   v = {v.real}")
    print(f"   Av = {Av.real}")
    print(f"   λv = {lam_v.real}")
    print(f"   Av = λv: {np.allclose(Av, lam_v)}")
````

## Exercise 4: Diagonalization (Intermediate)

For $A = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}$:

a) Find $P$ (eigenvector matrix) and $D$ (diagonal eigenvalue matrix) such that $A = PDP^{-1}$

b) Verify the decomposition

c) Use diagonalization to compute $A^4$ efficiently

````python
# Your work here
````

### Solution 4

````python
print("Exercise 4 Solution: Diagonalization\n")

A = np.array([[3, 1], [0, 2]])
print(f"A = \n{A}")

print("\na) Finding P and D:")
eigenvalues, eigenvectors = eig(A)
D = np.diag(eigenvalues)
P = eigenvectors

print(f"   Eigenvalues: λ₁ = {eigenvalues[0]:.1f}, λ₂ = {eigenvalues[1]:.1f}")
print(f"\n   D = diag(λ₁, λ₂) = \n{D.real}")
print(f"\n   P (eigenvectors as columns) = \n{P.real}")

print("\nb) Verification A = PDP⁻¹:")
P_inv = inv(P)
reconstructed = P @ D @ P_inv
print(f"   PDP⁻¹ = \n{reconstructed.real}")
print(f"   Equal to A: {np.allclose(A, reconstructed)}")

print("\nc) Computing A⁴ via diagonalization:")
D4 = np.diag(eigenvalues ** 4)
A4_diag = P @ D4 @ P_inv
A4_direct = matrix_power(A, 4)

print(f"   D⁴ = diag(3⁴, 2⁴) = diag(81, 16)")
print(f"   A⁴ = PD⁴P⁻¹ = \n{A4_diag.real}")
print(f"   Direct A⁴ = \n{A4_direct}")
print(f"   Match: {np.allclose(A4_diag, A4_direct)}")
````

## Exercise 5: Symmetric Matrix Properties (Intermediate)

For $A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$ (symmetric):

a) Verify eigenvalues are real

b) Find eigenvectors and verify they are orthogonal

c) Write the spectral decomposition: $A = \lambda_1 \mathbf{v}_1 \mathbf{v}_1^T + \lambda_2 \mathbf{v}_2 \mathbf{v}_2^T$

````python
# Your work here
````

### Solution 5

````python
print("Exercise 5 Solution: Symmetric Matrix\n")

A = np.array([[2, 1], [1, 2]])
print(f"Symmetric A = \n{A}")

eigenvalues, eigenvectors = eig(A)

print("\na) Eigenvalues are real:")
print(f"   λ₁ = {eigenvalues[0].real}")
print(f"   λ₂ = {eigenvalues[1].real}")
print(f"   Both real: {all(np.isreal(eigenvalues))}")

print("\nb) Eigenvector orthogonality:")
v1 = eigenvectors[:, 0].real
v2 = eigenvectors[:, 1].real
# Normalize
v1 = v1 / np.linalg.norm(v1)
v2 = v2 / np.linalg.norm(v2)

print(f"   v₁ = {v1}")
print(f"   v₂ = {v2}")
print(f"   v₁·v₂ = {np.dot(v1, v2):.6f}")
print(f"   Orthogonal: {np.isclose(np.dot(v1, v2), 0)}")
````

````python
print("c) Spectral decomposition:\n")
lam1, lam2 = eigenvalues.real
term1 = lam1 * np.outer(v1, v1)
term2 = lam2 * np.outer(v2, v2)

print(f"λ₁v₁v₁ᵀ = {lam1:.1f} × ")
print(f"{np.outer(v1, v1)}")
print(f"= \n{term1}")

print(f"\nλ₂v₂v₂ᵀ = {lam2:.1f} × ")
print(f"{np.outer(v2, v2)}")
print(f"= \n{term2}")

print(f"\nA = λ₁v₁v₁ᵀ + λ₂v₂v₂ᵀ = \n{term1 + term2}")
print(f"Matches A: {np.allclose(A, term1 + term2)}")
````

## Exercise 7: Power Method Implementation (Advanced)

Implement the power method to find the dominant eigenvalue and eigenvector of:

$$A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$$

Show convergence over 10 iterations.

**Algorithm:**
```
v₀ = random or fixed starting vector
for k iterations:
    w = A @ v
    λ estimate = Rayleigh quotient = (v^T A v) / (v^T v)
    v = w / ||w||  (normalize)
```

````python
# Your work here
````

### Solution 7

````python
print("Exercise 7 Solution: Power Method\n")

A = np.array([[2, 1], [1, 2]])
print(f"A = \n{A}")

# True values for comparison
true_vals, true_vecs = eig(A)
dom_idx = np.argmax(np.abs(true_vals))
true_lambda = true_vals[dom_idx].real
true_vec = true_vecs[:, dom_idx].real
true_vec = true_vec / np.linalg.norm(true_vec)

print(f"\nTrue dominant eigenvalue: {true_lambda}")
print(f"True eigenvector: {true_vec}")

# Power method implementation
v = np.array([1.0, 0.0])  # Starting vector

print(f"\nPower Method (starting with v = {v}):")
print(f"{'Iter':>4} {'λ estimate':>12} {'v':>24} {'Error':>12}")

for k in range(10):
    w = A @ v
    lambda_est = np.dot(w, v) / np.dot(v, v)  # Rayleigh quotient
    v = w / np.linalg.norm(w)  # Normalize
    error = abs(lambda_est - true_lambda)
    print(f"{k+1:4d} {lambda_est:12.8f} [{v[0]:10.6f}, {v[1]:10.6f}] {error:12.8f}")

print(f"\nFinal estimate: λ ≈ {lambda_est:.6f}, v ≈ {v}")
print(f"Converged!")
````

## Exercise 8: Markov Chain Stationary Distribution (Advanced)

A mouse can be in room A or room B:
- From A: 70% stay in A, 30% go to B
- From B: 40% go to A, 60% stay in B

a) Write the transition matrix $P$

b) Find the stationary distribution $\pi$ where $\pi P = \pi$

c) Verify $\pi$ sums to 1 and $\pi P = \pi$

**Hint:** The stationary distribution is the left eigenvector with eigenvalue 1.

````python
# Your work here
````

### Solution 8

````python
print("Exercise 8 Solution: Markov Chain\n")

print("a) Transition matrix P:")
P = np.array([[0.7, 0.3],   # From A
              [0.4, 0.6]])  # From B
print(f"   P = \n{P}")
print("   P[i,j] = P(go to j | currently in i)")

print("\nb) Finding stationary distribution:")
print("   πP = π means π is left eigenvector of P with λ=1")
print("   Equivalently, Pᵀπ = π, so π is eigenvector of Pᵀ with λ=1")

eigenvalues, eigenvectors = eig(P.T)
print(f"\n   Eigenvalues of Pᵀ: {eigenvalues}")

# Find λ = 1
idx = np.argmin(np.abs(eigenvalues - 1))
pi = eigenvectors[:, idx].real
pi = pi / np.sum(pi)  # Normalize to probability

print(f"\n   Stationary distribution π = {pi}")
print(f"   π(A) = {pi[0]:.4f}")
print(f"   π(B) = {pi[1]:.4f}")
````

````python
print("c) Verification:")
print(f"   Sum of π: {np.sum(pi):.4f}")
pi_P = pi @ P
print(f"   πP = {pi_P}")
print(f"   π = πP: {np.allclose(pi, pi_P)}")

# Alternative: solve linear system
print("\n   Alternative (solving πP = π directly):")
print("   π₁(0.7) + π₂(0.4) = π₁  →  -0.3π₁ + 0.4π₂ = 0")
print("   With π₁ + π₂ = 1:")
print("   0.4π₂ = 0.3π₁ = 0.3(1-π₂)  →  0.7π₂ = 0.3  →  π₂ = 3/7")
print(f"   π = [4/7, 3/7] = [{4/7:.4f}, {3/7:.4f}]")
````

## Exercise 9: Positive Definite Check (Advanced)

Determine if the following matrices are **positive definite** using the eigenvalue criterion (all eigenvalues > 0):

a) $A = \begin{bmatrix} 4 & 2 \\ 2 & 1 \end{bmatrix}$

b) $B = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}$

c) $C = \begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix}$

For non-positive-definite matrices, find an $\mathbf{x}$ such that $\mathbf{x}^T A \mathbf{x} \leq 0$.

````python
# Your work here
````

### Solution 9

````python
print("Exercise 9 Solution: Positive Definite Check\n")

matrices = {
    'A': np.array([[4, 2], [2, 1]]),
    'B': np.array([[3, 1], [1, 3]]),
    'C': np.array([[1, 2], [2, 1]])
}

for name, M in matrices.items():
    print(f"{name}) Matrix:\n{M}")
    eigenvalues, vecs = eig(M)
    eigenvalues = sorted(eigenvalues.real)
    print(f"   Eigenvalues: {eigenvalues}")
    is_pd = all(lam > 0 for lam in eigenvalues)
    print(f"   All positive: {is_pd}")
    print(f"   Positive definite: {is_pd}")
    
    if not is_pd:
        # Find eigenvector for non-positive eigenvalue
        eig_vals, eig_vecs = eig(M)
        idx = np.argmin(eig_vals.real)
        x = eig_vecs[:, idx].real
        quad = x.T @ M @ x
        print(f"   Counterexample: x = {x}")
        print(f"   xᵀMx = {quad:.4f}")
    print()
````

## Exercise 10: Spectral Norm and Condition Number (Advanced)

For $A = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}$:

a) Find singular values (= $\sqrt{\text{eigenvalues of } A^T A}$)

b) Compute spectral norm $\|A\|_2 = \sigma_{\max}$ (largest singular value)

c) Compute condition number $\kappa(A) = \sigma_{\max} / \sigma_{\min}$

**The condition number measures how sensitive solving $Ax = b$ is to perturbations.**

````python
# Your work here
````

### Solution 10

````python
print("Exercise 10 Solution: Spectral Norm\n")

A = np.array([[3, 1], [0, 2]])
print(f"A = \n{A}")

print("\na) Finding singular values:")
print("   Singular values = sqrt(eigenvalues of AᵀA)")

AtA = A.T @ A
print(f"\n   AᵀA = \n{AtA}")

eigenvalues, _ = eig(AtA)
eigenvalues = sorted(eigenvalues.real, reverse=True)
print(f"   Eigenvalues of AᵀA: {eigenvalues}")

singular_values = np.sqrt(eigenvalues)
print(f"   Singular values: σ₁ = {singular_values[0]:.4f}, σ₂ = {singular_values[1]:.4f}")

# Verify with SVD
_, s, _ = np.linalg.svd(A)
print(f"   NumPy SVD: {s}")
````

````python
print("b) Spectral norm:")
spectral_norm = singular_values[0]
print(f"   ||A||₂ = σ_max = {spectral_norm:.4f}")
print(f"   NumPy: {np.linalg.norm(A, 2):.4f}")

print("\nc) Condition number:")
condition = singular_values[0] / singular_values[1]
print(f"   κ(A) = σ_max / σ_min = {condition:.4f}")
print(f"   NumPy: {np.linalg.cond(A):.4f}")

print("\n   Interpretation:")
print("   - κ(A) ≈ 1: well-conditioned (stable)")
print("   - κ(A) >> 1: ill-conditioned (sensitive to errors)")
print(f"   - This matrix has moderate conditioning (κ ≈ {condition:.1f})")
````
