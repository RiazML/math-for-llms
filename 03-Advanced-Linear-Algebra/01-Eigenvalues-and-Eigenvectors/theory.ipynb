## 1. Basic Eigenvalues and Eigenvectors

For matrix $A$, vector $\mathbf{v}$ is an eigenvector with eigenvalue $\lambda$ if:
$$A\mathbf{v} = \lambda\mathbf{v}$$

The matrix transforms the eigenvector by only scaling it (by factor $\lambda$), without changing its direction.

````python
# Define a 2x2 matrix
A = np.array([[4, 1],
              [2, 3]])

print(f"Matrix A:\n{A}")

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = eig(A)

print(f"\nEigenvalues: {eigenvalues}")
print(f"\nEigenvectors (as columns):\n{eigenvectors}")
````

````python
# Verify Av = λv for each eigenvalue-eigenvector pair
print("Verification that Av = λv:")

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    Av = A @ v
    lam_v = lam * v
    
    print(f"\nλ_{i+1} = {lam:.4f}")
    print(f"v_{i+1} = {v}")
    print(f"Av = {Av}")
    print(f"λv = {lam_v}")
    print(f"Match: {np.allclose(Av, lam_v)}")
````

## 2. Trace and Determinant Relationships

For a matrix $A$ with eigenvalues $\lambda_1, \lambda_2, \ldots, \lambda_n$:

$$\text{tr}(A) = \sum_{i=1}^{n} \lambda_i \quad \text{and} \quad \det(A) = \prod_{i=1}^{n} \lambda_i$$

````python
A = np.array([[4, 1],
              [2, 3]])

eigenvalues, _ = eig(A)

print(f"Matrix A:\n{A}")
print(f"\nEigenvalues: {eigenvalues}")

# Trace = sum of eigenvalues
trace_A = np.trace(A)
sum_eigenvalues = np.sum(eigenvalues)
print(f"\n--- Trace ---")
print(f"tr(A) = {trace_A}")
print(f"λ₁ + λ₂ = {sum_eigenvalues.real:.4f}")
print(f"Equal: {np.isclose(trace_A, sum_eigenvalues.real)}")

# Determinant = product of eigenvalues
det_A = det(A)
prod_eigenvalues = np.prod(eigenvalues)
print(f"\n--- Determinant ---")
print(f"det(A) = {det_A}")
print(f"λ₁ × λ₂ = {prod_eigenvalues.real:.4f}")
print(f"Equal: {np.isclose(det_A, prod_eigenvalues.real)}")
````

## 3. Eigenvalue Properties Under Transformations

| Transformation | Eigenvalues |
|----------------|-------------|
| $A^k$ | $\lambda^k$ |
| $A^{-1}$ | $1/\lambda$ |
| $A + cI$ | $\lambda + c$ |
| $cA$ | $c\lambda$ |

````python
A = np.array([[3, 1],
              [0, 2]])

eigenvalues_A, _ = eig(A)
print(f"Matrix A:\n{A}")
print(f"Eigenvalues of A: {eigenvalues_A}")

# A^2
A2 = A @ A
eigenvalues_A2, _ = eig(A2)
print(f"\n--- A² ---")
print(f"A²:\n{A2}")
print(f"Eigenvalues of A²: {eigenvalues_A2}")
print(f"λ² values: {eigenvalues_A**2}")
````

````python
# A^(-1)
A_inv = inv(A)
eigenvalues_inv, _ = eig(A_inv)
print(f"--- A⁻¹ ---")
print(f"A⁻¹:\n{A_inv}")
print(f"Eigenvalues of A⁻¹: {eigenvalues_inv}")
print(f"1/λ values: {1/eigenvalues_A}")

# A + 2I
A_shift = A + 2 * np.eye(2)
eigenvalues_shift, _ = eig(A_shift)
print(f"\n--- A + 2I ---")
print(f"A + 2I:\n{A_shift}")
print(f"Eigenvalues of A + 2I: {eigenvalues_shift}")
print(f"λ + 2 values: {eigenvalues_A + 2}")

# 3A
A_scaled = 3 * A
eigenvalues_scaled, _ = eig(A_scaled)
print(f"\n--- 3A ---")
print(f"3A:\n{A_scaled}")
print(f"Eigenvalues of 3A: {eigenvalues_scaled}")
print(f"3λ values: {3 * eigenvalues_A}")
````

## 4. Symmetric Matrix Properties

For symmetric matrices ($A = A^T$):
1. All eigenvalues are **real**
2. Eigenvectors for distinct eigenvalues are **orthogonal**
3. Always **diagonalizable**: $A = Q\Lambda Q^T$ where $Q$ is orthogonal

````python
# Symmetric matrix
A = np.array([[4, 2, 0],
              [2, 5, 3],
              [0, 3, 6]])

print(f"Symmetric matrix A:\n{A}")
print(f"\nA = Aᵀ: {np.allclose(A, A.T)}")

eigenvalues, eigenvectors = eig(A)

print(f"\nEigenvalues: {eigenvalues.real}")
print("Note: All eigenvalues are real!")
````

````python
print(f"Eigenvectors:\n{eigenvectors.real}")

# Check orthogonality
print("\nOrthogonality check (vᵢ·vⱼ for i≠j):")
for i in range(3):
    for j in range(i+1, 3):
        dot = np.dot(eigenvectors[:, i], eigenvectors[:, j])
        print(f"  v{i+1}·v{j+1} = {dot.real:.6f}")

# Verify Q^T Q = I
Q = eigenvectors.real
print(f"\nQᵀQ (should be identity):\n{np.round(Q.T @ Q, 6)}")
````

## 5. Matrix Diagonalization

A matrix $A$ is diagonalizable if:
$$A = PDP^{-1}$$

where:
- $D$: diagonal matrix with eigenvalues
- $P$: matrix with eigenvectors as columns

````python
A = np.array([[4, 1],
              [2, 3]])

eigenvalues, P = eig(A)
D = np.diag(eigenvalues)

print(f"Matrix A:\n{A}")
print(f"\nDiagonal matrix D (eigenvalues):\n{D.real}")
print(f"\nEigenvector matrix P:\n{P.real}")
````

````python
# Verify A = PDP^(-1)
P_inv = inv(P)
reconstructed = P @ D @ P_inv

print(f"P⁻¹:\n{P_inv.real}")
print(f"\nPDP⁻¹:\n{reconstructed.real}")
print(f"\nA = PDP⁻¹: {np.allclose(A, reconstructed)}")
````

## 6. Matrix Powers via Diagonalization

If $A = PDP^{-1}$, then:
$$A^k = PD^kP^{-1}$$

This is efficient because $D^k$ is simply the diagonal matrix with $\lambda_i^k$ on the diagonal.

````python
A = np.array([[2, 1],
              [0, 3]])

print(f"Matrix A:\n{A}")

# Get eigendecomposition
eigenvalues, P = eig(A)
D = np.diag(eigenvalues)
P_inv = inv(P)

# Compute A^5 two ways
k = 5

# Method 1: Direct computation
A_k_direct = matrix_power(A, k)

# Method 2: Using diagonalization
D_k = np.diag(eigenvalues ** k)
A_k_diag = P @ D_k @ P_inv

print(f"\nA^{k} via direct multiplication:\n{A_k_direct}")
print(f"\nA^{k} via PD^{k}P⁻¹:\n{A_k_diag.real}")
print(f"\nResults match: {np.allclose(A_k_direct, A_k_diag)}")

# Show D^k is easy to compute
print(f"\nD^{k} = diag(λ₁^{k}, λ₂^{k}):")
print(f"  λ₁^{k} = {eigenvalues[0]**k:.0f}")
print(f"  λ₂^{k} = {eigenvalues[1]**k:.0f}")
````

## 7. Spectral Decomposition

For a symmetric matrix $A$:
$$A = \sum_{i=1}^{n} \lambda_i \mathbf{v}_i \mathbf{v}_i^T$$

This expresses $A$ as a sum of rank-1 matrices, each scaled by an eigenvalue.

````python
# Symmetric matrix
A = np.array([[3, 1],
              [1, 3]])

print(f"Symmetric matrix A:\n{A}")

eigenvalues, eigenvectors = eig(A)

# Normalize eigenvectors
v1 = eigenvectors[:, 0].real
v2 = eigenvectors[:, 1].real
v1 = v1 / np.linalg.norm(v1)
v2 = v2 / np.linalg.norm(v2)

lam1, lam2 = eigenvalues.real

print(f"\nEigenvalues: λ₁ = {lam1}, λ₂ = {lam2}")
print(f"Eigenvectors: v₁ = {v1}, v₂ = {v2}")
````

````python
# Spectral decomposition: A = λ₁v₁v₁ᵀ + λ₂v₂v₂ᵀ
term1 = lam1 * np.outer(v1, v1)
term2 = lam2 * np.outer(v2, v2)

print(f"λ₁v₁v₁ᵀ:\n{term1}")
print(f"\nλ₂v₂v₂ᵀ:\n{term2}")

reconstructed = term1 + term2
print(f"\nλ₁v₁v₁ᵀ + λ₂v₂v₂ᵀ:\n{reconstructed}")
print(f"\nMatches A: {np.allclose(A, reconstructed)}")
````

## 8. Power Method

The power method iteratively finds the **dominant eigenvalue** (largest magnitude):

```
v₀ = random vector
for k = 1, 2, ...:
    w = Av_{k-1}
    v_k = w / ||w||
    λ ≈ Rayleigh quotient = (v^T A v) / (v^T v)
```

````python
A = np.array([[4, 1],
              [2, 3]])

print(f"Matrix A:\n{A}")

# True eigenvalues
true_eigenvalues, true_eigenvectors = eig(A)
print(f"\nTrue eigenvalues: {true_eigenvalues}")
print(f"Dominant eigenvalue: {max(abs(true_eigenvalues))}")

# Power method
np.random.seed(42)
v = np.random.rand(2)
v = v / np.linalg.norm(v)

print(f"\nPower Method Iterations:")
print(f"{'Iter':>4} {'Estimated λ':>15} {'Error':>15}")

for i in range(10):
    w = A @ v
    eigenvalue_estimate = np.dot(w, v) / np.dot(v, v)  # Rayleigh quotient
    v = w / np.linalg.norm(w)
    error = abs(eigenvalue_estimate - max(true_eigenvalues.real))
    print(f"{i+1:4d} {eigenvalue_estimate:15.10f} {error:15.10f}")

print(f"\nFinal eigenvector estimate: {v}")
````

## 9. Covariance Matrix Eigenanalysis (PCA Preview)

In PCA, we analyze the eigendecomposition of the covariance matrix:
- **Eigenvalues** = variance along each principal axis
- **Eigenvectors** = principal component directions
- **Largest eigenvalue** → direction of maximum variance

````python
# Generate correlated 2D data
np.random.seed(42)
n = 100

# Create correlation
mean = [0, 0]
cov = [[3, 1.5],
       [1.5, 1]]

data = np.random.multivariate_normal(mean, cov, n)

print(f"Generated {n} data points with covariance:\n{np.array(cov)}")

# Compute sample covariance
X = data - data.mean(axis=0)  # Center the data
C = (X.T @ X) / (n - 1)

print(f"\nSample covariance matrix:\n{np.round(C, 4)}")
````

````python
# Eigendecomposition
eigenvalues, eigenvectors = eig(C)

# Sort by eigenvalue (descending)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print(f"Eigenvalues (variances along principal axes):")
for i, lam in enumerate(eigenvalues):
    print(f"  PC{i+1}: λ = {lam.real:.4f}")

print(f"\nPrincipal components (eigenvectors):")
for i in range(2):
    print(f"  PC{i+1}: {eigenvectors[:, i].real}")

# Variance explained
total_var = np.sum(eigenvalues)
print(f"\nVariance explained:")
for i, lam in enumerate(eigenvalues):
    pct = 100 * lam.real / total_var.real
    print(f"  PC{i+1}: {pct:.1f}%")
````

````python
# Visualize PCA
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(X[:, 0], X[:, 1], alpha=0.5, label='Data')

# Plot principal components
origin = [0, 0]
colors = ['red', 'blue']
for i in range(2):
    scale = np.sqrt(eigenvalues[i].real) * 2
    pc = eigenvectors[:, i].real * scale
    ax.annotate('', xy=pc, xytext=origin,
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=3))
    ax.annotate(f'PC{i+1} (λ={eigenvalues[i].real:.2f})', pc, fontsize=12, color=colors[i])

ax.set_xlabel('X₁')
ax.set_ylabel('X₂')
ax.set_title('PCA: Principal Components of Covariance Matrix')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
````

## 10. Markov Chain Stationary Distribution

For a Markov chain with transition matrix $P$, the **stationary distribution** $\pi$ satisfies:
$$\pi P = \pi$$

This is an eigenvector equation with $\lambda = 1$!

````python
# Transition matrix (rows sum to 1)
# State 1: Sunny, State 2: Rainy
P = np.array([[0.8, 0.2],   # From Sunny
              [0.4, 0.6]])  # From Rainy

print("Weather Markov Chain")
print(f"Transition matrix P:\n{P}")
print("  P[i,j] = P(tomorrow=j | today=i)")

# Find eigenvalues
eigenvalues, eigenvectors = eig(P.T)  # Note: using P^T for left eigenvector

print(f"\nEigenvalues of Pᵀ: {eigenvalues}")
````

````python
# Find eigenvector for λ = 1 (stationary distribution)
idx = np.argmin(np.abs(eigenvalues - 1))
stationary = eigenvectors[:, idx].real
stationary = stationary / np.sum(stationary)  # Normalize to probability

print(f"Stationary distribution π (eigenvector for λ=1):")
print(f"  π = {stationary}")
print(f"  P(Sunny) = {stationary[0]:.4f}")
print(f"  P(Rainy) = {stationary[1]:.4f}")

# Verify: π P = π
pi_P = stationary @ P
print(f"\nVerification: πP = {pi_P}")
print(f"πP = π: {np.allclose(pi_P, stationary)}")
````

````python
# Long-run behavior: P^k converges to stationary distribution
P_100 = matrix_power(P, 100)
print(f"P¹⁰⁰ (each row converges to stationary):")
print(np.round(P_100, 4))
print(f"\nNote: All rows are identical = stationary distribution!")
````

## 11. Complex Eigenvalues (Rotation)

Rotation matrices have complex eigenvalues:
$$R_\theta = \begin{bmatrix} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{bmatrix}$$

Eigenvalues: $e^{\pm i\theta} = \cos\theta \pm i\sin\theta$

Complex eigenvalues indicate rotation (no real eigenvector directions are preserved).

````python
# Rotation matrix by 90 degrees
theta = np.pi / 2  # 90 degrees
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])

print(f"Rotation matrix (90°):\n{np.round(R, 4)}")

eigenvalues, eigenvectors = eig(R)

print(f"\nEigenvalues: {eigenvalues}")
print(f"  λ₁ = {eigenvalues[0]:.4f} = e^(iπ/2) = i")
print(f"  λ₂ = {eigenvalues[1]:.4f} = e^(-iπ/2) = -i")

print(f"\n|λ₁| = {abs(eigenvalues[0]):.4f}")
print(f"|λ₂| = {abs(eigenvalues[1]):.4f}")
print("Note: |λ| = 1 for rotation matrices (preserve length)!")
````

````python
# Visualize rotation (no eigenvector direction preserved)
fig, ax = plt.subplots(figsize=(6, 6))

# Unit circle
t = np.linspace(0, 2*np.pi, 100)
ax.plot(np.cos(t), np.sin(t), 'b--', alpha=0.3, label='Unit circle')

# Test vectors and their rotations
test_vectors = [[1, 0], [0.7, 0.7], [0, 1], [-1, 0]]
colors = ['red', 'green', 'purple', 'orange']

for v, c in zip(test_vectors, colors):
    v = np.array(v)
    Rv = R @ v
    ax.arrow(0, 0, v[0]*0.9, v[1]*0.9, head_width=0.08, color=c, alpha=0.5)
    ax.arrow(0, 0, Rv[0]*0.9, Rv[1]*0.9, head_width=0.08, color=c, linestyle='--')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.set_title('90° Rotation: No real eigenvector\n(all vectors change direction)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 12. Positive Definite Matrices

A symmetric matrix $A$ is **positive definite** if:
- All eigenvalues are **positive** ($\lambda_i > 0$)
- Equivalently: $\mathbf{x}^T A \mathbf{x} > 0$ for all $\mathbf{x} \neq \mathbf{0}$

Positive definite matrices are crucial in optimization (Hessians) and statistics (covariance matrices).

````python
# Positive definite matrix
A = np.array([[4, 2],
              [2, 3]])

print(f"Matrix A:\n{A}")

eigenvalues, _ = eig(A)
print(f"\nEigenvalues: {eigenvalues.real}")
print(f"All positive: {all(eigenvalues.real > 0)}")
print(f"A is positive definite: {all(eigenvalues.real > 0)}")
````

````python
# Test quadratic form x^T A x > 0
print("Quadratic form xᵀAx for various x:")
test_vectors = [
    np.array([1, 0]),
    np.array([0, 1]),
    np.array([1, 1]),
    np.array([-1, 2]),
    np.array([0.5, -0.5])
]

for x in test_vectors:
    quad_form = x.T @ A @ x
    print(f"  x = {x}, xᵀAx = {quad_form:.4f} > 0: {quad_form > 0}")
````

````python
# Cholesky decomposition exists for positive definite
L = np.linalg.cholesky(A)
print(f"Cholesky decomposition exists (A = LLᵀ):")
print(f"\nL (lower triangular):\n{L}")
print(f"\nLLᵀ:\n{L @ L.T}")
print(f"\nA = LLᵀ: {np.allclose(A, L @ L.T)}")
````

## 13. Visualization: Eigenvector Transformation

Visualizing how eigenvectors are transformed compared to regular vectors.

````python
A = np.array([[2, 1],
              [1, 2]])

eigenvalues, eigenvectors = eig(A)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Unit circle transformation
ax = axes[0]

# Unit circle
theta = np.linspace(0, 2*np.pi, 100)
circle = np.array([np.cos(theta), np.sin(theta)])
transformed = A @ circle

ax.plot(circle[0], circle[1], 'b-', alpha=0.5, linewidth=2, label='Unit circle')
ax.plot(transformed[0], transformed[1], 'r-', alpha=0.5, linewidth=2, label='Transformed')

# Eigenvectors
colors = ['green', 'purple']
for i in range(2):
    v = eigenvectors[:, i].real
    v = v / np.linalg.norm(v)
    Av = A @ v
    
    ax.annotate('', xy=v, xytext=[0, 0],
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=2))
    ax.annotate('', xy=Av, xytext=[0, 0],
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=2, ls='--', alpha=0.7))
    ax.annotate(f'v{i+1}', v + 0.15, fontsize=12, color=colors[i])
    ax.annotate(f'Av{i+1}={eigenvalues[i].real:.1f}v{i+1}', Av + 0.1, fontsize=10, color=colors[i])

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.set_title(f'Eigenvectors remain on same line\nλ₁={eigenvalues[0].real:.2f}, λ₂={eigenvalues[1].real:.2f}')
ax.legend()

# Plot 2: Eigenvalue spectrum
ax = axes[1]
bars = ax.bar(range(1, len(eigenvalues)+1), eigenvalues.real, color=['green', 'purple'], alpha=0.7)
ax.set_xlabel('Eigenvalue index', fontsize=12)
ax.set_ylabel('Eigenvalue', fontsize=12)
ax.set_title('Eigenvalue Spectrum', fontsize=14)
ax.set_xticks([1, 2])
ax.set_xticklabels(['λ₁', 'λ₂'], fontsize=12)

for bar, val in zip(bars, eigenvalues):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
            f'{val.real:.2f}', ha='center', fontsize=12)

ax.axhline(0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
````

## Summary

| Concept | Key Formula | NumPy |
|---------|-------------|-------|
| Eigenvalue equation | $A\mathbf{v} = \lambda\mathbf{v}$ | `np.linalg.eig(A)` |
| Characteristic equation | $\det(A - \lambda I) = 0$ | - |
| Trace | $\text{tr}(A) = \sum \lambda_i$ | `np.trace(A)` |
| Determinant | $\det(A) = \prod \lambda_i$ | `np.linalg.det(A)` |
| Diagonalization | $A = PDP^{-1}$ | eigenvectors, eigenvalues |
| Matrix powers | $A^k = PD^kP^{-1}$ | `np.linalg.matrix_power(A, k)` |
| Spectral decomposition | $A = \sum \lambda_i v_i v_i^T$ | - |

### ML Applications
- **PCA**: Eigenvectors of covariance matrix
- **PageRank**: Dominant eigenvector
- **Spectral clustering**: Graph Laplacian eigenvectors
- **RNN stability**: Eigenvalue magnitudes
