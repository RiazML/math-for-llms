### Solution

```python
def classical_gram_schmidt_solution(V: np.ndarray) -> np.ndarray:
    n, k = V.shape
    Q = np.zeros((n, k))
    
    for j in range(k):
        q = V[:, j].copy()
        for i in range(j):
            q -= np.dot(Q[:, i], V[:, j]) * Q[:, i]
        norm = np.linalg.norm(q)
        if norm > 1e-10:
            Q[:, j] = q / norm
    return Q

def modified_gram_schmidt_solution(V: np.ndarray) -> np.ndarray:
    n, k = V.shape
    Q = V.copy().astype(float)
    
    for j in range(k):
        norm = np.linalg.norm(Q[:, j])
        if norm > 1e-10:
            Q[:, j] /= norm
        for i in range(j + 1, k):
            Q[:, i] -= np.dot(Q[:, j], Q[:, i]) * Q[:, j]
    return Q

# Test with solution
Q_classical = classical_gram_schmidt_solution(V)
Q_modified = modified_gram_schmidt_solution(V)
print(f"Classical orthogonality error: {np.linalg.norm(Q_classical.T @ Q_classical - np.eye(4)):.2e}")
print(f"Modified orthogonality error: {np.linalg.norm(Q_modified.T @ Q_modified - np.eye(4)):.2e}")
```

```python
# Test
print("Exercise 6: Eigenspaces and Diagonalization")
print("-" * 40)

A_not_diag = np.array([
    [2, 1],
    [0, 2]
], dtype=float)  # Not diagonalizable

B_diag = np.array([
    [3, 1],
    [0, 2]
], dtype=float)  # Diagonalizable

# Uncomment to test:
# print(f"  A diagonalizable: {is_diagonalizable(A_not_diag)}")
# print(f"  B diagonalizable: {is_diagonalizable(B_diag)}")
```

### Solution

```python
def is_diagonalizable_solution(A):
    eigenvalues, eigenvectors = np.linalg.eig(A)
    n = A.shape[0]
    
    # Check if we have n linearly independent eigenvectors
    _, s, _ = np.linalg.svd(eigenvectors)
    return np.sum(s > 1e-10) == n

def diagonalize_solution(A):
    if not is_diagonalizable_solution(A):
        return None
    
    eigenvalues, eigenvectors = np.linalg.eig(A)
    P = eigenvectors
    D = np.diag(eigenvalues)
    
    return P, D

def matrix_power_via_diagonalization_solution(A, k):
    result = diagonalize_solution(A)
    if result is None:
        return np.linalg.matrix_power(A, k)
    P, D = result
    D_k = np.diag(np.diag(D) ** k)
    return P @ D_k @ np.linalg.inv(P)

# Test
print(f"A (Jordan block) diagonalizable: {is_diagonalizable_solution(A_not_diag)}")
print(f"B diagonalizable: {is_diagonalizable_solution(B_diag)}")

# Verify diagonalization
P, D = diagonalize_solution(B_diag)
B_reconstructed = P @ D @ np.linalg.inv(P)
print(f"\nB = PDP^-1: {np.allclose(B_diag, B_reconstructed)}")

# Test matrix power
B_5 = matrix_power_via_diagonalization_solution(B_diag, 5)
B_5_direct = np.linalg.matrix_power(B_diag, 5)
print(f"B^5 via diagonalization matches direct: {np.allclose(B_5, B_5_direct)}")
```

### Solution

```python
def kernel_feature_space_solution(X, kernel='polynomial', degree=2):
    """Polynomial feature expansion."""
    from itertools import combinations_with_replacement
    
    n_samples, n_features = X.shape
    features = [np.ones(n_samples)]  # Bias term
    
    for d in range(1, degree + 1):
        for combo in combinations_with_replacement(range(n_features), d):
            feature = np.ones(n_samples)
            for idx in combo:
                feature *= X[:, idx]
            features.append(feature)
    
    return np.column_stack(features)

def attention_as_projection_solution(Q, K, V):
    """Scaled dot-product attention."""
    d_k = K.shape[-1]
    scores = Q @ K.T / np.sqrt(d_k)
    
    # Softmax
    exp_scores = np.exp(scores - scores.max(axis=-1, keepdims=True))
    attention_weights = exp_scores / exp_scores.sum(axis=-1, keepdims=True)
    
    return attention_weights @ V

# Test kernel feature space
X_expanded = kernel_feature_space_solution(X_ml, 'polynomial', degree=2)
print(f"Original dim: {X_ml.shape[1]}")
print(f"Expanded dim (degree 2): {X_expanded.shape[1]}")

# Test attention
np.random.seed(42)
seq_len, d_model = 4, 8
Q = np.random.randn(seq_len, d_model)
K = np.random.randn(seq_len, d_model)
V = np.random.randn(seq_len, d_model)

output = attention_as_projection_solution(Q, K, V)
print(f"\nAttention output shape: {output.shape}")
print(f"Each output is a weighted combination of V vectors.")
```
