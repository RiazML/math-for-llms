```python
# Exercise 2: Solution ✅

def power_set(s: set) -> Set[FrozenSet]:
    s = list(s)
    result = set()
    for size in range(len(s) + 1):
        for combo in itertools.combinations(s, size):
            result.add(frozenset(combo))
    return result

# (a) Subset relationships
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}
C = {1, 2, 3}

print("Subset relationships:")
print(f"  A ⊆ B: {A <= B} (A is subset of B)")
print(f"  A ⊂ B: {A < B}  (A is proper subset of B)")
print(f"  C ⊆ A: {C <= A} (C equals A)")
print(f"  C ⊂ A: {C < A}  (C is NOT proper subset — they're equal)")
print(f"  B ⊇ A: {B >= A} (B is superset of A)")

# (b) Power set
ps = power_set({1, 2, 3})
assert len(ps) == 8
print(f"\nP({{1,2,3}}) has {len(ps)} elements:")
for s in sorted(ps, key=lambda x: (len(x), sorted(x))):
    print(f"  {set(s) if s else '{}'}")

# (c) Verify |P(S)| = 2^|S| for sizes 0-6
print(f"\nVerify |P(S)| = 2^|S|:")
for n in range(7):
    S = set(range(n))
    ps_size = len(power_set(S))
    expected = 2 ** n
    assert ps_size == expected
    print(f"  |S|={n}: |P(S)|={ps_size} = 2^{n}={expected} ✓")

# (d) Vacuous truth
print(f"\nVacuous truth: ∅ ⊆ S for any S:")
for S in [set(), {1}, {1,2,3}, set(range(100))]:
    assert set() <= S
print(f"  Verified for empty, {{1}}, {{1,2,3}}, {{0..99}} ✓")

print(f'\nAll tests passed ✓')
print(f'→ Feature selection = choosing a member of P(features)')
```

```python
# Exercise 4: Solution ✅

U = set(range(50))
np.random.seed(42)

def random_subset(universe: set) -> set:
    elements = list(universe)
    size = np.random.randint(0, len(universe) + 1)
    chosen = np.random.choice(elements, size=size, replace=False)
    return set(chosen.tolist())

n_tests = 1000
dm1_pass = dm2_pass = abs1_pass = abs2_pass = 0

for _ in range(n_tests):
    A = random_subset(U)
    B = random_subset(U)
    comp = lambda S: U - S
    
    # De Morgan 1: C(A ∩ B) = C(A) ∪ C(B)
    if comp(A & B) == comp(A) | comp(B):
        dm1_pass += 1
    
    # De Morgan 2: C(A ∪ B) = C(A) ∩ C(B)
    if comp(A | B) == comp(A) & comp(B):
        dm2_pass += 1
    
    # Absorption 1: A ∩ (A ∪ B) = A
    if A & (A | B) == A:
        abs1_pass += 1
    
    # Absorption 2: A ∪ (A ∩ B) = A
    if A | (A & B) == A:
        abs2_pass += 1

print(f"Exhaustive Verification over {n_tests} random set pairs")
print(f"=" * 55)
print(f"  De Morgan 1: C(A∩B) = C(A)∪C(B)  →  {dm1_pass}/{n_tests} ✓")
print(f"  De Morgan 2: C(A∪B) = C(A)∩C(B)  →  {dm2_pass}/{n_tests} ✓")
print(f"  Absorption:  A∩(A∪B) = A          →  {abs1_pass}/{n_tests} ✓")
print(f"  Absorption:  A∪(A∩B) = A          →  {abs2_pass}/{n_tests} ✓")

assert dm1_pass == dm2_pass == abs1_pass == abs2_pass == n_tests
print(f'\nAll {4 * n_tests} law checks passed ✓')
print(f'→ These identities hold universally — use them to simplify set expressions')
```

```python
# Exercise 12: Solution ✅

def confusion_matrix_sets(y_true: list, y_pred: list, classes: list) -> dict:
    N = len(y_true)
    all_idx = set(range(N))
    results = {}
    
    for c in classes:
        actual_c = {i for i in range(N) if y_true[i] == c}
        pred_c = {i for i in range(N) if y_pred[i] == c}
        
        tp = actual_c & pred_c
        fp = pred_c - actual_c
        fn = actual_c - pred_c
        tn = all_idx - (actual_c | pred_c)
        
        assert len(tp) + len(fp) + len(fn) + len(tn) == N
        
        prec = len(tp) / len(pred_c) if pred_c else 0
        rec = len(tp) / len(actual_c) if actual_c else 0
        f1 = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0
        
        results[c] = {
            'TP': len(tp), 'FP': len(fp), 'FN': len(fn), 'TN': len(tn),
            'precision': prec, 'recall': rec, 'f1': f1
        }
    return results

# Test
np.random.seed(42)
n = 100
classes = ['cat', 'dog', 'bird']
y_true = [classes[i] for i in np.random.choice(3, n)]
y_pred = y_true.copy()
for i in np.random.choice(n, 20, replace=False):
    y_pred[i] = classes[np.random.choice(3)]

result = confusion_matrix_sets(y_true, y_pred, classes)

print("Per-Class Metrics (via Set Operations)")
print("=" * 60)
print(f"  {'Class':>6} │ {'TP':>3} {'FP':>3} {'FN':>3} {'TN':>3} │ {'Prec':>5} {'Rec':>5} {'F1':>5}")
print(f"  {'─'*6} │ {'─'*3} {'─'*3} {'─'*3} {'─'*3} │ {'─'*5} {'─'*5} {'─'*5}")

macro_p = macro_r = macro_f1 = 0
for c in classes:
    r = result[c]
    print(f"  {c:>6} │ {r['TP']:>3} {r['FP']:>3} {r['FN']:>3} {r['TN']:>3} │ "
          f"{r['precision']:.3f} {r['recall']:.3f} {r['f1']:.3f}")
    assert r['TP'] + r['FP'] + r['FN'] + r['TN'] == n
    macro_p += r['precision']
    macro_r += r['recall']
    macro_f1 += r['f1']

macro_p /= len(classes)
macro_r /= len(classes)
macro_f1 /= len(classes)
print(f"\n  Macro avg: Precision={macro_p:.3f}, Recall={macro_r:.3f}, F1={macro_f1:.3f}")
print(f"  TP+FP+FN+TN = {n} for all classes ✓")
print(f'\nAll tests passed ✓')
print(f'→ Every classification metric is a ratio of set cardinalities')
```

```python
# Exercise 14: Solution ✅

import random

def split_dataset(indices, ratios, seed=42):
    indices = list(indices)
    rng = random.Random(seed)
    rng.shuffle(indices)
    
    splits = {}
    start = 0
    names = list(ratios.keys())
    for i, name in enumerate(names):
        if i == len(names) - 1:
            splits[name] = set(indices[start:])
        else:
            end = start + int(len(indices) * ratios[name])
            splits[name] = set(indices[start:end])
            start = end
    return splits

def verify_partition(splits, universe):
    universe = set(universe)
    split_names = list(splits.keys())
    
    # Check disjointness
    leakage = {}
    is_disjoint = True
    for i in range(len(split_names)):
        for j in range(i+1, len(split_names)):
            overlap = splits[split_names[i]] & splits[split_names[j]]
            if overlap:
                is_disjoint = False
                leakage[f"{split_names[i]}∩{split_names[j]}"] = overlap
    
    # Check exhaustiveness
    covered = set()
    for s in splits.values():
        covered |= s
    is_exhaustive = (covered == universe)
    missing = universe - covered
    extra = covered - universe
    
    return {
        'is_disjoint': is_disjoint,
        'is_exhaustive': is_exhaustive,
        'is_partition': is_disjoint and is_exhaustive,
        'leakage': leakage,
        'missing': missing,
        'extra': extra
    }

def stratified_split(indices, labels, ratios, seed=42):
    rng = random.Random(seed)
    from collections import defaultdict
    
    # Group by label
    by_label = defaultdict(list)
    for idx, lbl in zip(indices, labels):
        by_label[lbl].append(idx)
    
    splits = {name: set() for name in ratios}
    names = list(ratios.keys())
    
    for lbl, group in by_label.items():
        rng.shuffle(group)
        start = 0
        for i, name in enumerate(names):
            if i == len(names) - 1:
                splits[name] |= set(group[start:])
            else:
                end = start + int(len(group) * ratios[name])
                splits[name] |= set(group[start:end])
                start = end
    return splits

def k_fold_splits(indices, k=5, seed=42):
    indices = list(indices)
    rng = random.Random(seed)
    rng.shuffle(indices)
    
    fold_size = len(indices) // k
    folds = []
    for i in range(k):
        if i == k - 1:
            val = set(indices[i * fold_size:])
        else:
            val = set(indices[i * fold_size:(i+1) * fold_size])
        train = set(indices) - val
        folds.append((train, val))
    return folds


# === Test splits ===
N = 1000
universe = set(range(N))
ratios = {'train': 0.7, 'val': 0.15, 'test': 0.15}

splits = split_dataset(universe, ratios)
result = verify_partition(splits, universe)

print("Basic 70/15/15 Split")
print("=" * 45)
for name, s in splits.items():
    print(f"  {name:6s}: {len(s):4d} samples ({len(s)/N:.1%})")
print(f"\n  Disjoint:    {result['is_disjoint']}")
print(f"  Exhaustive:  {result['is_exhaustive']}")
print(f"  Partition:   {result['is_partition']}")
assert result['is_partition'], "Split should form a partition"

# === Stratified split ===
labels = [i % 3 for i in range(N)]  # 3 balanced classes
strat_splits = stratified_split(list(universe), labels, ratios)
strat_result = verify_partition(strat_splits, universe)
assert strat_result['is_partition'], "Stratified split should be partition"

print(f"\nStratified Split (3 classes)")
print("=" * 45)
for name, s in strat_splits.items():
    label_dist = [sum(1 for idx in s if labels[idx] == c) for c in range(3)]
    pcts = [f'{x/len(s):.1%}' for x in label_dist]
    print(f"  {name:6s}: {len(s):4d} samples — class dist: {pcts}")

# === K-fold cross-validation ===
folds = k_fold_splits(universe, k=5)
print(f"\n5-Fold Cross-Validation")
print("=" * 45)
for i, (train, val) in enumerate(folds):
    fold_result = verify_partition({'train': train, 'val': val}, universe)
    assert fold_result['is_partition'], f"Fold {i} should be partition"
    print(f"  Fold {i}: train={len(train):4d}, val={len(val):3d} — partition: ✓")

# Verify each sample appears in exactly one validation fold
all_val = []
for _, val in folds:
    all_val.extend(val)
assert len(all_val) == N, f"Should have {N} val assignments, got {len(all_val)}"
assert set(all_val) == universe, "Each sample should appear in some val fold"

# === Detect data leakage ===
print(f"\nData Leakage Detection")
print("=" * 45)
leaked = splits.copy()
leaked['train'] = leaked['train'] | {list(leaked['val'])[0], list(leaked['test'])[0]}
leak_result = verify_partition(leaked, universe)
assert not leak_result['is_disjoint'], "Should detect leakage"
for pair, overlap in leak_result['leakage'].items():
    print(f"  ⚠ Leakage in {pair}: {len(overlap)} samples")
print(f"  Leakage detection working: ✓")

print(f'\nAll tests passed ✓')
print(f'→ Partitions guarantee no data leakage between train/val/test')
```

```python
# Exercise 16: Solution ✅

def simplify(clauses, literal):
    """Remove clauses containing literal, remove ¬literal from remaining."""
    new_clauses = []
    for clause in clauses:
        if literal in clause:
            continue  # clause satisfied
        new_clause = [l for l in clause if l != -literal]
        new_clauses.append(new_clause)
    return new_clauses

def unit_propagation(clauses, assignment):
    assignment = dict(assignment)
    changed = True
    while changed:
        changed = False
        for clause in clauses:
            if len(clause) == 0:
                return None, None  # conflict: empty clause
            if len(clause) == 1:
                lit = clause[0]
                var = abs(lit)
                val = lit > 0
                if var in assignment:
                    if assignment[var] != val:
                        return None, None  # conflict
                else:
                    assignment[var] = val
                    clauses = simplify(clauses, lit)
                    changed = True
                    break
    return clauses, assignment

def pure_literal_elimination(clauses, assignment):
    assignment = dict(assignment)
    all_literals = set()
    for clause in clauses:
        for lit in clause:
            all_literals.add(lit)
    
    for lit in list(all_literals):
        if -lit not in all_literals:
            var = abs(lit)
            if var not in assignment:
                assignment[var] = (lit > 0)
                clauses = simplify(clauses, lit)
    return clauses, assignment

def dpll(clauses, assignment=None):
    if assignment is None:
        assignment = {}
    
    # Unit propagation
    clauses, assignment = unit_propagation(clauses, assignment)
    if clauses is None:
        return False, None
    
    # Pure literal elimination
    clauses, assignment = pure_literal_elimination(clauses, assignment)
    
    # Base cases
    if len(clauses) == 0:
        return True, assignment  # all clauses satisfied
    if any(len(c) == 0 for c in clauses):
        return False, None  # empty clause = conflict
    
    # Choose unassigned variable (pick first literal of first clause)
    var = abs(clauses[0][0])
    
    # Try True
    new_clauses = simplify(clauses, var)
    new_assignment = dict(assignment)
    new_assignment[var] = True
    sat, result = dpll(new_clauses, new_assignment)
    if sat:
        return True, result
    
    # Try False
    new_clauses = simplify(clauses, -var)
    new_assignment = dict(assignment)
    new_assignment[var] = False
    sat, result = dpll(new_clauses, new_assignment)
    if sat:
        return True, result
    
    return False, None

def verify_assignment(clauses, assignment):
    """Check if assignment satisfies all clauses."""
    for clause in clauses:
        satisfied = False
        for lit in clause:
            var = abs(lit)
            val = assignment.get(var, False)
            if (lit > 0 and val) or (lit < 0 and not val):
                satisfied = True
                break
        if not satisfied:
            return False
    return True


# === (d) Test SAT instances ===
print("SAT Solver Tests")
print("=" * 55)

# SAT: (x1 ∨ x2) ∧ (¬x1 ∨ x3) ∧ (¬x2 ∨ ¬x3)
sat1 = [[1, 2], [-1, 3], [-2, -3]]
result, assignment = dpll(sat1)
assert result == True
assert verify_assignment(sat1, assignment)
print(f"  Test 1 (SAT):   {result}, assignment={assignment}")

# SAT: (x1) ∧ (¬x1 ∨ x2) ∧ (¬x2 ∨ x3)  — chain implication
sat2 = [[1], [-1, 2], [-2, 3]]
result, assignment = dpll(sat2)
assert result == True
assert verify_assignment(sat2, assignment)
assert assignment[1] == True and assignment[2] == True and assignment[3] == True
print(f"  Test 2 (SAT):   {result}, assignment={assignment}")

# UNSAT: (x1) ∧ (¬x1)
unsat1 = [[1], [-1]]
result, assignment = dpll(unsat1)
assert result == False
print(f"  Test 3 (UNSAT): {result}")

# UNSAT: (x1 ∨ x2) ∧ (x1 ∨ ¬x2) ∧ (¬x1 ∨ x2) ∧ (¬x1 ∨ ¬x2)
unsat2 = [[1, 2], [1, -2], [-1, 2], [-1, -2]]
result, assignment = dpll(unsat2)
assert result == False
print(f"  Test 4 (UNSAT): {result}")

# Larger SAT instance
sat3 = [[1, 2, 3], [-1, -2], [-2, -3], [1, 3], [-1, 2, -3]]
result, assignment = dpll(sat3)
assert result == True
assert verify_assignment(sat3, assignment)
print(f"  Test 5 (SAT):   {result}, assignment={assignment}")

# === (e) Encode constraint problem as SAT ===
# Problem: Color a 3-node triangle with 2 colors such that no adjacent nodes share a color
# Variables: x_i = True means node i gets color 1, False means color 2
# Constraint: adjacent nodes must differ → (x_i ∨ x_j) ∧ (¬x_i ∨ ¬x_j) for each edge
# Triangle edges: (1,2), (2,3), (1,3)
print(f"\nGraph Coloring as SAT")
print("=" * 55)

# 2-coloring a triangle (should be UNSAT — odd cycle needs 3 colors)
# x_i ≠ x_j  ↔  (x_i ∨ x_j) ∧ (¬x_i ∨ ¬x_j)
triangle_2color = [
    [1, 2], [-1, -2],    # edge (1,2): must differ
    [2, 3], [-2, -3],    # edge (2,3): must differ
    [1, 3], [-1, -3],    # edge (1,3): must differ
]
result, _ = dpll(triangle_2color)
assert result == False
print(f"  2-color triangle: {result} (odd cycle needs ≥3 colors) ✓")

# 2-coloring a path (should be SAT)
path_2color = [
    [1, 2], [-1, -2],    # edge (1,2)
    [2, 3], [-2, -3],    # edge (2,3)
]
result, assignment = dpll(path_2color)
assert result == True
colors = {v: 'R' if assignment.get(v, False) else 'B' for v in [1,2,3]}
print(f"  2-color path:    {result}, coloring={colors} ✓")

print(f'\nAll tests passed ✓')
print(f'→ SAT solvers underpin constraint satisfaction in AI planning & verification')
```

```python
# Exercise 17: Solution ✅

class DFA:
    def __init__(self, states, alphabet, transitions, start, accept):
        self.states = set(states)
        self.alphabet = set(alphabet)
        self.transitions = dict(transitions)
        self.start = start
        self.accept = set(accept)
    
    def accepts(self, string: str) -> bool:
        state = self.start
        for symbol in string:
            if symbol not in self.alphabet:
                return False
            key = (state, symbol)
            if key not in self.transitions:
                return False
            state = self.transitions[key]
        return state in self.accept
    
    def complement(self):
        new_accept = self.states - self.accept
        return DFA(self.states, self.alphabet, self.transitions, self.start, new_accept)
    
    def enumerate_accepted(self, max_len: int) -> set:
        accepted = set()
        # BFS over all strings up to max_len
        from collections import deque
        queue = deque([("", self.start)])
        while queue:
            string, state = queue.popleft()
            if state in self.accept:
                accepted.add(string)
            if len(string) < max_len:
                for symbol in sorted(self.alphabet):
                    key = (state, symbol)
                    if key in self.transitions:
                        queue.append((string + symbol, self.transitions[key]))
        return accepted


# === (b) DFA: binary strings divisible by 3 ===
# States represent remainder mod 3: {0, 1, 2}
# Reading binary left-to-right: new_value = 2*old + bit
# So new_remainder = (2*old_remainder + bit) mod 3
div3_transitions = {}
for state in range(3):
    for bit in ['0', '1']:
        div3_transitions[(state, bit)] = (2 * state + int(bit)) % 3

div3_dfa = DFA(
    states={0, 1, 2},
    alphabet={'0', '1'},
    transitions=div3_transitions,
    start=0,
    accept={0}
)

print("DFA: Binary Strings Divisible by 3")
print("=" * 50)

# Test specific values
test_cases = [
    ("", True),    # empty = 0, divisible by 3
    ("0", True),   # 0
    ("11", True),  # 3
    ("110", True), # 6
    ("1001", True),# 9
    ("1", False),  # 1
    ("10", False), # 2
    ("100", False),# 4
    ("101", False),# 5
]

for binary, expected in test_cases:
    result = div3_dfa.accepts(binary)
    val = int(binary, 2) if binary else 0
    status = "✓" if result == expected else "✗"
    print(f"  '{binary}' (={val:2d}): {result:5} {status}")
    assert result == expected

# === (c) Enumerate all accepted strings up to length 6 ===
accepted = div3_dfa.enumerate_accepted(6)
print(f"\nAccepted strings (len ≤ 6): {len(accepted)} strings")
for s in sorted(accepted, key=lambda x: (len(x), x)):
    val = int(s, 2) if s else 0
    if val <= 20:
        print(f"  '{s}' = {val}", end="")
        assert val % 3 == 0
print()

# Verify ALL accepted strings are divisible by 3
for s in accepted:
    val = int(s, 2) if s else 0
    assert val % 3 == 0, f"'{s}' = {val} not divisible by 3"
print(f"  All {len(accepted)} strings verified divisible by 3: ✓")

# === (d) Complement DFA ===
not_div3 = div3_dfa.complement()
print(f"\nComplement DFA (NOT divisible by 3)")
print("=" * 50)

# === (e) Verify closure: L(complement) = Σ* \ L(M) ===
all_strings = set()
for length in range(7):
    for bits in itertools.product('01', repeat=length):
        all_strings.add(''.join(bits))

accepted_orig = {s for s in all_strings if div3_dfa.accepts(s)}
accepted_comp = {s for s in all_strings if not_div3.accepts(s)}

assert accepted_orig | accepted_comp == all_strings, "Should cover Σ*"
assert accepted_orig & accepted_comp == set(), "Should be disjoint"
assert accepted_comp == all_strings - accepted_orig, "Complement = Σ* \\ L"

print(f"  |Σ*| (len ≤ 6):        {len(all_strings)}")
print(f"  |L(M)| (div by 3):     {len(accepted_orig)}")
print(f"  |L(M̄)| (not div by 3): {len(accepted_comp)}")
print(f"  L(M) ∪ L(M̄) = Σ*:     ✓")
print(f"  L(M) ∩ L(M̄) = ∅:      ✓")
print(f"  L(M̄) = Σ* \\ L(M):     ✓")

print(f'\nAll tests passed ✓')
print(f'→ DFAs formalize tokenizer patterns & constrained decoding in LLMs')
```

```python
# Exercise 20: Solution ✅

def generate_dataset(n_samples: int, n_features: int, seed: int = 42):
    rng = np.random.RandomState(seed)
    feature_names = [f"f{i}" for i in range(n_features)]
    X = rng.randn(n_samples, n_features)
    # Label based on features 0 and 1 (with noise)
    signal = 0.8 * X[:, 0] + 0.5 * X[:, 1] + 0.2 * rng.randn(n_samples)
    y = (signal > 0).astype(int)
    return X, y, feature_names

def feature_subset_scores(X_train, y_train, feature_names):
    n_features = len(feature_names)
    results = []
    
    for r in range(1, n_features + 1):
        for subset in itertools.combinations(range(n_features), r):
            # Simple scoring: for each feature in subset, find best threshold
            best_acc = 0
            X_sub = X_train[:, list(subset)]
            # Use mean of each feature as threshold, predict by majority vote
            thresholds = X_sub.mean(axis=0)
            # For each sample: predict 1 if any feature > threshold
            preds = np.any(X_sub > thresholds, axis=1).astype(int)
            acc = np.mean(preds == y_train)
            # Also try: predict 1 if mean of features > 0
            preds2 = (X_sub.mean(axis=1) > 0).astype(int)
            acc2 = np.mean(preds2 == y_train)
            best_acc = max(acc, acc2)
            
            names = tuple(feature_names[i] for i in subset)
            results.append((names, subset, best_acc))
    
    results.sort(key=lambda x: -x[2])
    return results

def threshold_classifier(X, feature_indices, threshold=0.0):
    X_sub = X[:, list(feature_indices)]
    return (X_sub.mean(axis=1) > threshold).astype(int)

def full_evaluation(y_true, y_pred, classes):
    results = {}
    n = len(y_true)
    
    for c in classes:
        true_set = set(np.where(y_true == c)[0])
        pred_set = set(np.where(y_pred == c)[0])
        
        tp = len(true_set & pred_set)
        fp = len(pred_set - true_set)
        fn = len(true_set - pred_set)
        tn = n - tp - fp - fn
        
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
        
        # Jaccard of true vs predicted sets for this class
        jac = len(true_set & pred_set) / len(true_set | pred_set) if len(true_set | pred_set) > 0 else 1
        
        results[c] = {
            'TP': tp, 'FP': fp, 'FN': fn, 'TN': tn,
            'precision': precision, 'recall': recall, 'f1': f1,
            'jaccard': jac
        }
    
    # Macro averages
    results['macro_precision'] = np.mean([results[c]['precision'] for c in classes])
    results['macro_recall'] = np.mean([results[c]['recall'] for c in classes])
    results['macro_f1'] = np.mean([results[c]['f1'] for c in classes])
    results['accuracy'] = np.mean(y_true == y_pred)
    
    return results


# === Pipeline ===
print("End-to-End Set Theory ML Pipeline")
print("=" * 60)

# (a) Generate and partition data
X, y, feature_names = generate_dataset(200, 5, seed=42)
indices = set(range(len(X)))
splits = split_dataset(indices, {'train': 0.7, 'val': 0.15, 'test': 0.15}, seed=42)
partition_result = verify_partition(splits, indices)
assert partition_result['is_partition'], "Must be a valid partition"

print(f"(a) Dataset: {len(X)} samples, {len(feature_names)} features")
print(f"    Features: {feature_names}")
print(f"    Class distribution: {dict(zip(*np.unique(y, return_counts=True)))}")
for name, s in splits.items():
    print(f"    {name:6s}: {len(s):3d} samples")
print(f"    Valid partition: ✓")

# Split data
train_idx = sorted(splits['train'])
val_idx = sorted(splits['val'])
test_idx = sorted(splits['test'])
X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val = X[val_idx], y[val_idx]
X_test, y_test = X[test_idx], y[test_idx]

# (b) Feature subset enumeration via power set
print(f"\n(b) Feature Selection via Power Set")
print("=" * 60)
# Limit to subsets of size ≤ 3 to keep it tractable
scored = feature_subset_scores(X_train, y_train, feature_names)
print(f"    Total subsets evaluated: {len(scored)}")
print(f"\n    Top 5 feature subsets:")
for names, indices_sub, acc in scored[:5]:
    print(f"      {str(names):30s} → train acc = {acc:.3f}")
print(f"\n    Bottom 3 feature subsets:")
for names, indices_sub, acc in scored[-3:]:
    print(f"      {str(names):30s} → train acc = {acc:.3f}")

# (c) Train with best features
best_names, best_indices, best_train_acc = scored[0]
print(f"\n(c) Best Features: {best_names}")
print("=" * 60)

y_val_pred = threshold_classifier(X_val, best_indices)
y_test_pred = threshold_classifier(X_test, best_indices)
val_acc = np.mean(y_val_pred == y_val)
test_acc = np.mean(y_test_pred == y_test)
print(f"    Train accuracy: {best_train_acc:.3f}")
print(f"    Val accuracy:   {val_acc:.3f}")
print(f"    Test accuracy:  {test_acc:.3f}")

# (d) Set-based evaluation
print(f"\n(d) Set-Based Confusion Matrix")
print("=" * 60)
eval_result = full_evaluation(y_test, y_test_pred, classes=[0, 1])

for c in [0, 1]:
    m = eval_result[c]
    print(f"    Class {c}: TP={m['TP']:3d}  FP={m['FP']:3d}  FN={m['FN']:3d}  TN={m['TN']:3d}")
    print(f"             P={m['precision']:.3f}  R={m['recall']:.3f}  F1={m['f1']:.3f}  J={m['jaccard']:.3f}")

print(f"\n    Macro:   P={eval_result['macro_precision']:.3f}  "
      f"R={eval_result['macro_recall']:.3f}  F1={eval_result['macro_f1']:.3f}")
print(f"    Accuracy: {eval_result['accuracy']:.3f}")

# Verify TP+FP+FN+TN = N for each class
n_test = len(y_test)
for c in [0, 1]:
    m = eval_result[c]
    assert m['TP'] + m['FP'] + m['FN'] + m['TN'] == n_test

# (e) Jaccard comparison across feature subsets
print(f"\n(e) Prediction Similarity (Jaccard)")
print("=" * 60)
# Compare predictions from top 5 feature subsets
top_subsets = scored[:5]
pred_sets = {}
for names, idx_sub, _ in top_subsets:
    preds = threshold_classifier(X_test, idx_sub)
    pred_sets[names] = set(np.where(preds == 1)[0])

print(f"    Pairwise Jaccard of positive-class predictions:")
subset_names = list(pred_sets.keys())
for i in range(len(subset_names)):
    for j in range(i+1, len(subset_names)):
        s1 = pred_sets[subset_names[i]]
        s2 = pred_sets[subset_names[j]]
        j_sim = jaccard(s1, s2)
        print(f"      {str(subset_names[i]):25s} ↔ {str(subset_names[j]):25s}  J={j_sim:.3f}")

print(f'\n{"=" * 60}')
print(f'Pipeline complete — all set theory concepts applied ✓')
print(f'  ✓ Partitions (data splits)')
print(f'  ✓ Power sets (feature selection)')
print(f'  ✓ Set operations (confusion matrix)')
print(f'  ✓ Jaccard similarity (prediction comparison)')
print(f'  ✓ Verification (partition validity, De Morgan\'s)')
print(f'\n→ Set theory provides the mathematical backbone of every ML pipeline')
```
