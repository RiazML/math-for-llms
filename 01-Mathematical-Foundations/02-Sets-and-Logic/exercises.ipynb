## Exercise 1: Set Operations

Given the following sets representing features selected by different methods:
- `method_a = {1, 2, 3, 5, 8, 13}`
- `method_b = {2, 3, 5, 7, 11, 13}`
- `method_c = {1, 3, 5, 9, 13}`

Calculate:
1. Features selected by ALL methods (consensus)
2. Features selected by ANY method
3. Features unique to method_a
4. Features in method_a OR method_b, but NOT method_c

```python
# Exercise 1 - Your Solution
method_a = {1, 2, 3, 5, 8, 13}
method_b = {2, 3, 5, 7, 11, 13}
method_c = {1, 3, 5, 9, 13}

# 1. Consensus (all methods)
consensus = method_a & method_b & method_c
print(f"1. Consensus: {consensus}")

# 2. Any method
any_method = method_a | method_b | method_c
print(f"2. Any method: {any_method}")

# 3. Unique to method_a
unique_a = method_a - method_b - method_c
print(f"3. Unique to A: {unique_a}")

# 4. (A ∪ B) - C
result = (method_a | method_b) - method_c
print(f"4. (A ∪ B) - C: {result}")
```

## Exercise 2: Confusion Matrix

A medical diagnosis model predicts disease for 20 patients:
- `actual_sick = {0, 3, 5, 8, 12, 15, 18}`
- `predicted_sick = {0, 2, 3, 5, 7, 15, 17, 18}`

Calculate using set operations:
1. True Positives (TP)
2. False Positives (FP)
3. False Negatives (FN)
4. Precision = TP / (TP + FP)
5. Recall = TP / (TP + FN)

```python
# Exercise 2 - Your Solution
actual_sick = {0, 3, 5, 8, 12, 15, 18}
predicted_sick = {0, 2, 3, 5, 7, 15, 17, 18}

# 1. True Positives
TP = actual_sick & predicted_sick
print(f"1. TP: {TP} (count: {len(TP)})")

# 2. False Positives
FP = predicted_sick - actual_sick
print(f"2. FP: {FP} (count: {len(FP)})")

# 3. False Negatives
FN = actual_sick - predicted_sick
print(f"3. FN: {FN} (count: {len(FN)})")

# 4. Precision
precision = len(TP) / len(predicted_sick)
print(f"4. Precision: {precision:.3f}")

# 5. Recall
recall = len(TP) / len(actual_sick)
print(f"5. Recall: {recall:.3f}")
```

## Exercise 3: Data Split Validation

You have a dataset with indices 0-99 split as follows:
- `train = set(range(0, 70))`
- `val = set(range(65, 85))`
- `test = set(range(80, 100))`

Tasks:
1. Identify any data leakage (overlapping indices)
2. Find missing indices (not in any split)
3. Fix the splits to be non-overlapping with full coverage

```python
# Exercise 3 - Your Solution
train = set(range(0, 70))
val = set(range(65, 85))
test = set(range(80, 100))
all_indices = set(range(100))

# 1. Find overlaps (data leakage!)
train_val_overlap = train & val
val_test_overlap = val & test
print(f"1. Train-Val overlap: {train_val_overlap}")
print(f"   Val-Test overlap: {val_test_overlap}")

# 2. Find missing indices
covered = train | val | test
missing = all_indices - covered
print(f"2. Missing indices: {missing}")

# 3. Fix the splits
train_fixed = set(range(0, 70))
val_fixed = set(range(70, 85))
test_fixed = set(range(85, 100))
print(f"\n3. Fixed splits:")
print(f"   Train: 0-69, Val: 70-84, Test: 85-99")
print(f"   Coverage: {train_fixed | val_fixed | test_fixed == all_indices}")
```

## Exercise 4: Jaccard Similarity

Implement Jaccard similarity and use it for:
1. Calculate similarity between user purchase histories:
   - `user1 = {"laptop", "mouse", "keyboard", "monitor"}`
   - `user2 = {"laptop", "headphones", "keyboard", "webcam"}`
   - `user3 = {"phone", "case", "charger"}`

2. Which pair of users is most similar?
3. What products would you recommend to user1 based on user2?

```python
# Exercise 4 - Your Solution
def jaccard(a, b):
    return len(a & b) / len(a | b) if len(a | b) > 0 else 0

user1 = {"laptop", "mouse", "keyboard", "monitor"}
user2 = {"laptop", "headphones", "keyboard", "webcam"}
user3 = {"phone", "case", "charger"}

# 1. Calculate similarities
sim_12 = jaccard(user1, user2)
sim_13 = jaccard(user1, user3)
sim_23 = jaccard(user2, user3)

print(f"1. Similarities:")
print(f"   User1-User2: {sim_12:.3f}")
print(f"   User1-User3: {sim_13:.3f}")
print(f"   User2-User3: {sim_23:.3f}")

# 2. Most similar pair
print(f"\n2. Most similar: User1 & User2 ({sim_12:.3f})")

# 3. Recommendations (items user2 has but user1 doesn't)
recommendations = user2 - user1
print(f"\n3. Recommend to User1: {recommendations}")
```

## Exercise 5: De Morgan's Laws

Verify De Morgan's laws with sets:
- Universal set: `U = set(range(20))`
- `A = {0, 2, 4, 6, 8}` (even < 10)
- `B = {0, 5, 10, 15}` (multiples of 5)

Prove:
1. NOT(A AND B) = NOT(A) OR NOT(B)
2. NOT(A OR B) = NOT(A) AND NOT(B)

(Use `U - set` for complement)

```python
# Exercise 5 - Your Solution
U = set(range(20))
A = {0, 2, 4, 6, 8}
B = {0, 5, 10, 15}

# Complements
not_A = U - A
not_B = U - B

# Law 1: NOT(A ∩ B) = NOT(A) ∪ NOT(B)
lhs1 = U - (A & B)
rhs1 = not_A | not_B
print("1. NOT(A ∩ B) = NOT(A) ∪ NOT(B)")
print(f"   LHS: {lhs1}")
print(f"   RHS: {rhs1}")
print(f"   Equal: {lhs1 == rhs1} ✓")

# Law 2: NOT(A ∪ B) = NOT(A) ∩ NOT(B)
lhs2 = U - (A | B)
rhs2 = not_A & not_B
print("\n2. NOT(A ∪ B) = NOT(A) ∩ NOT(B)")
print(f"   LHS: {lhs2}")
print(f"   RHS: {rhs2}")
print(f"   Equal: {lhs2 == rhs2} ✓")
```

## Exercise 6: Boolean Logic Neural Network

Implement XOR using only AND, OR, NOT operations:
- XOR(a, b) = (a OR b) AND NOT(a AND b)

Build truth table and verify for all inputs.

```python
# Exercise 6 - Your Solution
def AND(a, b): return a and b
def OR(a, b): return a or b
def NOT(a): return not a

def XOR(a, b):
    """XOR = (a OR b) AND NOT(a AND b)"""
    return AND(OR(a, b), NOT(AND(a, b)))

# Truth table
print("XOR Truth Table:")
print("a | b | XOR(a,b)")
print("-" * 17)
for a in [False, True]:
    for b in [False, True]:
        result = XOR(a, b)
        print(f"{int(a)} | {int(b)} |    {int(result)}")

# Verify XOR property: True only when inputs differ
print("\n✓ XOR is True only when a ≠ b")
```
