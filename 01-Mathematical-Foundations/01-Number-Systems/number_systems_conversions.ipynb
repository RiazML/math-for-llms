# Number System Conversions in Python

This notebook demonstrates how to convert numbers between decimal, binary, octal, and hexadecimal systems using Python. It also includes a custom function for base conversion and an interactive widget for hands-on exploration.

## 1. Import Required Libraries
We will use Python's built-in functions and `ipywidgets` for interactivity.

```python
import ipywidgets as widgets
from IPython.display import display
```

## 2. Convert Decimal to Binary
Use Python's built-in `bin()` function to convert decimal numbers to binary representation.

```python
# Convert decimal numbers to binary
for dec in [5, 10, 42, 255]:
    print(f"Decimal: {dec} -> Binary: {bin(dec)}")
```

## 3. Convert Decimal to Octal
Use Python's built-in `oct()` function to convert decimal numbers to octal representation.

```python
# Convert decimal numbers to octal
for dec in [5, 10, 42, 255]:
    print(f"Decimal: {dec} -> Octal: {oct(dec)}")
```

## 4. Convert Decimal to Hexadecimal
Use Python's built-in `hex()` function to convert decimal numbers to hexadecimal representation.

```python
# Convert decimal numbers to hexadecimal
for dec in [5, 10, 42, 255]:
    print(f"Decimal: {dec} -> Hexadecimal: {hex(dec)}")
```

## 5. Convert Binary to Decimal
Use the `int()` function with base 2 to convert binary strings to decimal numbers.

```python
# Convert binary strings to decimal
binaries = ['101', '1010', '101010', '11111111']
for b in binaries:
    print(f"Binary: {b} -> Decimal: {int(b, 2)}")
```

## 6. Convert Octal to Decimal
Use the `int()` function with base 8 to convert octal strings to decimal numbers.

```python
# Convert octal strings to decimal
octals = ['7', '12', '52', '377']
for o in octals:
    print(f"Octal: {o} -> Decimal: {int(o, 8)}")
```

## 7. Convert Hexadecimal to Decimal
Use the `int()` function with base 16 to convert hexadecimal strings to decimal numbers.

```python
# Convert hexadecimal strings to decimal
hexes = ['a', '1f', '2a', 'ff']
for h in hexes:
    print(f"Hexadecimal: {h} -> Decimal: {int(h, 16)}")
```

## 8. Custom Function for Base Conversion
This function converts a number from any base to any other base.

```python
def base_convert(number_str, from_base, to_base):
    """Convert a number string from one base to another."""
    decimal = int(number_str, from_base)
    if to_base == 2:
        return bin(decimal)
    elif to_base == 8:
        return oct(decimal)
    elif to_base == 16:
        return hex(decimal)
    else:
        # General conversion for bases > 10
        digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        if decimal == 0:
            return '0'
        result = ''
        while decimal > 0:
            result = digits[decimal % to_base] + result
            decimal //= to_base
        return result

# Examples:
print(base_convert('1010', 2, 10))  # binary to decimal
print(base_convert('ff', 16, 2))    # hex to binary
print(base_convert('77', 8, 16))    # octal to hex
print(base_convert('255', 10, 2))   # decimal to binary
```

## 9. Interactive Base Conversion Example
Use `ipywidgets` to create an interactive widget for base conversion.

```python
def interactive_base_convert():
    number = widgets.Text(value='1010', description='Number:')
    from_base = widgets.Dropdown(options=[2, 8, 10, 16], value=2, description='From base:')
    to_base = widgets.Dropdown(options=[2, 8, 10, 16], value=10, description='To base:')
    output = widgets.Output()

    def on_change(change=None):
        with output:
            output.clear_output()
            try:
                result = base_convert(number.value, from_base.value, to_base.value)
                print(f"Result: {result}")
            except Exception as e:
                print(f"Error: {e}")

    number.observe(on_change, names='value')
    from_base.observe(on_change, names='value')
    to_base.observe(on_change, names='value')
    display(widgets.HBox([number, from_base, to_base]))
    display(output)
    on_change()

interactive_base_convert()
```
