## 1. Direct Proof

**Structure:** Assume hypothesis P → chain of logical steps → arrive at conclusion Q

```
P (Hypothesis) → Step 1 → Step 2 → ... → Q (Conclusion)
```

### Theorem: Sum of Two Even Numbers is Even

**Proof:**
1. Let a and b be even numbers (hypothesis)
2. By definition: a = 2m and b = 2n for some integers m, n
3. Then a + b = 2m + 2n = 2(m + n)
4. Since (m + n) is an integer, a + b has form 2k
5. By definition, a + b is even ∎

````python
# Verification: Sum of two even numbers is even
def is_even(n):
    return n % 2 == 0

test_pairs = [(4, 6), (10, 22), (100, 200), (-8, 14)]

print("Testing: Sum of two even numbers is even")
print("=" * 45)
for a, b in test_pairs:
    result = a + b
    print(f"{a:4d} + {b:4d} = {result:4d}  (even: {is_even(result)})")
````

### Theorem: Product of Two Odd Numbers is Odd

**Proof:**
1. Let a and b be odd numbers
2. Then a = 2m + 1 and b = 2n + 1 for some integers m, n
3. a · b = (2m + 1)(2n + 1)
4. = 4mn + 2m + 2n + 1
5. = 2(2mn + m + n) + 1
6. This has form 2k + 1, so a · b is odd ∎

````python
# Verification: Product of two odd numbers is odd
def is_odd(n):
    return n % 2 == 1

test_pairs = [(3, 5), (7, 9), (11, 13), (-3, 7)]

print("Testing: Product of two odd numbers is odd")
print("=" * 45)
for a, b in test_pairs:
    result = a * b
    print(f"{a:4d} × {b:4d} = {result:4d}  (odd: {is_odd(result)})")
````

## 2. Proof by Contradiction

**Structure:** Assume the statement is FALSE → derive a contradiction → original must be TRUE

```
Assume ¬Q (Q is false)
    ↓
Logical steps
    ↓
Arrive at contradiction (P ∧ ¬P)
    ↓
Therefore Q must be true
```

### Theorem: √2 is Irrational

**Proof by contradiction:**
1. Assume √2 is rational (negation of what we want)
2. Then √2 = p/q where p, q are integers with gcd(p,q) = 1 (lowest terms)
3. Squaring: 2 = p²/q², so p² = 2q²
4. Therefore p² is even, so p must be even
5. Let p = 2k, then 4k² = 2q², so q² = 2k²
6. Therefore q² is even, so q must be even
7. But both p and q being even contradicts gcd(p,q) = 1
8. **Contradiction!** Therefore √2 is irrational ∎

````python
# Numerical evidence: √2 has no repeating decimal pattern
sqrt2 = sqrt(2)
print(f"√2 = {sqrt2:.50f}...")
print("\nThe decimal never repeats (irrational numbers don't have repeating decimals)")

# Show that no fraction perfectly equals √2
print("\nBest rational approximations:")
print("="*40)
approximations = [(1, 1), (3, 2), (7, 5), (17, 12), (41, 29), (99, 70), (239, 169)]
for p, q in approximations:
    approx = p / q
    error = abs(approx - sqrt2)
    print(f"{p:3d}/{q:3d} = {approx:.10f}  (error: {error:.2e})")
````

### Theorem: There is No Largest Integer

**Proof by contradiction:**
1. Assume N is the largest integer
2. Consider N + 1
3. N + 1 > N and N + 1 is an integer
4. This contradicts N being the largest
5. Therefore no largest integer exists ∎

````python
# Demonstration: For any N, we can always find N+1
def find_larger(N):
    """Given any integer N, return a larger integer."""
    return N + 1

N = 1000000000000  # A trillion
print(f"If N = {N:,}")
print(f"Then N+1 = {find_larger(N):,}")
print(f"N+1 > N? {find_larger(N) > N}")
````

## 3. Proof by Contrapositive

**Key insight:** To prove "P → Q", we can equivalently prove "¬Q → ¬P"

```
Logical Equivalence:

P → Q  ≡  ¬Q → ¬P

"If it's raining, the ground is wet"
                ≡
"If the ground is NOT wet, it's NOT raining"
```

### Theorem: If n² is odd, then n is odd

**Original:** n² odd → n odd

**Contrapositive:** n even → n² even (easier to prove!)

**Proof of contrapositive:**
1. Assume n is even
2. Then n = 2k for some integer k
3. n² = (2k)² = 4k² = 2(2k²)
4. Since 2k² is an integer, n² is even ∎

````python
# Verification: Relationship between n and n² oddness
print("Verifying: n² odd ⟹ n odd (equivalently: n even ⟹ n² even)")
print("=" * 55)
print(f"{'n':>3} │ {'n²':>4} │ n odd │ n² odd │ Consistent")
print("-" * 55)

for n in range(1, 13):
    n_sq = n ** 2
    n_odd = n % 2 == 1
    n2_odd = n_sq % 2 == 1
    # If n² is odd, n should also be odd
    consistent = (not n2_odd) or n_odd  # n² odd implies n odd
    print(f"{n:3d} │ {n_sq:4d} │ {str(n_odd):>5} │ {str(n2_odd):>6} │ {consistent}")
````

### Theorem: If xy is odd, then both x and y are odd

**Contrapositive:** If x is even OR y is even, then xy is even

**Proof:**
1. Assume x is even (WLOG - without loss of generality)
2. Then x = 2k for some integer k
3. xy = 2ky
4. Since ky is an integer, xy is even ∎

````python
# Verification: If xy is odd, both x and y must be odd
print("Verifying: xy odd ⟹ (x odd AND y odd)")
print("=" * 55)

test_cases = [(3, 5), (3, 4), (4, 5), (4, 6), (7, 9), (2, 7)]
for x, y in test_cases:
    product = x * y
    x_odd = x % 2 == 1
    y_odd = y % 2 == 1
    prod_odd = product % 2 == 1
    both_odd = x_odd and y_odd
    print(f"{x} × {y} = {product:3d} | x odd: {x_odd}, y odd: {y_odd} | "
          f"Product odd: {prod_odd} | Both odd: {both_odd}")
````

## 4. Mathematical Induction

**Structure:** Prove a statement P(n) holds for all n ≥ n₀

```
┌────────────────┐     ┌──────────────────────┐
│  Base Case     │     │  Inductive Step      │
│  P(n₀) is true │ ──▶ │  P(k) ⟹ P(k+1)      │
└────────────────┘     └──────────────────────┘
                              ↓
                    P(n) true for all n ≥ n₀

Like dominoes: knock first one, each knocks the next
```

### Theorem: Σᵢ₌₁ⁿ i = n(n+1)/2

**Base case (n = 1):**
- LHS: Σᵢ₌₁¹ i = 1
- RHS: 1(2)/2 = 1 ✓

**Inductive hypothesis:** Assume Σᵢ₌₁ᵏ i = k(k+1)/2 for some k ≥ 1

**Inductive step (show for k+1):**
$$\sum_{i=1}^{k+1} i = \sum_{i=1}^{k} i + (k+1) = \frac{k(k+1)}{2} + (k+1)$$
$$= \frac{k(k+1) + 2(k+1)}{2} = \frac{(k+1)(k+2)}{2}$$

This is exactly the formula with n = k+1 ∎

````python
# Verification: Sum formula Σi = n(n+1)/2
def sum_direct(n):
    """Compute Σᵢ₌₁ⁿ i directly."""
    return sum(range(1, n + 1))

def sum_formula(n):
    """Compute using closed-form formula."""
    return n * (n + 1) // 2

print("Verifying: Σᵢ₌₁ⁿ i = n(n+1)/2")
print("=" * 45)
print(f"{'n':>5} │ {'Direct Sum':>12} │ {'Formula':>12} │ Equal")
print("-" * 45)

for n in [1, 5, 10, 50, 100, 1000]:
    direct = sum_direct(n)
    formula = sum_formula(n)
    print(f"{n:5d} │ {direct:12,d} │ {formula:12,d} │ {direct == formula}")
````

### Theorem: For n ≥ 3, 2ⁿ > 2n + 1

**Base case (n = 3):**
- LHS: 2³ = 8
- RHS: 2(3) + 1 = 7
- 8 > 7 ✓

**Inductive hypothesis:** Assume 2ᵏ > 2k + 1 for some k ≥ 3

**Inductive step:**
$$2^{k+1} = 2 \cdot 2^k > 2(2k + 1) = 4k + 2$$

Need to show 4k + 2 > 2(k+1) + 1 = 2k + 3:
$$4k + 2 - (2k + 3) = 2k - 1$$

For k ≥ 3: 2k - 1 ≥ 5 > 0 ✓

````python
# Verification: 2ⁿ > 2n + 1 for n ≥ 3
print("Verifying: 2ⁿ > 2n + 1 for n ≥ 3")
print("=" * 50)
print(f"{'n':>3} │ {'2ⁿ':>8} │ {'2n+1':>6} │ {'Ratio':>8} │ 2ⁿ > 2n+1")
print("-" * 50)

for n in range(1, 15):
    lhs = 2 ** n
    rhs = 2 * n + 1
    ratio = lhs / rhs
    holds = lhs > rhs
    marker = "✓" if n >= 3 and holds else "" if holds else "✗"
    print(f"{n:3d} │ {lhs:8d} │ {rhs:6d} │ {ratio:8.2f} │ {holds} {marker}")
````

## 5. Strong Induction

**Difference from regular induction:**
- Regular: Assume P(k), prove P(k+1)
- Strong: Assume P(n₀), P(n₀+1), ..., P(k) ALL true, prove P(k+1)

**Use when:** P(k+1) depends on multiple previous cases.

### Theorem: Every integer n > 1 has a prime factorization

**Base case (n = 2):**
- 2 is prime, so its factorization is just {2} ✓

**Strong inductive hypothesis:**
- Assume every integer from 2 to k has a prime factorization

**Inductive step (show for k+1):**
- **Case 1:** k+1 is prime → factorization is {k+1} ✓
- **Case 2:** k+1 is composite → k+1 = a × b where 1 < a, b < k+1
  - By strong hypothesis, a and b have prime factorizations
  - Combining them gives k+1's factorization ✓ ∎

````python
def prime_factors(n):
    """Return the prime factorization of n."""
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
    return factors

print("Prime factorizations (every n > 1 has one):")
print("=" * 45)

test_numbers = [2, 6, 12, 28, 100, 127, 360, 1024, 9999]
for n in test_numbers:
    factors = prime_factors(n)
    factorization = ' × '.join(map(str, factors))
    product = np.prod(factors)
    print(f"{n:5d} = {factorization}")
    assert product == n, f"Product mismatch for {n}"
````

## 6. Proof by Cases

**Structure:** Divide into exhaustive cases, prove each separately.

```
Statement to prove: P

Case 1: If condition A → P holds
Case 2: If condition B → P holds
...
Since A, B, ... cover all possibilities, P is true.
```

### Theorem: |xy| = |x| · |y| for all real x, y

**Case 1:** x ≥ 0 and y ≥ 0
- |xy| = xy = |x| · |y| ✓

**Case 2:** x ≥ 0 and y < 0
- |xy| = |−xy| = −xy = x · (−y) = |x| · |y| ✓

**Case 3:** x < 0 and y ≥ 0
- |xy| = |−xy| = −xy = (−x) · y = |x| · |y| ✓

**Case 4:** x < 0 and y < 0
- |xy| = xy (product of negatives is positive) = (−x)(−y) = |x| · |y| ✓

All cases covered ∎

````python
# Verification: |xy| = |x| · |y| for all cases
print("Verifying: |xy| = |x| · |y|")
print("=" * 60)
print(f"{'x':>4} {'y':>4} │ {'Case':^18} │ {'|xy|':>5} {'|x||y|':>6} │ Equal")
print("-" * 60)

test_cases = [
    (3, 4, "x≥0, y≥0"),
    (3, -4, "x≥0, y<0"),
    (-3, 4, "x<0, y≥0"),
    (-3, -4, "x<0, y<0"),
    (0, 5, "x=0"),
    (-2, 0, "y=0"),
    (0, 0, "both zero"),
]

for x, y, case in test_cases:
    lhs = abs(x * y)
    rhs = abs(x) * abs(y)
    print(f"{x:4d} {y:4d} │ {case:^18} │ {lhs:5d} {rhs:6d} │ {lhs == rhs}")
````

## 7. Existence and Uniqueness Proofs

**Existence (∃):** Show at least one object with the property exists
- **Constructive:** Build an explicit example
- **Non-constructive:** Show existence without construction

**Uniqueness:** Show at most one object exists
- Assume two objects x and y both satisfy the property, prove x = y

### Theorem: Division Algorithm

For integers a ≥ 0 and d > 0, there exist **unique** integers q and r such that:
$$a = qd + r, \quad 0 \leq r < d$$

**Existence proof (constructive):**
- q = ⌊a/d⌋ (floor division)
- r = a - qd (remainder)
- By construction: a = qd + r and 0 ≤ r < d ✓

**Uniqueness proof:**
1. Suppose a = q₁d + r₁ = q₂d + r₂ with 0 ≤ r₁, r₂ < d
2. Then (q₁ - q₂)d = r₂ - r₁
3. Since |r₂ - r₁| < d and d|(r₂ - r₁), we must have r₁ = r₂
4. Therefore q₁ = q₂ ∎

````python
# Division algorithm demonstration
print("Division Algorithm: a = qd + r where 0 ≤ r < d")
print("=" * 55)
print(f"{'a':>6} {'d':>4} │ {'q':>4} {'r':>4} │ Verify: qd + r │ 0 ≤ r < d")
print("-" * 55)

test_cases = [(17, 5), (100, 7), (25, 4), (0, 3), (123, 11), (1000, 37)]
for a, d in test_cases:
    q, r = divmod(a, d)
    verify = q * d + r
    valid_r = 0 <= r < d
    print(f"{a:6d} {d:4d} │ {q:4d} {r:4d} │ {q}×{d}+{r} = {verify:6d} │ {valid_r}")
````

## 8. ML Application: Convergence Proof

### Theorem: Gradient Descent Convergence for Convex Functions

**Setup:**
- f: convex function, L-smooth (Lipschitz gradient)
- Update rule: x_{t+1} = x_t - η∇f(x_t) with η = 1/L

**Claim:** f(x_T) - f(x*) ≤ L||x₀ - x*||² / (2T)

**Proof sketch:**
1. By L-smoothness: f(x_{t+1}) ≤ f(x_t) - (1/2L)||∇f(x_t)||²
2. Function value decreases each iteration
3. By convexity and telescoping: convergence rate is O(1/T) ∎

````python
# Numerical demonstration: Gradient descent on f(x) = x²
def f(x):
    """Simple quadratic: f(x) = x²"""
    return x ** 2

def grad_f(x):
    """Gradient: f'(x) = 2x"""
    return 2 * x

# For f(x) = x², the Lipschitz constant L = 2
L = 2
eta = 1 / L  # Optimal learning rate
x = 10.0     # Starting point
x_star = 0.0 # Optimal point

print("Gradient Descent Convergence Demonstration")
print(f"f(x) = x², x₀ = {x}, x* = {x_star}")
print(f"Learning rate η = 1/L = {eta}")
print("=" * 55)
print(f"{'T':>4} │ {'x_t':>10} │ {'f(x_t)':>10} │ {'f(x_t) - f(x*)':>14}")
print("-" * 55)

for t in range(16):
    f_diff = f(x) - f(x_star)
    print(f"{t:4d} │ {x:10.6f} │ {f(x):10.6f} │ {f_diff:14.6f}")
    x = x - eta * grad_f(x)

print("\nObserve: Error decreases by factor of 4 each step (matches O(1/T) theory)")
````

## 9. The Well-Ordering Principle

**Principle:** Every non-empty set of positive integers has a smallest element.

This is equivalent to the principle of mathematical induction and can be used for alternative proofs.

### Alternative Proof: √2 is Irrational (Using Well-Ordering)

**Proof:**
1. Assume √2 = p/q where p, q are positive integers
2. Let S = {n ∈ ℤ⁺ : n√2 ∈ ℤ⁺}  (positive integers n where n√2 is also an integer)
3. Then q ∈ S (since q√2 = p)
4. By well-ordering, S has a smallest element m
5. Let m' = m√2 - m = m(√2 - 1)
6. Since 0 < √2 - 1 < 1, we have 0 < m' < m
7. Now m'√2 = m(√2 - 1)√2 = 2m - m√2, which is an integer
8. So m' ∈ S with m' < m, contradicting minimality of m
9. Therefore √2 is irrational ∎

````python
# Illustrating the well-ordering principle
import random

# Any non-empty set of positive integers has a minimum
sets = [
    [7, 3, 9, 1, 5],
    [100, 200, 50, 75],
    [42],  # Singleton set
    list(range(1, 101)),  # First 100 positive integers
]

print("Well-Ordering Principle: Every non-empty set of positive integers has a minimum")
print("=" * 70)

for s in sets:
    minimum = min(s)
    preview = str(s[:5])[:-1] + ", ...]" if len(s) > 5 else str(s)
    print(f"Set: {preview:<35} → Minimum: {minimum}")

print("\nThe minimum always exists for any non-empty set of positive integers.")
````

## 10. Summary: Choosing the Right Technique

| Situation | Technique |
|-----------|:----------|
| Show P implies Q directly | Direct proof |
| Q is hard to prove directly | Contrapositive (prove ¬Q ⟹ ¬P) |
| Statement seems impossible otherwise | Contradiction |
| Statement about all n ∈ ℕ | Induction |
| P(k+1) needs multiple previous cases | Strong induction |
| Natural division into scenarios | Proof by cases |
| "There exists" | Existence (constructive or non-constructive) |
| "There is exactly one" | Existence + Uniqueness |

````python
# Quick reference: Logic symbols
print("PROOF TECHNIQUES QUICK REFERENCE")
print("=" * 55)
print("""
Direct:        P ⟹ Q (chain of implications)
Contrapositive: ¬Q ⟹ ¬P (equivalent to P ⟹ Q)
Contradiction: Assume ¬S, derive P ∧ ¬P
Induction:     P(n₀) ∧ [P(k) ⟹ P(k+1)] ⟹ ∀n≥n₀: P(n)
Strong Ind:    P(n₀...k) ⟹ P(k+1)
Cases:         (C₁ ⟹ P) ∧ (C₂ ⟹ P) ∧ ... ∧ (C₁ ∨ C₂ ∨ ...) ⟹ P

COMMON MISTAKES:
• Induction: Forgetting the base case
• Contradiction: Not clearly stating the assumption
• Contrapositive: Confusing with converse (Q ⟹ P)
• Cases: Not covering all possibilities
• Existence: Claiming uniqueness without proof
""")
````
