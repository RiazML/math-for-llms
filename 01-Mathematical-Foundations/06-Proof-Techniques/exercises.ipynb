## Exercise 11: Binary Representation (Strong Induction) ⭐⭐

**Technique:** Strong Induction

**Theorem:** Every integer $n \geq 1$ can be expressed as a sum of distinct powers of 2.

**Proof structure:**
- **Base case** ($n = 1$): $1 = 2^0$ ✓
- **Strong inductive hypothesis:** Assume every integer $1 \leq m < n$ can be written as a sum of distinct powers of 2
- **Inductive step:**

```python
# Exercise 11: Solution ✅

def to_binary_representation_sol(n):
    """Express n as sum of distinct powers of 2 (binary decomposition)."""
    powers = set()
    bit_position = 0
    while n > 0:
        if n % 2 == 1:
            powers.add(bit_position)
        n //= 2
        bit_position += 1
    return powers

def build_strong_induction_proof_sol():
    proof = {
        'theorem': 'Every integer n ≥ 1 can be expressed as a sum of distinct powers of 2',
        'technique': 'Strong Induction',
        'base_case': 'n=1: 1 = 2^0 ✓',
        'strong_ih': 'Assume every integer 1 ≤ m < n has a binary representation',
        'even_case': ('n = 2k, k < n: by SIH k = Σ 2^{aᵢ}, '
                     'so n = Σ 2^{aᵢ+1} (shift all exponents up by 1, still distinct)'),
        'odd_case': ('n = 2k+1, k < n: by SIH k = Σ 2^{aᵢ} with min(aᵢ) ≥ 0, '
                    'so n = 2^0 + Σ 2^{aᵢ+1} (2^0 is new, shifted powers ≥ 2^1, all distinct)'),
        'conclusion': 'By strong induction, every n ≥ 1 has a unique binary representation. QED'
    }
    return proof

proof = build_strong_induction_proof_sol()

print('=== STRONG INDUCTION PROOF: Binary Representation ===')
for key, val in proof.items():
    print(f'  {key}: {val}')

# Verify for n = 1..50
print(f'\n=== Verification for n=1..50 ===')
all_correct = True
for n in range(1, 51):
    powers = to_binary_representation_sol(n)
    reconstructed = sum(2**p for p in powers)
    is_distinct = len(powers) == len(set(powers))  # always true for sets
    if reconstructed != n or not is_distinct:
        all_correct = False
        print(f'  FAIL at n={n}')
print(f'All correct: {all_correct} ✓')

# Show detailed decomposition
print(f'\n=== Detailed binary decomposition ===')
print(f'{"n":>4} | {"Binary":>10} | {"Powers of 2":>25} | {"Sum":>6}')
print('-' * 55)
for n in [1, 2, 3, 7, 10, 13, 42, 100, 255, 1024]:
    powers = to_binary_representation_sol(n)
    binary_str = bin(n)[2:]
    power_str = ' + '.join(f'2^{p}' for p in sorted(powers, reverse=True))
    reconstructed = sum(2**p for p in powers)
    print(f'{n:4d} | {binary_str:>10} | {power_str:>25} | {reconstructed:6d}')

# ML context: binary encoding of features
print('\n=== ML Context: Feature encoding ===')
print('Binary representation is the basis for:')
print('  - One-hot encoding (each category = power of 2)')
print('  - Binary feature hashing')
print('  - Quantized neural network weights (powers of 2)')
```
